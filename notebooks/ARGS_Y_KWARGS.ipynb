```python
# ===================================================================
# GUÍA DEFINITIVA DE PYTHON: SECCIÓN DE *ARGS Y **KWARGS
# Código completo y mejorado para Google Colab
# ===================================================================

from IPython.core.display import display, HTML

# --- 1. PRE-CÁLCULO DE LOS RESULTADOS DE CÓDIGO ---

# --- Sección 13: *args y **kwargs ---

# Ejemplo 1: *args
def sumar_todo(*numeros):
    # 'numeros' es una tupla con todos los argumentos posicionales extra.
    resultado_str = []
    resultado_str.append(f"Argumentos recibidos como tupla: {numeros}")
    resultado_str.append(f"La suma total es: {sum(numeros)}")
    return "\n".join(resultado_str)

output_s13_1 = sumar_todo(10, 20, 30, 5)

# Ejemplo 2: **kwargs
def crear_perfil_usuario(nombre, **datos_extra):
    # 'datos_extra' es un diccionario con todos los argumentos de palabra clave extra.
    resultado_str = []
    resultado_str.append(f"Perfil para: {nombre}")
    resultado_str.append("Datos adicionales:")
    for clave, valor in datos_extra.items():
        resultado_str.append(f"  - {clave.capitalize()}: {valor}")
    return "\n".join(resultado_str)

output_s13_2 = crear_perfil_usuario('Ana', edad=30, ciudad='Madrid', profesion='Ingeniera')

# Ejemplo 3: Combinando todo
def procesar_pedido(id_cliente, *productos, **opciones):
    resultado_str = []
    resultado_str.append(f"--- Procesando Pedido para Cliente: {id_cliente} ---")
    resultado_str.append(f"Productos en el carrito: {productos}")
    resultado_str.append("Opciones de envío:")
    opciones.setdefault('metodo', 'estándar') # Valor por defecto
    opciones.setdefault('urgente', False)
    for k, v in opciones.items():
        resultado_str.append(f"  - {k.capitalize()}: {v}")
    return "\n".join(resultado_str)

output_s13_3 = procesar_pedido('CLI-123', 'Laptop', 'Mouse', 'Teclado', metodo='exprés', regalo=True)

# Ejemplo 4: Desempaquetado
lista_nums = [15, 25, 60]
output_s13_4a = sumar_todo(*lista_nums) # Desempaqueta la lista en argumentos posicionales

dict_perfil = {'edad': 42, 'ciudad': 'Barcelona'}
output_s13_4b = crear_perfil_usuario('Carlos', **dict_perfil) # Desempaqueta el dict en argumentos de palabra clave


# --- 2. PLANTILLA HTML CON LA NUEVA SECCIÓN ---
html_python_guide_template = f"""
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Guía Definitiva de Python: De Cero a Experto</title>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">
  <style>
    :root {{
      --bg-gradient: linear-gradient(135deg, #e0eafc 0%, #cfdef3 100%); --bg-card: rgba(255, 255, 255, 0.7); --bg-content: rgba(248, 250, 252, 0.9); --text-primary: #2c3e50; --text-secondary: #5a6a7a; --text-header: #ffffff; --accent-primary: #3498db; --accent-secondary: #8e44ad; --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%); --border-color: rgba(0, 0, 0, 0.08); --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.07); --code-bg: #2d3748; --code-text: #e2e8f0; --code-output-bg: #1a202c;
    }}
    [data-theme="dark"] {{
      --bg-gradient: linear-gradient(135deg, #141E30 0%, #243B55 100%); --bg-card: rgba(26, 32, 44, 0.7); --bg-content: rgba(45, 55, 72, 0.8); --text-primary: #f7fafc; --text-secondary: #a0aec0; --text-header: #f7fafc; --accent-primary: #4eacfa; --accent-secondary: #c471ed; --border-color: rgba(255, 255, 255, 0.1); --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.2); --code-bg: #1a202c; --code-text: #f7fafc; --code-output-bg: #11151c;
    }}
    * {{ margin: 0; padding: 0; box-sizing: border-box; }}
    html {{ scroll-behavior: smooth; }}
    body {{ font-family: 'Inter', sans-serif; line-height: 1.8; background: var(--bg-gradient); color: var(--text-primary); transition: background 0.5s ease, color 0.5s ease; min-height: 100vh; }}
    .container {{ max-width: 1000px; margin: 0 auto; padding: 2rem; }}
    .header {{ text-align: center; margin-bottom: 3rem; }}
    .main-title {{ font-size: clamp(2.8rem, 5vw, 4.5rem); font-weight: 800; background: var(--accent-gradient); -webkit-background-clip: text; -webkit-text-fill-color: transparent; background-clip: text; margin-bottom: 1rem; }}
    .subtitle {{ font-size: 1.4rem; color: var(--text-header); font-weight: 400; opacity: 0.95; text-shadow: 0 2px 4px rgba(0, 0, 0, 0.3); max-width: 900px; margin: auto; }}
    .theme-toggle {{ position: fixed; top: 1.5rem; right: 1.5rem; width: 50px; height: 50px; border-radius: 50%; background: var(--bg-card); backdrop-filter: blur(10px); box-shadow: var(--shadow-card); border: 1px solid var(--border-color); cursor: pointer; display: flex; align-items: center; justify-content: center; font-size: 1.3rem; color: var(--accent-primary); transition: all 0.4s cubic-bezier(0.25, 0.8, 0.25, 1); z-index: 1000; }}
    .theme-toggle:hover {{ transform: scale(1.1) rotate(15deg); }}
    .lesson-container {{ display: flex; flex-direction: column; gap: 1.5rem; }}
    .topic-card {{ background: var(--bg-card); backdrop-filter: blur(15px); border-radius: 20px; box-shadow: var(--shadow-card); border: 1px solid var(--border-color); overflow: hidden; transition: all 0.4s ease; }}
    .topic-header {{ cursor: pointer; padding: 1.5rem 2rem; display: flex; justify-content: space-between; align-items: center; }}
    .topic-title {{ font-size: 1.3rem; font-weight: 600; color: var(--text-primary); }}
    .expand-icon {{ font-size: 1.2rem; color: var(--text-secondary); transition: transform 0.4s ease; }}
    .topic-card.open .expand-icon {{ transform: rotate(180deg); }}
    .topic-content {{ max-height: 0; overflow: hidden; transition: max-height 1s ease, padding 1s ease; background: var(--bg-content); }}
    .topic-card.open .topic-content {{ max-height: 4500px; padding: 1.5rem 2rem; border-top: 1px solid var(--border-color); }}
    .topic-content h3 {{ font-size: 1.2rem; color: var(--text-primary); margin: 1.5rem 0 1rem 0; padding-left: 1rem; border-left: 4px solid var(--accent-secondary); }}
    pre {{ background: var(--code-bg); color: var(--code-text); padding: 1.5rem; border-radius: 15px; border: 1px solid var(--border-color); overflow-x: auto; margin: 1.5rem 0; font-family: 'JetBrains Mono', monospace; font-size: 0.9em; }}
    .code-output {{ background: var(--code-output-bg); color: #e0e0e0; font-family: 'JetBrains Mono', monospace; padding: 1rem; border-radius: 0 0 15px 15px; margin-top: -1.5rem; border: 1px solid var(--border-color); border-top: none; white-space: pre; }}
    .tip-box {{ background: rgba(var(--accent-primary), 0.1); border-left: 4px solid var(--accent-primary); padding: 1rem; margin: 1.5rem 0; }}
  </style>
</head>
<body data-theme="light">
  <div class="theme-toggle" id="themeToggleButton" title="Cambiar tema"><i class="fas fa-moon" id="themeIcon"></i></div>
  <div class="container">
    <header class="header">
      <h1 class="main-title">Guía Definitiva de Python</h1>
      <p class="subtitle">Una inmersión interactiva en la filosofía, sintaxis y ecosistema de uno de los lenguajes de programación más influyentes del mundo.</p>
    </header>
    <div class="lesson-container">

        <!-- SECCIÓN 13: *ARGS Y **KWARGS -->
        <div class="topic-card">
            <div class="topic-header"><span class="topic-title">13. Argumentos Flexibles: *args y **kwargs</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <h3>¿Por qué necesitamos argumentos flexibles?</h3>
                <p>A veces, no sabemos de antemano cuántos argumentos necesitará recibir una función. Queremos crear funciones "genéricas" que puedan manejar un número variable de entradas. Aquí es donde entran en juego <code>*args</code> y <code>**kwargs</code>.</p>
                <p>Imagínalo como organizar una fiesta: tienes algunos invitados confirmados (argumentos normales), pero quieres permitir que traigan a sus amigos (<code>*args</code>) y que hagan peticiones especiales (<code>**kwargs</code>).</p>

                <h3><code>*args</code>: Múltiples Argumentos Posicionales</h3>
                <p>El asterisco simple (<code>*</code>) antes de un nombre de parámetro (por convención, <code>args</code>) empaqueta cualquier número de <strong>argumentos posicionales</strong> adicionales en una <strong>tupla</strong>.</p>
                <pre>def sumar_todo(*numeros):
    # 'numeros' es una tupla con todos los argumentos posicionales extra.
    print(f"Argumentos recibidos como tupla: {{numeros}}")
    return sum(numeros)

sumar_todo(10, 20, 30, 5)</pre>
                <div class="code-output">{output_s13_1}</div>

                <h3><code>**kwargs</code>: Múltiples Argumentos de Palabra Clave</h3>
                <p>El doble asterisco (<code>**</code>) antes de un nombre de parámetro (por convención, <code>kwargs</code>, de "keyword arguments") empaqueta cualquier número de <strong>argumentos de palabra clave</strong> adicionales en un <strong>diccionario</strong>.</p>
                <pre>def crear_perfil_usuario(nombre, **datos_extra):
    # 'datos_extra' es un diccionario con todos los argumentos de palabra clave extra.
    print(f"Perfil para: {{nombre}}")
    print("Datos adicionales:")
    for clave, valor in datos_extra.items():
        print(f"  - {{clave.capitalize()}}: {{valor}}")

crear_perfil_usuario('Ana', edad=30, ciudad='Madrid', profesion='Ingeniera')</pre>
                <div class="code-output">{output_s13_2}</div>

                <h3>La Firma Completa: Orden Canónico</h3>
                <p>Cuando usas todo junto, Python exige un orden estricto: <strong>1. Argumentos posicionales, 2. `*args`, 3. `**kwargs`</strong>. Esto asegura que no haya ambigüedad.</p>
                <pre>def procesar_pedido(id_cliente, *productos, **opciones):
    print(f"--- Procesando Pedido para Cliente: {{id_cliente}} ---")
    print(f"Productos en el carrito: {{productos}}")
    print("Opciones de envío:")
    # Se puede acceder y modificar el diccionario de opciones
    opciones.setdefault('metodo', 'estándar') # Establece un valor por defecto
    opciones.setdefault('urgente', False)
    for k, v in opciones.items():
        print(f"  - {{k.capitalize()}}: {{v}}")

procesar_pedido('CLI-123', 'Laptop', 'Mouse', 'Teclado', metodo='exprés', regalo=True)</pre>
                <div class="code-output">{output_s13_3}</div>

                <h3>El Superpoder Oculto: Desempaquetar al Llamar</h3>
                <p>La misma sintaxis se puede usar a la inversa: para <strong>desempaquetar</strong> una lista o un diccionario y pasarlos como argumentos a una función.</p>
                <div class="tip-box">Esta técnica es extremadamente útil para llamar a funciones con argumentos que has generado dinámicamente o que has recibido de una API.</div>
                <pre># Desempaquetar una lista con *
lista_nums = [15, 25, 60]
print("Llamando a sumar_todo(*lista_nums)...")
sumar_todo(*lista_nums) # Equivale a sumar_todo(15, 25, 60)

print("\\n" + "="*20 + "\\n")

# Desempaquetar un diccionario con **
dict_perfil = {{'edad': 42, 'ciudad': 'Barcelona'}}
print("Llamando a crear_perfil_usuario('Carlos', **dict_perfil)...")
crear_perfil_usuario('Carlos', **dict_perfil) # Equivale a crear_perfil_usuario('Carlos', edad=42, ciudad='Barcelona')</pre>
                <div class="code-output">Llamando a sumar_todo(*lista_nums)...
{output_s13_4a}

====================

Llamando a crear_perfil_usuario('Carlos', **dict_perfil)...
{output_s13_4b}</div>
            </div>
        </div>
    </div>
  </div>
  <script>
    (function() {{
        const themeToggleButton = document.getElementById('themeToggleButton');
        const themeIcon = document.getElementById('themeIcon');
        const body = document.body;
        const applyTheme = (theme) => {{
            body.setAttribute('data-theme', theme);
            localStorage.setItem('theme', theme);
            themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon';
        }};
        const savedTheme = localStorage.getItem('theme') || 'light';
        applyTheme(savedTheme);
        themeToggleButton.addEventListener('click', () => {{
            applyTheme(body.getAttribute('data-theme') === 'dark' ? 'light' : 'dark');
        }});
        document.querySelectorAll('.topic-header').forEach(header => {{
            header.addEventListener('click', () => header.parentElement.classList.toggle('open'));
        }});
        const firstTopic = document.querySelector('.topic-card');
        if(firstTopic) firstTopic.classList.add('open');
    }})();
  </script>
</body>
</html>
"""

# --- 3. MOSTRAR EL RESULTADO FINAL ---
display(HTML(html_python_guide_template))
```
