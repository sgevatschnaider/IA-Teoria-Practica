```python
# Celda 10
print("--- 10. ¿Aprobaste el Curso? ---")
nota_final = 75
asistencia_porcentaje = 90

# Requisitos para aprobar: Nota >= 60 Y Asistencia >= 80%
if nota_final >= 60:
    # Si la nota es suficiente, ahora comprobamos la asistencia
    if asistencia_porcentaje >= 80:
        resultado = "¡Felicidades! Has aprobado el curso."
    else:
        resultado = f"No aprobaste. Tu nota es buena ({nota_final}), pero tu asistencia ({asistencia_porcentaje}%) es muy baja."
else:
    resultado = f"No aprobaste. Tu nota ({nota_final}) no es suficiente."

print(resultado)
```
