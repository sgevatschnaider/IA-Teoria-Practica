```python
# -*- coding: utf-8 -*-
from IPython.core.display import display, HTML
import re
import traceback

# 1. Función para resaltar palabras clave
def highlight_keywords(text, keywords):
    parts = []
    keyword_pattern = '|'.join(r'(?<![\w])(?:' + re.escape(kw) + r')(?![\w])' for kw in keywords)
    pattern = f'({keyword_pattern})'
    try:
        for match in re.finditer(pattern, text, flags=re.IGNORECASE):
            start, end = match.span()
            last_open_bracket = text.rfind('<', 0, start)
            last_close_bracket = text.rfind('>', 0, start)
            if last_open_bracket > last_close_bracket:
                continue
            last_double_quote = text.rfind('"', 0, start)
            last_single_quote = text.rfind("'", 0, start)
            latest_quote = max(last_double_quote, last_single_quote)
            if latest_quote != -1 and latest_quote > last_close_bracket:
                quote_char = '"' if last_double_quote > last_single_quote else "'"
                segment_before = text[last_close_bracket + 1 : start]
                if segment_before.count(quote_char) % 2 != 0:
                    continue
            parts.append({'start': start, 'end': end, 'text': match.group(1)})
    except re.error as e:
        print(f"Advertencia: Error en el patrón regex combinado: {e}")
        return text
    except Exception as e:
        print(f"Error inesperado en highlight_keywords: {e}")
        traceback.print_exc()
        return text
    parts.sort(key=lambda x: x['start'])
    filtered_parts = []
    last_part_end = -1
    for part in parts:
        if part['start'] >= last_part_end:
            filtered_parts.append(part)
            last_part_end = part['end']
    highlighted_text = ""
    current_pos = 0
    for part in filtered_parts:
        highlighted_text += text[current_pos:part['start']]
        highlighted_text += f'<strong class="keyword">{part["text"]}</strong>'
        current_pos = part['end']
    highlighted_text += text[current_pos:]
    return highlighted_text

# 2. Palabras clave para resaltar
keywords_turing = [
    "Test de Turing", "Turing", "Alan Turing", "Computing Machinery and Intelligence",
    "máquina", "máquinas", "interrogador", "juez", "humano", "comportamiento inteligente",
    "conversación", "texto", "canal de texto", "terminal", "chat", "lenguaje corporal",
    "objetivo", "distinguir", "simular", "simulación", "inteligencia artificial",
    "comprensión contextual", "respuestas lógicas", "coherentes", "idioma", "críticas",
    "conciencia", "emociones", "inteligencia general",
    "trucos lingüísticos", "preprogramados", "importancia", "modelo de lenguaje",
    "GPT", "Claude", "Gemini", "límites", "inteligencia humana", "debate", "filosófico"
]

# 3. Estructura del texto con subtítulos
sections_turing = {
    "title": "El Test de Turing: ¿Pueden las máquinas pensar?",
    "subtitle": "Historia, funcionamiento y debates en torno a la inteligencia artificial",
    "intro": """
El Test de Turing es un experimento mental propuesto en 1950 por el matemático y pionero de la computación Alan Turing, en su artículo <b>Computing Machinery and Intelligence</b>, donde planteó la famosa pregunta: "¿Pueden las máquinas pensar?". Para responder a esa cuestión, propuso una prueba que se conoce hoy como el Test de Turing, cuyo objetivo es evaluar si una máquina puede mostrar un comportamiento inteligente indistinguible del de un ser humano.
    """,
    "en_que_consiste": """
<h2>¿En qué consiste el Test de Turing?</h2>
El test simula una conversación escrita entre tres participantes:
<ul>
<li>Un interrogador humano (también llamado juez).</li>
<li>Un humano que responde preguntas.</li>
<li>Una máquina que también responde preguntas.</li>
</ul>
El interrogador no sabe cuál de los dos es la máquina. La interacción ocurre a través de un canal de texto (por ejemplo, una terminal o chat), de manera que la voz, la apariencia y el lenguaje corporal no influyan en el juicio.
<br><br>
<b>Objetivo:</b> Si el interrogador no puede distinguir consistentemente quién es el humano y quién es la máquina —es decir, si la máquina logra "engañarlo" en una proporción significativa de las veces—, entonces se dice que la máquina ha pasado el Test de Turing.
    """,
    "que_evalua": """
<h2>¿Qué evalúa realmente el test?</h2>
Más que medir inteligencia real en términos humanos, el test evalúa la capacidad de una máquina para simular una conversación humana de forma creíble, utilizando:
<ul>
<li>Comprensión contextual del lenguaje</li>
<li>Respuestas lógicas y coherentes</li>
<li>Uso apropiado del idioma</li>
</ul>
    """,
    "criticas": """
<h2>Críticas y debates sobre el Test de Turing</h2>
Aunque influyente, el test ha recibido críticas por varias razones:
<ul>
<li><b>Simulación no es comprensión:</b> Una máquina puede imitar la conversación humana sin tener conciencia, emociones o comprensión real.</li>
<li><b>Limitación del lenguaje:</b> El test se basa solo en texto; no considera otras formas de inteligencia como la visual, emocional o corporal.</li>
<li><b>Falsos positivos:</b> Una máquina podría engañar al juez sin tener inteligencia general, solo usando trucos lingüísticos o datos preprogramados.</li>
</ul>
    """,
    "importancia": """
<h2>Importancia histórica y actual</h2>
El Test de Turing no se utiliza como evaluación formal en la investigación moderna de IA, pero sigue siendo un símbolo filosófico y una herramienta para debatir sobre los límites entre la inteligencia humana y artificial.
<br><br>
Con el avance de los modelos de lenguaje como GPT, Claude o Gemini, se ha renovado el interés por esta prueba, ya que algunas conversaciones con estas IA pueden parecer indistinguibles de las humanas en muchos contextos.
    """
}

# 4. Resaltar palabras clave en cada sección
highlighted_sections = {k: highlight_keywords(v, keywords_turing) for k, v in sections_turing.items()}

# 5. CSS y JS para cambio de tema
css_styles = """
<style>
:root {
  --primary-color: #2e86de;
  --secondary-color: #00b894;
  --text-color: #2d3436;
  --bg-color: #f9fafb;
  --container-bg: #ffffff;
  --keyword-bg: #dff9fb;
  --keyword-text: #0652dd;
  --button-bg: #f1f2f6;
  --button-hover-bg: #dfe4ea;
  --button-text-color: #222f3e;
  --shadow-color: rgba(0,0,0,0.07);
}
body.dark-mode {
  --primary-color: #0984e3;
  --secondary-color: #00cec9;
  --text-color: #f5f6fa;
  --bg-color: #222f3e;
  --container-bg: #1e272e;
  --keyword-bg: #576574;
  --keyword-text: #00d2d3;
  --button-bg: #222f3e;
  --button-hover-bg: #576574;
  --button-text-color: #dff9fb;
  --shadow-color: rgba(0,0,0,0.4);
}
body {
  font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
  line-height: 1.7;
  background-color: var(--bg-color);
  color: var(--text-color);
  transition: background-color .3s, color .3s;
  padding: 18px;
  margin: 0;
}
.container {
  max-width: 880px;
  margin: 28px auto;
  padding: 32px;
  background-color: var(--container-bg);
  border-radius: 9px;
  box-shadow: 0 4px 12px var(--shadow-color);
  position: relative;
  transition: background-color .3s, box-shadow .3s;
  border-top: 6px solid var(--primary-color);
}
h1 {
  color: var(--primary-color);
  font-size: 2.3em;
  margin-bottom: .2em;
  line-height: 1.2;
  text-align: center;
}
h2, h3 {
  color: var(--secondary-color);
  text-align: left;
  margin-top: 2.1em;
  margin-bottom: .7em;
}
.keyword {
  background-color: var(--keyword-bg);
  color: var(--keyword-text);
  padding: 0.14em 0.36em;
  border-radius: 4px;
  font-weight: 600;
  display: inline-block;
  margin: 0 1px;
  transition: background-color .2s, color .2s;
}
.theme-toggle {
  background-color: var(--button-bg);
  color: var(--button-text-color);
  border: 1.5px solid var(--secondary-color);
  padding: 9px 14px;
  border-radius: 5px;
  cursor: pointer;
  position: absolute;
  top: 18px;
  right: 18px;
  font-size: 1em;
  transition: background-color .2s, color .2s, border-color .2s;
  z-index: 10;
}
.theme-toggle:hover { background-color: var(--button-hover-bg);}
.theme-toggle:focus {
    outline: 2px solid var(--primary-color);
    outline-offset: 2px;
    border-color: var(--primary-color);
}
ul { margin-top:0.7em; margin-bottom:1.2em; }
li { margin-bottom:0.5em; }
b { color: var(--primary-color);}
</style>
"""

javascript_code = """
<script>
function toggleTheme() {
    const body = document.body;
    const themeToggleButton = document.getElementById("theme-toggle-btn");
    const isDarkMode = body.classList.toggle("dark-mode");
    localStorage.setItem("theme", isDarkMode ? "dark" : "light");
    if (themeToggleButton) {
        themeToggleButton.textContent = isDarkMode ? "Modo Claro" : "Modo Oscuro";
        themeToggleButton.setAttribute("aria-pressed", isDarkMode ? "true" : "false");
    }
}
window.onload = function() {
    const themeToggleButton = document.getElementById("theme-toggle-btn");
    const savedTheme = localStorage.getItem("theme");
    if (savedTheme === "dark") {
        document.body.classList.add("dark-mode");
        if (themeToggleButton) {
             themeToggleButton.textContent = "Modo Claro";
             themeToggleButton.setAttribute("aria-pressed", "true");
        }
    } else {
        document.body.classList.remove("dark-mode");
        if (themeToggleButton) {
             themeToggleButton.textContent = "Modo Oscuro";
             themeToggleButton.setAttribute("aria-pressed", "false");
        }
    }
    if (themeToggleButton && !themeToggleButton.hasAttribute('aria-pressed')) {
         themeToggleButton.setAttribute("aria-pressed", document.body.classList.contains("dark-mode") ? "true" : "false");
    }
};
</script>
"""

# 6. Ensamblar el HTML
html_structure = f"""
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <title>{highlighted_sections['title']}</title>
    {css_styles}
</head>
<body>
<div class="container">
    <button id="theme-toggle-btn" class="theme-toggle" onclick="toggleTheme()" title="Cambiar tema de color">Modo Oscuro</button>
    <h1>{highlighted_sections['title']}</h1>
    <h2 style="font-style:italic; font-weight:400; color:#636e72; text-align:center; margin-top:-0.5em;">{highlighted_sections['subtitle']}</h2>
    <div style="margin-top: 1.3em;">{highlighted_sections['intro']}</div>
    {highlighted_sections['en_que_consiste']}
    {highlighted_sections['que_evalua']}
    {highlighted_sections['criticas']}
    {highlighted_sections['importancia']}
</div>
{javascript_code}
</body>
</html>
"""

# 7. Mostrar el HTML en Colab
display(HTML(html_structure))
```

```python
# ----------------------------------------
# Análisis de Sentimientos con Scikit-learn
# ----------------------------------------

# Paso 1: Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from IPython.display import display  # Para mostrar tablas en Google Colab/Notebooks

import warnings
warnings.filterwarnings('ignore')

# Paso 2: Crear dataset manual de ejemplo
data = {
    'texto': [
        "Me encantó la película, fue emocionante",
        "Odié el servicio, muy malo",
        "Increíble experiencia, volvería sin dudas",
        "No lo recomiendo, fue una pérdida de tiempo",
        "El producto llegó roto, muy decepcionado",
        "Excelente atención y calidad",
        "La comida estaba fría y sin sabor",
        "Fue el mejor día de mi vida",
        "Nada que ver con lo prometido",
        "Superó todas mis expectativas"
    ],
    'sentimiento': [
        'positivo',
        'negativo',
        'positivo',
        'negativo',
        'negativo',
        'positivo',
        'negativo',
        'positivo',
        'negativo',
        'positivo'
    ]
}
df = pd.DataFrame(data)
print("Dataset de ejemplo:")
display(df)

# Paso 3: Vectorizar el texto (convertir texto a números)
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['texto'])
y = df['sentimiento']

# Paso 4: Dividir el dataset en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Paso 5: Entrenar el modelo Naive Bayes
modelo = MultinomialNB()
modelo.fit(X_train, y_train)

# Paso 6: Evaluar el modelo
y_pred = modelo.predict(X_test)
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred, zero_division=0))

# Paso 7: Visualizar predicciones sobre los textos de prueba
test_texts = df.iloc[y_test.index]['texto'].reset_index(drop=True)
predictions_df = pd.DataFrame({
    'Texto': test_texts,
    'Sentimiento real': y_test.reset_index(drop=True),
    'Predicción': y_pred
})
predictions_df['Coincide con la realidad'] = predictions_df.apply(
    lambda row: 'Sí' if row['Sentimiento real'] == row['Predicción'] else 'No', axis=1
)

print("\nResultados de las predicciones en el set de prueba:")
display(predictions_df)

# Paso 8: Gráfico de distribución de predicciones
plt.figure(figsize=(6,4))
pred_counts = predictions_df['Predicción'].value_counts()
bars = plt.bar(pred_counts.index, pred_counts.values, color=['skyblue', 'salmon'])
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
             str(int(bar.get_height())), ha='center', va='bottom', fontsize=12)
plt.title("Distribución de Sentimientos Predichos")
plt.xlabel("Sentimiento")
plt.ylabel("Cantidad")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Paso 9: Prueba con texto personalizado (input del usuario)
texto_usuario = ["El servicio fue impecable y rápido"]
texto_vec = vectorizer.transform(texto_usuario)
prediccion = modelo.predict(texto_vec)

print("\nTexto personalizado de prueba:")
print(f"Texto ingresado: \"{texto_usuario[0]}\"")
print(f"Sentimiento predicho: {prediccion[0]}")

# (Opcional) Puedes agregar un input interactivo en Colab así:
texto_usuario = [input("Escribe una opinión para analizar su sentimiento: ")]
texto_vec = vectorizer.transform(texto_usuario)
print(f"Sentimiento predicho: {modelo.predict(texto_vec)[0]}")
```

```python
from IPython.core.display import display, HTML
import re
import traceback

# 1. Función para resaltar palabras clave
def highlight_keywords(text, keywords):
    parts = []
    keyword_pattern = '|'.join(r'(?<![\w])(?:' + re.escape(kw) + r')(?![\w])' for kw in keywords)
    pattern = f'({keyword_pattern})'
    try:
        for match in re.finditer(pattern, text, flags=re.IGNORECASE):
            start, end = match.span()
            last_open_bracket = text.rfind('<', 0, start)
            last_close_bracket = text.rfind('>', 0, start)
            if last_open_bracket > last_close_bracket:
                continue
            last_double_quote = text.rfind('"', 0, start)
            last_single_quote = text.rfind("'", 0, start)
            latest_quote = max(last_double_quote, last_single_quote)
            if latest_quote != -1 and latest_quote > last_close_bracket:
                quote_char = '"' if last_double_quote > last_single_quote else "'"
                segment_before = text[last_close_bracket + 1 : start]
                if segment_before.count(quote_char) % 2 != 0:
                    continue
            parts.append({'start': start, 'end': end, 'text': match.group(1)})
    except re.error as e:
        print(f"Advertencia: Error en el patrón regex combinado: {e}")
        return text
    except Exception as e:
        print(f"Error inesperado en highlight_keywords: {e}")
        traceback.print_exc()
        return text
    parts.sort(key=lambda x: x['start'])
    filtered_parts = []
    last_part_end = -1
    for part in parts:
        if part['start'] >= last_part_end:
            filtered_parts.append(part)
            last_part_end = part['end']
    highlighted_text = ""
    current_pos = 0
    for part in filtered_parts:
        highlighted_text += text[current_pos:part['start']]
        highlighted_text += f'<strong class="keyword">{part["text"]}</strong>'
        current_pos = part['end']
    highlighted_text += text[current_pos:]
    return highlighted_text

# 2. Palabras clave para resaltar
keywords_token = [
    "token", "tokens", "tokenización", "tokenizar", "NLP", "modelo de lenguaje",
    "GPT", "BERT", "T5", "transformer", "representaciones numéricas", "embedding",
    "Byte Pair Encoding", "BPE", "WordPiece", "SentencePiece", "parámetros", "límite de tokens",
    "API", "OpenAI", "entrada", "salida", "contexto", "tokenizador", "preentrenamiento",
    "fine-tune", "red neuronal", "vector", "corpus", "caracteres", "palabra", "subpalabra",
    "símbolo", "espacio", "tiktoken"
]

# 3. Secciones didácticas
sections_token = {
    "title": "¿Qué es un token en procesamiento de lenguaje natural (NLP)?",
    "subtitle": "Definición, tipos, ejemplos y relevancia en modelos como GPT y BERT",
    "intro": """
En <b>procesamiento de lenguaje natural (NLP)</b>, un <b>token</b> es la unidad mínima de texto que un modelo de lenguaje puede procesar. Es el bloque básico que alimenta al modelo durante el análisis, entrenamiento o generación de texto.
<br><br>
Modelos como GPT, BERT o T5 no procesan texto directamente en palabras ni en caracteres, sino que lo <b>tokenizan</b>, es decir, lo convierten en una secuencia de <b>tokens</b> antes de transformarlo en <b>representaciones numéricas</b> (vectores).
""",
    "que_puede_ser": """
<h2>¿Qué puede ser un token?</h2>
Un token puede ser:
<ul>
<li>Una <b>palabra completa</b>: por ejemplo, "feliz" puede ser un único token.</li>
<li>Una <b>parte de una palabra</b> (subpalabra): por ejemplo, "inteligencia" puede dividirse en "intel", "igencia".</li>
<li>Una sola letra o <b>símbolo especial</b>: como "@", "$" o "."</li>
<li><b>Espacios en blanco</b> (en algunos tokenizadores): por ejemplo, " Chat" puede contar como un token distinto de "Chat".</li>
</ul>
Esto depende del tipo de <b>tokenizador</b> utilizado.
""",
    "tipos": """
<h2>Tipos de tokenización</h2>
Existen distintos enfoques para convertir texto en tokens:
<ul>
<li><b>Tokenización basada en palabras</b>: Divide el texto por espacios.<br>
<code>"GPT genera texto" → ["GPT", "genera", "texto"]</code>
<br>Es simple, pero falla con palabras nuevas, errores ortográficos o lenguajes altamente flexionados.</li>
<li><b>Tokenización basada en caracteres</b>: Cada letra es un token.<br>
<code>"Hola" → ["H", "o", "l", "a"]</code>
<br>Muy granular, pero poco eficiente para modelos grandes.</li>
<li><b>Tokenización basada en subpalabras</b> (BPE, WordPiece, SentencePiece):<br>
Este es el tipo usado por modelos como GPT, BERT y T5. Divide las palabras en unidades más pequeñas según su frecuencia en el corpus de entrenamiento. <br>
<code>"Desarrollador" → ["Des", "arroll", "ador"]</code>
<br>Esto permite manejar palabras desconocidas o nuevas (como "transformerización") dividiéndolas en subcomponentes conocidos por el modelo.</li>
</ul>
""",
    "ejemplo": """
<h2>Ejemplo real: tokenización de GPT</h2>
GPT usa un tokenizador basado en <b>Byte Pair Encoding (BPE)</b>. Veamos la frase:
<pre><code>"ChatGPT es genial"</code></pre>
Puede tokenizarse como:
<pre><code>["Chat", "G", "PT", " es", " genial"]</code></pre>
Observaciones:
<ul>
<li>"ChatGPT" se divide en partes conocidas.</li>
<li>Los espacios previos pueden ser parte del token (" es", " genial").</li>
<li>La tokenización es sensible a mayúsculas, espacios y símbolos.</li>
</ul>
""",
    "importancia": """
<h2>¿Por qué importa cuántos tokens hay?</h2>
<ul>
<li>Los modelos trabajan en función de <b>tokens</b>, no de palabras.</li>
<li>GPT-3.5 tiene un límite de <b>4,096 tokens</b> por entrada/salida combinada, GPT-4 puede llegar hasta 32,000 tokens.</li>
<li>El costo de uso en APIs como OpenAI se mide por token.</li>
<li>A mayor cantidad de tokens, mayor carga computacional y memoria necesaria.</li>
</ul>
""",
    "embedding": """
<h2>¿Cómo se usa un token?</h2>
Una vez tokenizado el texto, cada token se convierte en un <b>embedding vectorial</b>, es decir, un vector numérico de alta dimensión que representa su significado y contexto. Esos vectores son los que realmente procesa el modelo usando la arquitectura Transformer.
""",
    "tabla": """
<h2>En resumen</h2>
<table style="border-collapse:collapse;width:100%;margin-top:1em;">
<tr style="background:#f1f2f6;">
  <th style="padding:6px;border:1px solid #aaa;">Concepto</th>
  <th style="padding:6px;border:1px solid #aaa;">Explicación clara</th>
</tr>
<tr>
  <td style="padding:6px;border:1px solid #aaa;">Token</td>
  <td style="padding:6px;border:1px solid #aaa;">Unidad mínima de texto procesada por un modelo de lenguaje</td>
</tr>
<tr>
  <td style="padding:6px;border:1px solid #aaa;">No equivale a palabra</td>
  <td style="padding:6px;border:1px solid #aaa;">Puede ser una palabra, parte de una palabra, un símbolo, espacio o puntuación</td>
</tr>
<tr>
  <td style="padding:6px;border:1px solid #aaa;">Importancia</td>
  <td style="padding:6px;border:1px solid #aaa;">Es la base del procesamiento en GPT, BERT, etc.</td>
</tr>
<tr>
  <td style="padding:6px;border:1px solid #aaa;">Tokenizador</td>
  <td style="padding:6px;border:1px solid #aaa;">Algoritmo que convierte texto en tokens, como BPE o WordPiece</td>
</tr>
<tr>
  <td style="padding:6px;border:1px solid #aaa;">Relación con el modelo</td>
  <td style="padding:6px;border:1px solid #aaa;">Cada token se transforma en un vector numérico para alimentar a la red neuronal</td>
</tr>
</table>
""",
    "demo": """
<h2>¿Cómo ver el número de tokens en un texto?</h2>
Con <b>tiktoken</b> (el tokenizador de OpenAI), podés calcularlo así en Colab:
<pre><code>!pip install -q tiktoken

import tiktoken

tokenizer = tiktoken.get_encoding("cl100k_base")  # Compatible con GPT-4
texto = "ChatGPT es genial"
tokens = tokenizer.encode(texto)

print("Tokens:", tokens)
print("Cantidad de tokens:", len(tokens))
</code></pre>
"""
}

# 4. Resaltado de palabras clave
highlighted_sections = {k: highlight_keywords(v, keywords_token) for k, v in sections_token.items()}

# 5. CSS + JS para modo claro/oscuro
css_styles = """
<style>
:root {
  --primary-color: #2e86de;
  --secondary-color: #00b894;
  --text-color: #2d3436;
  --bg-color: #f9fafb;
  --container-bg: #ffffff;
  --keyword-bg: #dff9fb;
  --keyword-text: #0652dd;
  --button-bg: #f1f2f6;
  --button-hover-bg: #dfe4ea;
  --button-text-color: #222f3e;
  --shadow-color: rgba(0,0,0,0.07);
}
body.dark-mode {
  --primary-color: #0984e3;
  --secondary-color: #00cec9;
  --text-color: #f5f6fa;
  --bg-color: #222f3e;
  --container-bg: #1e272e;
  --keyword-bg: #576574;
  --keyword-text: #00d2d3;
  --button-bg: #222f3e;
  --button-hover-bg: #576574;
  --button-text-color: #dff9fb;
  --shadow-color: rgba(0,0,0,0.4);
}
body {
  font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
  line-height: 1.7;
  background-color: var(--bg-color);
  color: var(--text-color);
  transition: background-color .3s, color .3s;
  padding: 18px;
  margin: 0;
}
.container {
  max-width: 880px;
  margin: 28px auto;
  padding: 32px;
  background-color: var(--container-bg);
  border-radius: 9px;
  box-shadow: 0 4px 12px var(--shadow-color);
  position: relative;
  transition: background-color .3s, box-shadow .3s;
  border-top: 6px solid var(--primary-color);
}
h1 {
  color: var(--primary-color);
  font-size: 2.1em;
  margin-bottom: .2em;
  line-height: 1.2;
  text-align: center;
}
h2, h3 {
  color: var(--secondary-color);
  text-align: left;
  margin-top: 2.1em;
  margin-bottom: .7em;
}
.keyword {
  background-color: var(--keyword-bg);
  color: var(--keyword-text);
  padding: 0.14em 0.36em;
  border-radius: 4px;
  font-weight: 600;
  display: inline-block;
  margin: 0 1px;
  transition: background-color .2s, color .2s;
}
.theme-toggle {
  background-color: var(--button-bg);
  color: var(--button-text-color);
  border: 1.5px solid var(--secondary-color);
  padding: 9px 14px;
  border-radius: 5px;
  cursor: pointer;
  position: absolute;
  top: 18px;
  right: 18px;
  font-size: 1em;
  transition: background-color .2s, color .2s, border-color .2s;
  z-index: 10;
}
.theme-toggle:hover { background-color: var(--button-hover-bg);}
.theme-toggle:focus {
    outline: 2px solid var(--primary-color);
    outline-offset: 2px;
    border-color: var(--primary-color);
}
ul { margin-top:0.7em; margin-bottom:1.2em; }
li { margin-bottom:0.5em; }
b { color: var(--primary-color);}
table {font-size:1em;}
pre {background: #f3f3f3; color: #262626; border-radius: 7px; padding: 10px; font-size: 1em;}
body.dark-mode pre {background: #263445; color: #f5f6fa;}
</style>
"""

javascript_code = """
<script>
function toggleTheme() {
    const body = document.body;
    const themeToggleButton = document.getElementById("theme-toggle-btn");
    const isDarkMode = body.classList.toggle("dark-mode");
    localStorage.setItem("theme", isDarkMode ? "dark" : "light");
    if (themeToggleButton) {
        themeToggleButton.textContent = isDarkMode ? "Modo Claro" : "Modo Oscuro";
        themeToggleButton.setAttribute("aria-pressed", isDarkMode ? "true" : "false");
    }
}
window.onload = function() {
    const themeToggleButton = document.getElementById("theme-toggle-btn");
    const savedTheme = localStorage.getItem("theme");
    if (savedTheme === "dark") {
        document.body.classList.add("dark-mode");
        if (themeToggleButton) {
             themeToggleButton.textContent = "Modo Claro";
             themeToggleButton.setAttribute("aria-pressed", "true");
        }
    } else {
        document.body.classList.remove("dark-mode");
        if (themeToggleButton) {
             themeToggleButton.textContent = "Modo Oscuro";
             themeToggleButton.setAttribute("aria-pressed", "false");
        }
    }
    if (themeToggleButton && !themeToggleButton.hasAttribute('aria-pressed')) {
         themeToggleButton.setAttribute("aria-pressed", document.body.classList.contains("dark-mode") ? "true" : "false");
    }
};
</script>
"""

# 6. Ensamblar el HTML
html_structure = f"""
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <title>{highlighted_sections['title']}</title>
    {css_styles}
</head>
<body>
<div class="container">
    <button id="theme-toggle-btn" class="theme-toggle" onclick="toggleTheme()" title="Cambiar tema de color">Modo Oscuro</button>
    <h1>{highlighted_sections['title']}</h1>
    <h2 style="font-style:italic; font-weight:400; color:#636e72; text-align:center; margin-top:-0.5em;">{highlighted_sections['subtitle']}</h2>
    <div style="margin-top: 1.3em;">{highlighted_sections['intro']}</div>
    {highlighted_sections['que_puede_ser']}
    {highlighted_sections['tipos']}
    {highlighted_sections['ejemplo']}
    {highlighted_sections['importancia']}
    {highlighted_sections['embedding']}
    {highlighted_sections['tabla']}
    {highlighted_sections['demo']}
</div>
{javascript_code}
</body>
</html>
"""

display(HTML(html_structure))
```

```python
from IPython.core.display import display, HTML
import re
import traceback

# 1. Función para resaltar palabras clave
def highlight_keywords(text, keywords):
    parts = []
    keyword_pattern = '|'.join(r'(?<![\w])(?:' + re.escape(kw) + r')(?![\w])' for kw in keywords)
    pattern = f'({keyword_pattern})'
    try:
        for match in re.finditer(pattern, text, flags=re.IGNORECASE):
            start, end = match.span()
            last_open_bracket = text.rfind('<', 0, start)
            last_close_bracket = text.rfind('>', 0, start)
            if last_open_bracket > last_close_bracket:
                continue
            last_double_quote = text.rfind('"', 0, start)
            last_single_quote = text.rfind("'", 0, start)
            latest_quote = max(last_double_quote, last_single_quote)
            if latest_quote != -1 and latest_quote > last_close_bracket:
                quote_char = '"' if last_double_quote > last_single_quote else "'"
                segment_before = text[last_close_bracket + 1 : start]
                if segment_before.count(quote_char) % 2 != 0:
                    continue
            parts.append({'start': start, 'end': end, 'text': match.group(1)})
    except re.error as e:
        print(f"Advertencia: Error en el patrón regex combinado: {e}")
        return text
    except Exception as e:
        print(f"Error inesperado en highlight_keywords: {e}")
        traceback.print_exc()
        return text
    parts.sort(key=lambda x: x['start'])
    filtered_parts = []
    last_part_end = -1
    for part in parts:
        if part['start'] >= last_part_end:
            filtered_parts.append(part)
            last_part_end = part['end']
    highlighted_text = ""
    current_pos = 0
    for part in filtered_parts:
        highlighted_text += text[current_pos:part['start']]
        highlighted_text += f'<strong class="keyword">{part["text"]}</strong>'
        current_pos = part['end']
    highlighted_text += text[current_pos:]
    return highlighted_text

# 2. Palabras clave a resaltar
keywords_top_p = [
    "Top-p", "nucleus sampling", "probabilidad acumulada", "umbral", "p", "modelo de lenguaje",
    "GPT", "token", "aleatoriedad", "núcleo", "Top-k", "argmax", "temperatura", "exploración", "explotación"
]

# 3. Secciones didácticas
sections_top_p = {
    "title": "¿Qué es Top-p sampling (nucleus sampling)?",
    "subtitle": "Selección probabilística adaptativa en modelos de lenguaje",
    "intro": """
<b>Top-p</b>, también conocido como <b>nucleus sampling</b>, es un método probabilístico para seleccionar la siguiente palabra (token) durante la generación de texto en modelos de lenguaje como GPT.<br>
Se utiliza como alternativa o complemento al parámetro <b>temperatura</b>, y su propósito es el mismo: controlar la <b>aleatoriedad</b> y la calidad del texto generado, pero con un enfoque más dinámico y ajustado a cada contexto.
""",
    "como_funciona": """
<h2>¿Cómo funciona?</h2>
En cada paso de generación de texto, el modelo predice una <b>distribución de probabilidad</b> sobre el vocabulario completo. Por ejemplo:<br>
<code>“Hoy me siento muy…”</code>
<br><br>
El modelo evalúa cientos de posibles palabras siguientes y les asigna una probabilidad según el contexto.<br><br>
Con <b>Top-p</b>, el modelo:
<ul>
<li>Ordena las palabras por probabilidad, de mayor a menor.</li>
<li>Suma las probabilidades acumuladas hasta superar un <b>umbral p</b> (por ejemplo, 0.9).</li>
<li>Elimina las palabras cuya probabilidad acumulada queda fuera de ese núcleo.</li>
<li>Selecciona aleatoriamente una palabra solo entre ese <b>conjunto reducido</b> (núcleo).</li>
</ul>
""",
    "que_es_p": """
<h2>¿Qué significa "p"?</h2>
<ul>
<li>p = 0.9 → se seleccionan las palabras más probables cuyo total sume el 90% de la probabilidad total.</li>
<li>Cuanto menor sea "p", más conservador es el modelo.</li>
<li>Cuanto mayor sea "p", más diversidad se permite.</li>
</ul>
""",
    "ejemplo": """
<h2>Ejemplo ilustrativo</h2>
Supongamos que el modelo genera esta distribución:
<table style="border-collapse:collapse;width:90%;margin-top:1em;">
<tr style="background:#f1f2f6;">
  <th style="padding:6px;border:1px solid #aaa;">Palabra posible</th>
  <th style="padding:6px;border:1px solid #aaa;">Probabilidad individual</th>
  <th style="padding:6px;border:1px solid #aaa;">Probabilidad acumulada</th>
</tr>
<tr><td style="padding:6px;border:1px solid #aaa;">feliz</td><td style="padding:6px;border:1px solid #aaa;">0.45</td><td style="padding:6px;border:1px solid #aaa;">0.45</td></tr>
<tr><td style="padding:6px;border:1px solid #aaa;">cansado</td><td style="padding:6px;border:1px solid #aaa;">0.25</td><td style="padding:6px;border:1px solid #aaa;">0.70</td></tr>
<tr><td style="padding:6px;border:1px solid #aaa;">inspirado</td><td style="padding:6px;border:1px solid #aaa;">0.12</td><td style="padding:6px;border:1px solid #aaa;">0.82</td></tr>
<tr><td style="padding:6px;border:1px solid #aaa;">vacío</td><td style="padding:6px;border:1px solid #aaa;">0.07</td><td style="padding:6px;border:1px solid #aaa;">0.89</td></tr>
<tr><td style="padding:6px;border:1px solid #aaa;">aburrido</td><td style="padding:6px;border:1px solid #aaa;">0.05</td><td style="padding:6px;border:1px solid #aaa;">0.94</td></tr>
<tr><td style="padding:6px;border:1px solid #aaa;">perdido</td><td style="padding:6px;border:1px solid #aaa;">0.03</td><td style="padding:6px;border:1px solid #aaa;">0.97</td></tr>
<tr><td style="padding:6px;border:1px solid #aaa;">eufórico</td><td style="padding:6px;border:1px solid #aaa;">0.02</td><td style="padding:6px;border:1px solid #aaa;">0.99</td></tr>
<tr><td style="padding:6px;border:1px solid #aaa;">desencajado</td><td style="padding:6px;border:1px solid #aaa;">0.01</td><td style="padding:6px;border:1px solid #aaa;">1.00</td></tr>
</table>
Si <b>p = 0.9</b>: el modelo selecciona solo entre: <b>feliz, cansado, inspirado, vacío</b>.
""",
    "por_que": """
<h2>¿Por qué usar Top-p?</h2>
<ul>
<li>Evita resultados demasiado deterministas (como en argmax o temperatura baja).</li>
<li>También evita resultados absurdamente improbables, como podría pasar con temperatura alta.</li>
<li>El número de opciones se ajusta según la forma de la distribución en cada paso, no con un número fijo como en Top-k.</li>
</ul>
""",
    "comparativa": """
<h2>Comparación rápida: Top-p vs. otros métodos</h2>
<table style="border-collapse:collapse;width:100%;margin-top:1em;">
<tr style="background:#f1f2f6;">
  <th style="padding:6px;border:1px solid #aaa;">Método</th>
  <th style="padding:6px;border:1px solid #aaa;">¿Qué hace?</th>
</tr>
<tr><td style="padding:6px;border:1px solid #aaa;">Argmax</td><td style="padding:6px;border:1px solid #aaa;">Siempre elige la palabra más probable (determinista, sin creatividad)</td></tr>
<tr><td style="padding:6px;border:1px solid #aaa;">Temperatura</td><td style="padding:6px;border:1px solid #aaa;">Ajusta toda la distribución suavizándola o agudizándola</td></tr>
<tr><td style="padding:6px;border:1px solid #aaa;">Top-k</td><td style="padding:6px;border:1px solid #aaa;">Selecciona aleatoriamente entre las k palabras más probables</td></tr>
<tr><td style="padding:6px;border:1px solid #aaa;">Top-p</td><td style="padding:6px;border:1px solid #aaa;">Selecciona entre todas las palabras cuya probabilidad acumulada ≤ p</td></tr>
</table>
""",
    "resumen": """
<h2>En resumen</h2>
<ul>
<li><b>Top-p</b> controla la diversidad y la calidad del texto generado.</li>
<li>Define un “núcleo” de palabras relevantes cuya probabilidad acumulada no supere un umbral p.</li>
<li>Aporta un equilibrio inteligente entre exploración (creatividad) y explotación (coherencia).</li>
<li>Suele combinarse con temperatura para un control más fino del estilo de salida.</li>
</ul>
"""
}

# 4. Resaltado de palabras clave en cada sección
highlighted_sections = {k: highlight_keywords(v, keywords_top_p) for k, v in sections_top_p.items()}

# 5. CSS + JS para modo claro/oscuro
css_styles = """
<style>
:root {
  --primary-color: #2e86de;
  --secondary-color: #00b894;
  --text-color: #2d3436;
  --bg-color: #f9fafb;
  --container-bg: #ffffff;
  --keyword-bg: #dff9fb;
  --keyword-text: #0652dd;
  --button-bg: #f1f2f6;
  --button-hover-bg: #dfe4ea;
  --button-text-color: #222f3e;
  --shadow-color: rgba(0,0,0,0.07);
}
body.dark-mode {
  --primary-color: #0984e3;
  --secondary-color: #00cec9;
  --text-color: #f5f6fa;
  --bg-color: #222f3e;
  --container-bg: #1e272e;
  --keyword-bg: #576574;
  --keyword-text: #00d2d3;
  --button-bg: #222f3e;
  --button-hover-bg: #576574;
  --button-text-color: #dff9fb;
  --shadow-color: rgba(0,0,0,0.4);
}
body {
  font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
  line-height: 1.7;
  background-color: var(--bg-color);
  color: var(--text-color);
  transition: background-color .3s, color .3s;
  padding: 18px;
  margin: 0;
}
.container {
  max-width: 880px;
  margin: 28px auto;
  padding: 32px;
  background-color: var(--container-bg);
  border-radius: 9px;
  box-shadow: 0 4px 12px var(--shadow-color);
  position: relative;
  transition: background-color .3s, box-shadow .3s;
  border-top: 6px solid var(--primary-color);
}
h1 {
  color: var(--primary-color);
  font-size: 2.1em;
  margin-bottom: .2em;
  line-height: 1.2;
  text-align: center;
}
h2, h3 {
  color: var(--secondary-color);
  text-align: left;
  margin-top: 2.1em;
  margin-bottom: .7em;
}
.keyword {
  background-color: var(--keyword-bg);
  color: var(--keyword-text);
  padding: 0.14em 0.36em;
  border-radius: 4px;
  font-weight: 600;
  display: inline-block;
  margin: 0 1px;
  transition: background-color .2s, color .2s;
}
.theme-toggle {
  background-color: var(--button-bg);
  color: var(--button-text-color);
  border: 1.5px solid var(--secondary-color);
  padding: 9px 14px;
  border-radius: 5px;
  cursor: pointer;
  position: absolute;
  top: 18px;
  right: 18px;
  font-size: 1em;
  transition: background-color .2s, color .2s, border-color .2s;
  z-index: 10;
}
.theme-toggle:hover { background-color: var(--button-hover-bg);}
.theme-toggle:focus {
    outline: 2px solid var(--primary-color);
    outline-offset: 2px;
    border-color: var(--primary-color);
}
ul { margin-top:0.7em; margin-bottom:1.2em; }
li { margin-bottom:0.5em; }
b { color: var(--primary-color);}
table {font-size:1em;}
pre {background: #f3f3f3; color: #262626; border-radius: 7px; padding: 10px; font-size: 1em;}
body.dark-mode pre {background: #263445; color: #f5f6fa;}
</style>
"""

javascript_code = """
<script>
function toggleTheme() {
    const body = document.body;
    const themeToggleButton = document.getElementById("theme-toggle-btn");
    const isDarkMode = body.classList.toggle("dark-mode");
    localStorage.setItem("theme", isDarkMode ? "dark" : "light");
    if (themeToggleButton) {
        themeToggleButton.textContent = isDarkMode ? "Modo Claro" : "Modo Oscuro";
        themeToggleButton.setAttribute("aria-pressed", isDarkMode ? "true" : "false");
    }
}
window.onload = function() {
    const themeToggleButton = document.getElementById("theme-toggle-btn");
    const savedTheme = localStorage.getItem("theme");
    if (savedTheme === "dark") {
        document.body.classList.add("dark-mode");
        if (themeToggleButton) {
             themeToggleButton.textContent = "Modo Claro";
             themeToggleButton.setAttribute("aria-pressed", "true");
        }
    } else {
        document.body.classList.remove("dark-mode");
        if (themeToggleButton) {
             themeToggleButton.textContent = "Modo Oscuro";
             themeToggleButton.setAttribute("aria-pressed", "false");
        }
    }
    if (themeToggleButton && !themeToggleButton.hasAttribute('aria-pressed')) {
         themeToggleButton.setAttribute("aria-pressed", document.body.classList.contains("dark-mode") ? "true" : "false");
    }
};
</script>
"""

# 6. Ensamblar el HTML (sin la visualización, que va aparte en matplotlib)
html_structure = f"""
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <title>{highlighted_sections['title']}</title>
    {css_styles}
</head>
<body>
<div class="container">
    <button id="theme-toggle-btn" class="theme-toggle" onclick="toggleTheme()" title="Cambiar tema de color">Modo Oscuro</button>
    <h1>{highlighted_sections['title']}</h1>
    <h2 style="font-style:italic; font-weight:400; color:#636e72; text-align:center; margin-top:-0.5em;">{highlighted_sections['subtitle']}</h2>
    <div style="margin-top: 1.3em;">{highlighted_sections['intro']}</div>
    {highlighted_sections['como_funciona']}
    {highlighted_sections['que_es_p']}
    {highlighted_sections['ejemplo']}
    {highlighted_sections['por_que']}
    {highlighted_sections['comparativa']}
    {highlighted_sections['resumen']}
    <hr style="margin:30px 0 18px 0;">
    <h2 style="margin-bottom:0.5em;">Visualización interactiva: núcleo Top-p</h2>
    <div id="top-p-widget"></div>
    <div style='color:#7f8c8d;font-size:0.97em;'>
    Ajusta el valor de <b>p</b> en el widget abajo para ver cómo cambia el “núcleo” de palabras candidatas.<br>
    Las barras en verde indican las palabras seleccionables según p.
    </div>
</div>
{javascript_code}
</body>
</html>
"""

# 7. Mostrar la explicación HTML
display(HTML(html_structure))

# 8. Visualización interactiva de Top-p (nucleus) con matplotlib
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets

# Ejemplo fijo, reproducible
palabras = ["feliz", "cansado", "inspirado", "vacío", "aburrido", "perdido", "eufórico", "desencajado"]
prob_ind = np.array([0.45, 0.25, 0.12, 0.07, 0.05, 0.03, 0.02, 0.01])

def visualizar_top_p(p=0.9):
    # Calcular acumuladas
    prob_acum = np.cumsum(prob_ind)
    # Determinar el núcleo (nucleus)
    mask = prob_acum <= p
    # Siempre incluir la palabra donde supera p
    if not mask[-1]:
        idx_first_above = np.argmax(prob_acum > p)
        mask[idx_first_above] = True
    colores = ['#00b894' if m else '#dff9fb' for m in mask]
    edge_colors = ['#0984e3' if m else '#636e72' for m in mask]
    plt.figure(figsize=(8,3))
    bars = plt.bar(palabras, prob_ind, color=colores, edgecolor=edge_colors, linewidth=2)
    plt.title(f"Núcleo de palabras para Top-p = {p:.2f}", fontsize=13)
    plt.ylabel("Probabilidad individual")
    plt.ylim(0, 0.5)
    plt.grid(axis='y', linestyle=':', alpha=0.5)
    # Mostrar probabilidad acumulada sobre las barras
    prob_labels = [f"{acum:.2f}" for acum in prob_acum]
    for bar, label, m in zip(bars, prob_labels, mask):
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.015,
                 f"P={label}", ha='center', va='bottom', color=('#0652dd' if m else '#636e72'), fontsize=10)
    # Etiqueta visual para las barras núcleo
    for i, m in enumerate(mask):
        if m:
            plt.text(i, prob_ind[i]+0.07, "núcleo", color="#00b894", ha='center', fontsize=11, fontweight='bold')
    plt.tight_layout()
    plt.show()

widgets.interact(
    visualizar_top_p,
    p=widgets.FloatSlider(
        value=0.9, min=0.2, max=1.0, step=0.01,
        description="Valor de p (umbral):",
        readout_format='.2f',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='70%')
    )
)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

# --- Funciones Auxiliares ---
def softmax(x):
    """Calcula softmax de forma estable."""
    if not isinstance(x, np.ndarray):
        x = np.array(x, dtype=float)
    if x.size == 0:
        return np.array([])

    # Restar el máximo para estabilidad numérica
    x_exp = np.exp(x - np.max(x))
    sum_x_exp = np.sum(x_exp)

    if sum_x_exp == 0: # Evitar división por cero si todos los exp son ~0
        return np.zeros_like(x)
    return x_exp / sum_x_exp

# --- Función Principal para Graficar y Explicar ---
def plot_and_explain_penalty_effect(
    tokens_str="café, té, agua, jugo, mate, refresco, leche",
    logits_original_str="2.0, 1.8, 1.3, 0.8, 0.5, 1.1, 0.9",
    frecuencias_iniciales_str="3, 1, 0, 2, 4, 0, 1",
    frequency_penalty=0.8
):
    # Limpiar el output anterior del gráfico y HTML
    # El clear_output se gestionará por interactive_output, pero lo dejamos por si se llama manualmente.
    # clear_output(wait=True)

    # Output para mensajes de error o información
    # Esto se manejará mejor si la función devuelve los elementos a mostrar.
    # Por ahora, lo mantenemos simple.

    # 1. Parseo y Validación de Entradas
    error_messages = []
    try:
        tokens = [t.strip() for t in tokens_str.split(',') if t.strip()]
        if not tokens:
            error_messages.append("La lista de tokens no puede estar vacía.")
    except Exception as e:
        error_messages.append(f"Error al parsear tokens: {e}")
        tokens = []

    try:
        logits_original = np.array([float(x.strip()) for x in logits_original_str.split(',') if x.strip()])
    except Exception as e:
        error_messages.append(f"Error al parsear logits originales: {e}")
        logits_original = np.array([])

    try:
        frecuencias_iniciales = np.array([int(x.strip()) for x in frecuencias_iniciales_str.split(',') if x.strip()])
    except Exception as e:
        error_messages.append(f"Error al parsear frecuencias iniciales: {e}")
        frecuencias_iniciales = np.array([])

    if not error_messages: # Solo si no hay errores de parseo individuales
        if not (len(tokens) == len(logits_original) == len(frecuencias_iniciales)):
            error_messages.append(
                f"Las longitudes de tokens ({len(tokens)}), "
                f"logits ({len(logits_original)}), y frecuencias ({len(frecuencias_iniciales)}) "
                "deben coincidir."
            )

    # Si hay errores, mostrarlos y un gráfico de error
    if error_messages:
        print("\n--- Errores en la Entrada ---")
        for msg in error_messages:
            print(f"- {msg}")
        print("Por favor, corrige los valores en los campos de arriba.")

        plt.figure(figsize=(10, 6))
        plt.text(0.5, 0.5, "Error en los datos de entrada.\nRevisa los mensajes y corrige los campos.",
                 horizontalalignment='center', verticalalignment='center',
                 fontsize=14, color='red', bbox=dict(facecolor='white', alpha=0.8, edgecolor='red'))
        plt.xticks([])
        plt.yticks([])
        plt.title("Error de Configuración", fontsize=16, color='red')
        plt.show()
        return

    # 2. Aplicar Penalización y Calcular Probabilidades
    frecuencias = frecuencias_iniciales.copy()
    penalizaciones = frequency_penalty * frecuencias
    logits_ajustados = logits_original - penalizaciones

    probs_original = softmax(logits_original)
    probs_ajustadas = softmax(logits_ajustados)

    # 3. Visualización
    ancho = 0.37
    x = np.arange(len(tokens))

    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(12, 7)) # Ajustar tamaño

    # Barras originales
    ax.bar(x - ancho/2, probs_original, ancho, label='Prob. Original', color='#b2bec3', alpha=0.7, edgecolor='#636e72', linewidth=1.5, zorder=1)
    # Barras ajustadas
    ax.bar(x + ancho/2, probs_ajustadas, ancho, label='Prob. Ajustada (con penalización)', color='#00b894', alpha=0.9, edgecolor='#0984e3', linewidth=1.5, zorder=2)

    # Texto encima de cada barra ajustada y detalles debajo
    for i in range(len(tokens)):
        # Probabilidad ajustada
        ax.text(x[i] + ancho/2, probs_ajustadas[i] + 0.008,
                f"{probs_ajustadas[i]:.3f}", ha='center', va='bottom', fontsize=9, color="#005A64", fontweight='bold')
        # Frecuencia y Penalización
        ax.text(x[i], -0.045, # Ajustar posición y contenido
                f"F:{frecuencias[i]} P:{penalizaciones[i]:.1f}",
                ha='center', va='top', fontsize=8.5,
                color="#c0392b" if frecuencias[i] > 0 and penalizaciones[i] != 0 else "#34495e",
                fontweight='bold' if frecuencias[i] > 0 and penalizaciones[i] != 0 else 'normal')

    ax.set_xticks(x)
    ax.set_xticklabels(tokens, fontsize=11, rotation=35, ha="right") # Rotar para mejor lectura

    max_prob_val = 0
    if probs_original.size > 0: max_prob_val = max(max_prob_val, np.max(probs_original))
    if probs_ajustadas.size > 0: max_prob_val = max(max_prob_val, np.max(probs_ajustadas))

    y_lim_upper = max(0.1, max_prob_val * 1.20) # Límite Y dinámico con un mínimo
    ax.set_ylim(-0.10, y_lim_upper) # Espacio inferior para texto

    ax.set_ylabel("Probabilidad de Selección", fontsize=13)
    ax.set_title(f"Efecto de Frequency Penalty = {frequency_penalty:.2f}", fontsize=16, weight='bold', color='#2c3e50')
    ax.legend(fontsize=11, loc='upper right')
    ax.grid(True, axis='y', linestyle=':', alpha=0.7)

    plt.tight_layout(rect=[0, 0.05, 1, 0.95]) # Ajustar para que todo quepa
    plt.show()

    # 4. Explicación Didáctica Interactiva
    explicacion_html = f"""
    <div style="font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif; max-width:850px; margin:20px auto; background:#f8f9fa; border-left:6px solid #28a745; padding:20px 35px; border-radius:10px; box-shadow: 0 4px 8px rgba(0,0,0,0.1);">
        <h3 style="margin-top:0; color:#1a535c; border-bottom: 2px solid #4ecdc4; padding-bottom:10px;">Entendiendo la Penalización por Frecuencia (<span style='color:#28a745;'>frequency_penalty</span>)</h3>
        <ul style="line-height:1.7; padding-left:20px;">
            <li>Cada palabra (<b>token</b>) tiene una <b>frecuencia de aparición</b> (<code>F=...</code> debajo de cada barra).</li>
            <li>La <b>penalización aplicada</b> (<code>P=...</code>) es <code>frequency_penalty × frecuencia</code>. Esta penalización se resta de los 'logits' (puntuaciones internas) originales del modelo.</li>
            <li>Las barras <span style="color:#6c757d; font-weight:bold;">grises</span> indican la probabilidad <u>original</u> (sin penalización).</li>
            <li>Las barras <span style="color:#00b894; font_weight:bold;">verdes</span> muestran la probabilidad <u>ajustada</u> después de aplicar la penalización.</li>
            <li>Tokens con mayor frecuencia (y por ende, mayor penalización) ven su probabilidad de ser seleccionados <b>reducida significativamente</b>.</li>
            <li>Experimenta cambiando el valor de <code>frequency_penalty</code> y los campos de <b>Tokens, Logits y Frecuencias</b> para observar cómo se altera la distribución de probabilidad.</li>
        </ul>
        <p style="font-size:0.95em; color:#495057; margin-top:15px;">
            <b>Objetivo:</b> El <code>frequency_penalty</code> es un hiperparámetro crucial en modelos de lenguaje generativos. Ayuda a <b>reducir la repetición</b> de palabras y frases, fomentando la generación de texto más <b>diverso, natural y menos predecible</b>.
        </p>
    </div>
    """
    display(HTML(explicacion_html))

# --- Configuración de Widgets ---
style_desc = {'description_width': '200px'} # Ajustar ancho de descripción
layout_text = widgets.Layout(width='calc(100% - 210px)', margin='5px 0px 5px 5px') # Ajustar ancho para que quepa con descripción
layout_slider = widgets.Layout(width='calc(100% - 210px)', margin='10px 0px 20px 5px')

tokens_widget = widgets.Text(
    value="café, té, agua, jugo, mate, refresco, leche",
    description="Tokens:",
    style=style_desc, layout=layout_text
)
logits_widget = widgets.Text(
    value="2.0, 1.8, 1.3, 0.8, 0.5, 1.1, 0.9",
    description="Logits Originales:",
    style=style_desc, layout=layout_text
)
frecuencias_widget = widgets.Text(
    value="3, 1, 0, 2, 4, 0, 1",
    description="Frecuencias Iniciales:",
    style=style_desc, layout=layout_text
)
penalty_widget = widgets.FloatSlider(
    value=0.8, min=0.0, max=3.0, step=0.05,
    description="Penalización Frecuencia:",
    readout_format='.2f',
    style=style_desc, layout=layout_slider,
    continuous_update=False # Actualizar solo al soltar
)

# --- Ensamblaje de la Interfaz ---
input_section_title = widgets.HTML(
    "<h2>🔧 Configuración de Parámetros de Penalización por Frecuencia</h2>"
    "<p>Modifica los tokens, sus logits (puntuaciones base), sus frecuencias de aparición previas, y el factor de penalización.</p><hr>"
)

# Usar HBox para alinear descripción y widget en la misma línea si el espacio lo permite
# (o VBox si se prefiere descripción arriba)
# Para inputs de texto, la descripción ya está integrada.
controls_box = widgets.VBox([
    tokens_widget,
    logits_widget,
    frecuencias_widget,
    penalty_widget
])

# El output donde se mostrará el gráfico y la explicación
# interactive_output maneja clear_output(wait=True) internamente para su celda de salida.
output_display_area = widgets.interactive_output(
    plot_and_explain_penalty_effect,
    {
        'tokens_str': tokens_widget,
        'logits_original_str': logits_widget,
        'frecuencias_iniciales_str': frecuencias_widget,
        'frequency_penalty': penalty_widget
    }
)

# Mostrar la interfaz completa
display(input_section_title, controls_box, output_display_area)
```

```python
# Importar las funciones necesarias de IPython para mostrar HTML
from IPython.core.display import display, HTML

# Definir el contenido HTML del Glosario con diseño mejorado
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Glosario de Términos - Clase 1: Inteligencia Artificial</title>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">
  <style>
    :root {
      /* Tema claro */
      --bg-primary: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
      --bg-secondary: #ffffff;
      --bg-tertiary: #f8fafc;
      --bg-quaternary: #f1f5f9;
      --text-primary: #1e293b;
      --text-secondary: #64748b;
      --text-accent: #3b82f6;
      --border-color: #e2e8f0;
      --shadow-sm: 0 1px 2px 0 rgb(0 0 0 / 0.05);
      --shadow-md: 0 4px 6px -1px rgb(0 0 0 / 0.1), 0 2px 4px -2px rgb(0 0 0 / 0.1);
      --shadow-lg: 0 10px 15px -3px rgb(0 0 0 / 0.1), 0 4px 6px -4px rgb(0 0 0 / 0.1);
      --shadow-xl: 0 20px 25px -5px rgb(0 0 0 / 0.1), 0 8px 10px -6px rgb(0 0 0 / 0.1);

      --color-blue: #3b82f6;
      --color-emerald: #10b981;
      --color-purple: #8b5cf6;
      --color-rose: #f43f5e;
      --color-amber: #f59e0b;
      --color-indigo: #6366f1;
      --color-cyan: #06b6d4;
    }

    [data-theme="dark"] {
      --bg-primary: linear-gradient(135deg, #0f172a 0%, #1e293b 100%);
      --bg-secondary: #1e293b;
      --bg-tertiary: #334155;
      --bg-quaternary: #475569;
      --text-primary: #f1f5f9;
      --text-secondary: #94a3b8;
      --text-accent: #60a5fa;
      --border-color: #475569;
      --shadow-sm: 0 1px 2px 0 rgb(0 0 0 / 0.3);
      --shadow-md: 0 4px 6px -1px rgb(0 0 0 / 0.3), 0 2px 4px -2px rgb(0 0 0 / 0.3);
      --shadow-lg: 0 10px 15px -3px rgb(0 0 0 / 0.3), 0 4px 6px -4px rgb(0 0 0 / 0.3);
      --shadow-xl: 0 20px 25px -5px rgb(0 0 0 / 0.3), 0 8px 10px -6px rgb(0 0 0 / 0.3);
    }

    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    html {
        scroll-behavior: smooth;
    }

    body {
      font-family: 'Inter', -apple-system, BlinkMacSystemFont, sans-serif;
      line-height: 1.7;
      background: var(--bg-primary);
      color: var(--text-primary);
      min-height: 100vh;
      overflow-x: hidden;
      transition: background 0.3s ease, color 0.3s ease;
    }

    .container {
      max-width: 1200px;
      margin: 0 auto;
      padding: 2rem;
      position: relative;
    }

    .header {
      text-align: center;
      margin-bottom: 3rem;
      position: relative;
      z-index: 10;
    }

    .header::before {
      content: '';
      position: absolute;
      top: -50px;
      left: 50%;
      transform: translateX(-50%);
      width: 120px;
      height: 120px;
      background: var(--color-purple);
      border-radius: 50%;
      opacity: 0.1;
      animation: pulse 3s ease-in-out infinite;
    }

    @keyframes pulse {
      0%, 100% { transform: translateX(-50%) scale(1); opacity: 0.1; }
      50% { transform: translateX(-50%) scale(1.2); opacity: 0.2; }
    }

    .main-title {
      font-size: clamp(2.5rem, 5vw, 3.5rem);
      font-weight: 700;
      background: linear-gradient(135deg, var(--color-purple), var(--color-indigo));
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
      margin-bottom: 0.5rem;
      position: relative;
    }

    .subtitle {
      font-size: 1.25rem;
      color: var(--text-secondary);
      font-weight: 400;
      max-width: 600px;
      margin: 0 auto 1.5rem auto;
    }

    .academic-info {
      display: flex;
      align-items: center;
      justify-content: center;
      flex-wrap: wrap;
      gap: 1rem;
      font-size: 0.95rem;
      color: var(--text-secondary);
      background: var(--bg-secondary);
      padding: 1rem 1.5rem;
      border-radius: 1rem;
      border: 1px solid var(--border-color);
      box-shadow: var(--shadow-sm);
      max-width: fit-content;
      margin: 0 auto;
      transition: all 0.3s ease;
    }

    .academic-info:hover {
      transform: translateY(-2px);
      box-shadow: var(--shadow-md);
      border-color: var(--text-accent);
    }

    .academic-badge {
      display: flex;
      align-items: center;
      gap: 0.5rem;
      padding: 0.5rem 0.75rem;
      background: var(--bg-tertiary);
      border-radius: 0.5rem;
      border: 1px solid var(--border-color);
      font-weight: 500;
    }

    .academic-badge i {
      color: var(--text-accent);
      font-size: 0.9rem;
    }

    .academic-badge strong {
      color: var(--text-primary);
    }

    .theme-toggle {
      position: fixed;
      top: 2rem;
      right: 2rem;
      width: 60px;
      height: 60px;
      border: 2px solid var(--border-color);
      border-radius: 50%;
      background: var(--bg-secondary);
      box-shadow: var(--shadow-lg);
      cursor: pointer;
      display: flex;
      align-items: center;
      justify-content: center;
      font-size: 1.5rem;
      color: var(--text-primary);
      transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
      z-index: 1000;
      backdrop-filter: blur(10px);
    }

    .theme-toggle:hover {
      transform: translateY(-2px) scale(1.05);
      box-shadow: var(--shadow-xl);
      background: var(--text-accent);
      color: white;
    }

    .theme-toggle:active {
      transform: translateY(0) scale(0.95);
    }

    .theme-icon {
      transition: all 0.3s ease;
      position: absolute;
    }

    .theme-toggle .fa-sun {
      opacity: 0;
      transform: rotate(-90deg) scale(0.5);
    }

    .theme-toggle .fa-moon {
      opacity: 1;
      transform: rotate(0deg) scale(1);
    }

    [data-theme="dark"] .theme-toggle .fa-moon {
      opacity: 0;
      transform: rotate(90deg) scale(0.5);
    }

    [data-theme="dark"] .theme-toggle .fa-sun {
      opacity: 1;
      transform: rotate(0deg) scale(1);
    }

    .stats-container {
      display: grid;
      grid-template-columns: repeat(auto-fit, minmax(200px, 1fr));
      gap: 1rem;
      margin-bottom: 2rem;
    }

    .stat-card {
      background: var(--bg-secondary);
      padding: 1.5rem;
      border-radius: 1rem;
      border: 1px solid var(--border-color);
      box-shadow: var(--shadow-sm);
      text-align: center;
      transition: all 0.3s ease;
    }

    .stat-card:hover {
      transform: translateY(-2px);
      box-shadow: var(--shadow-md);
    }

    .stat-number {
      font-size: 2rem;
      font-weight: 700;
      color: var(--color-purple);
      display: block;
    }

    .stat-label {
      font-size: 0.9rem;
      color: var(--text-secondary);
      margin-top: 0.5rem;
    }

    .search-container {
      margin-bottom: 2rem;
      position: relative;
      max-width: 500px;
      margin-left: auto;
      margin-right: auto;
    }

    .search-input {
      width: 100%;
      padding: 1rem 1rem 1rem 3rem;
      font-size: 1rem;
      border: 2px solid var(--border-color);
      border-radius: 2rem;
      background: var(--bg-secondary);
      color: var(--text-primary);
      transition: all 0.3s ease;
      outline: none;
    }

    .search-input:focus {
      border-color: var(--color-purple);
      box-shadow: 0 0 0 3px rgba(139, 92, 246, 0.1);
      transform: translateY(-1px);
    }

    .search-icon {
      position: absolute;
      left: 1.25rem;
      top: 50%;
      transform: translateY(-50%);
      color: var(--text-secondary);
      font-size: 1rem;
      transition: color 0.3s ease;
    }

    .search-input:focus + .search-icon {
      color: var(--color-purple);
    }

    .alphabet-filter {
      display: flex;
      justify-content: center;
      flex-wrap: wrap;
      gap: 0.5rem;
      margin-bottom: 2rem;
      padding: 1.5rem;
      background: var(--bg-secondary);
      border-radius: 1rem;
      box-shadow: var(--shadow-md);
      border: 1px solid var(--border-color);
    }

    .letter-btn {
      width: 40px;
      height: 40px;
      border: 1px solid var(--border-color);
      border-radius: 50%;
      background: var(--bg-tertiary);
      color: var(--text-secondary);
      font-weight: 500;
      cursor: pointer;
      transition: all 0.3s ease;
      display: flex;
      align-items: center;
      justify-content: center;
    }

    .letter-btn:hover {
      background: var(--color-purple);
      color: white;
      transform: translateY(-2px);
      box-shadow: var(--shadow-md);
      border-color: var(--color-purple);
    }

    .letter-btn.active {
      background: var(--color-purple);
      color: white;
      border-color: var(--color-purple);
      box-shadow: var(--shadow-md);
    }

    .clear-filter {
      padding: 0.5rem 1rem;
      background: var(--color-rose);
      color: white;
      border: none;
      border-radius: 1rem;
      font-size: 0.9rem;
      cursor: pointer;
      transition: all 0.3s ease;
      font-weight: 500;
      display: inline-flex;
      align-items: center;
      gap: 0.3rem;
    }

    .clear-filter:hover {
      background: #dc2626;
      transform: translateY(-1px);
      box-shadow: var(--shadow-sm);
    }

    .term-counter {
      text-align: center;
      margin-bottom: 1.5rem;
      padding: 0.75rem 1.5rem;
      background: var(--bg-secondary);
      border-radius: 2rem;
      border: 1px solid var(--border-color);
      color: var(--text-secondary);
      font-size: 0.95rem;
      font-weight: 500;
      max-width: fit-content;
      margin-left: auto;
      margin-right: auto;
      box-shadow: var(--shadow-sm);
      display: inline-flex;
      align-items: center;
      gap: 0.5rem;
    }

    .glossary-grid {
      display: grid;
      grid-template-columns: repeat(auto-fit, minmax(350px, 1fr));
      gap: 1.5rem;
      margin-top: 2rem;
    }

    .term-card {
      background: var(--bg-secondary);
      border-radius: 1rem;
      padding: 1.5rem;
      box-shadow: var(--shadow-md);
      border: 1px solid var(--border-color);
      transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
      position: relative;
      overflow: hidden;
    }

    .term-card::before {
      content: '';
      position: absolute;
      top: 0;
      left: 0;
      width: 4px;
      height: 100%;
      background: linear-gradient(45deg, var(--color-purple), var(--color-indigo));
      transition: width 0.3s ease;
    }

    .term-card:hover {
      transform: translateY(-4px);
      box-shadow: var(--shadow-xl);
      border-color: var(--color-purple);
    }

    .term-card:hover::before {
      width: 8px;
    }

    .term-name {
      font-size: 1.25rem;
      font-weight: 600;
      color: var(--color-purple);
      margin-bottom: 0.75rem;
      display: flex;
      align-items: center;
      gap: 0.5rem;
    }

    .term-icon {
      font-size: 1rem;
      color: var(--color-indigo);
      flex-shrink: 0;
    }

    .term-definition {
      color: var(--text-primary);
      line-height: 1.8;
      font-size: 0.95rem;
    }

    .term-category {
      position: absolute;
      top: 1rem;
      right: 1rem;
      padding: 0.25rem 0.75rem;
      background: var(--color-amber);
      color: white;
      font-size: 0.75rem;
      font-weight: 500;
      border-radius: 1rem;
      text-transform: uppercase;
      letter-spacing: 0.5px;
      box-shadow: var(--shadow-sm);
    }

    .floating-shapes {
      position: fixed;
      top: 0;
      left: 0;
      width: 100%;
      height: 100%;
      pointer-events: none;
      z-index: -1;
      overflow: hidden;
    }

    .shape {
      position: absolute;
      border-radius: 50%;
      opacity: 0.03;
      animation: float 8s ease-in-out infinite;
      backdrop-filter: blur(1px);
    }

    .shape:nth-child(1) {
      width: 100px;
      height: 100px;
      top: 20%;
      left: 10%;
      background: var(--color-purple);
      animation-delay: 0s;
    }

    .shape:nth-child(2) {
      width: 150px;
      height: 150px;
      top: 60%;
      right: 15%;
      background: var(--color-indigo);
      animation-delay: 2s;
    }

    .shape:nth-child(3) {
      width: 80px;
      height: 80px;
      bottom: 20%;
      left: 20%;
      background: var(--color-cyan);
      animation-delay: 4s;
    }

    .shape:nth-child(4) {
      width: 120px;
      height: 120px;
      top: 40%;
      right: 40%;
      background: var(--color-emerald);
      animation-delay: 6s;
    }

    @keyframes float {
      0%, 100% { transform: translateY(0) rotate(0deg); }
      25% { transform: translateY(-30px) rotate(90deg); }
      50% { transform: translateY(-20px) rotate(180deg); }
      75% { transform: translateY(-40px) rotate(270deg); }
    }

    .term-card.hidden {
      display: none;
    }

    .term-card.visible {
      display: block;
      animation: fadeInCard 0.5s ease forwards;
    }

    @keyframes fadeInCard {
      from { opacity: 0; transform: translateY(10px); }
      to { opacity: 1; transform: translateY(0); }
    }

    .no-results {
      text-align: center;
      padding: 3rem;
      color: var(--text-secondary);
      font-size: 1.1rem;
      grid-column: 1 / -1;
      background: var(--bg-secondary);
      border-radius: 1rem;
      border: 2px dashed var(--border-color);
      display: flex;
      flex-direction: column;
      align-items: center;
      justify-content: center;
    }

    .no-results i {
      font-size: 3rem;
      margin-bottom: 1rem;
      color: var(--color-purple);
      opacity: 0.3;
    }

    @media (max-width: 768px) {
      .container {
        padding: 1rem;
      }

      .glossary-grid {
        grid-template-columns: 1fr;
        gap: 1rem;
      }

      .alphabet-filter {
        padding: 1rem;
        gap: 0.25rem;
      }

      .letter-btn {
        width: 35px;
        height: 35px;
        font-size: 0.9rem;
      }

      .theme-toggle {
        width: 50px;
        height: 50px;
        top: 1rem;
        right: 1rem;
        font-size: 1.25rem;
      }

      .main-title {
        font-size: 2rem;
      }

      .stats-container {
        grid-template-columns: repeat(auto-fit, minmax(150px, 1fr));
      }

      .academic-info {
        flex-direction: column;
        align-items: stretch;
      }

      .academic-badge {
        justify-content: center;
      }
    }

    .fade-in {
      opacity: 0;
      transform: translateY(20px);
      animation: fadeInUp 0.6s ease forwards;
    }

    .term-card.fade-in {
        animation: fadeInUp 0.6s ease forwards;
        animation-delay: calc(var(--card-index) * 0.05s);
    }

    @keyframes fadeInUp {
      to {
        opacity: 1;
        transform: translateY(0);
      }
    }

    ::-webkit-scrollbar {
      width: 8px;
    }

    ::-webkit-scrollbar-track {
      background: var(--bg-tertiary);
      border-radius: 4px;
    }

    ::-webkit-scrollbar-thumb {
      background: var(--text-secondary);
      border-radius: 4px;
      transition: background 0.3s ease;
    }

    ::-webkit-scrollbar-thumb:hover {
      background: var(--text-primary);
    }
  </style>
</head>
<body data-theme="light">
  <div class="floating-shapes">
    <div class="shape"></div>
    <div class="shape"></div>
    <div class="shape"></div>
    <div class="shape"></div>
  </div>

  <button class="theme-toggle" id="themeToggleButton" title="Cambiar tema" aria-label="Cambiar tema">
    <i class="fas fa-moon theme-icon"></i>
    <i class="fas fa-sun theme-icon"></i>
  </button>

  <div class="container">
    <header class="header">
      <h1 class="main-title fade-in">Glosario de Términos</h1>
      <p class="subtitle fade-in">Clase 1: Inteligencia Artificial</p>

      <div class="academic-info fade-in">
        <div class="academic-badge">
          <i class="fas fa-graduation-cap"></i>
          <span><strong>Curso:</strong> IA Fundamentals</span>
        </div>
        <div class="academic-badge">
          <i class="fas fa-book-open"></i>
          <span><strong>Clase:</strong> 1</span>
        </div>
      </div>
    </header>

    <!-- Sección eliminada: colab-notice -->

    <div class="stats-container fade-in">
      <div class="stat-card">
        <span class="stat-number" id="totalTerms">20</span>
        <div class="stat-label">Términos Totales</div>
      </div>
      <div class="stat-card">
        <span class="stat-number" id="categoriesCount">8</span>
        <div class="stat-label">Categorías</div>
      </div>
      <div class="stat-card">
        <span class="stat-number" id="visibleTerms">20</span>
        <div class="stat-label">Términos Visibles</div>
      </div>
    </div>

    <div class="search-container fade-in">
      <input type="text" class="search-input" placeholder="Buscar término o definición..." id="searchInput" autocomplete="off">
      <i class="fas fa-search search-icon"></i>
    </div>

    <div class="alphabet-filter fade-in">
      <button class="letter-btn" data-letter="A">A</button>
      <button class="letter-btn" data-letter="B">B</button>
      <button class="letter-btn" data-letter="C">C</button>
      <button class="letter-btn" data-letter="D">D</button>
      <button class="letter-btn" data-letter="E">E</button>
      <button class="letter-btn" data-letter="F">F</button>
      <button class="letter-btn" data-letter="G">G</button>
      <button class="letter-btn" data-letter="I">I</button>
      <button class="letter-btn" data-letter="M">M</button>
      <button class="letter-btn" data-letter="N">N</button>
      <button class="letter-btn" data-letter="R">R</button>
      <button class="letter-btn" data-letter="S">S</button>
      <button class="letter-btn" data-letter="T">T</button>
      <button class="letter-btn" data-letter="V">V</button>
      <button class="clear-filter" id="clearFilterButton">
        <i class="fas fa-times"></i> Limpiar
      </button>
    </div>

    <div class="term-counter fade-in">
      <i class="fas fa-list-ul"></i>
      <span id="termCount">20 términos encontrados</span>
    </div>

    <div class="glossary-grid" id="glossaryGrid">

      <div class="term-card fade-in" data-term="algoritmo" style="--card-index: 0;">
        <div class="term-category">Fundamentos</div>
        <h3 class="term-name">
          <i class="fas fa-cogs term-icon"></i>
          Algoritmo
        </h3>
        <p class="term-definition">
          Conjunto finito de pasos definidos y ordenados que permiten resolver un problema o realizar una tarea específica.
        </p>
      </div>
      <div class="term-card fade-in" data-term="análisis de sentimientos" style="--card-index: 1;">
        <div class="term-category">NLP</div>
        <h3 class="term-name">
          <i class="fas fa-heart term-icon"></i>
          Análisis de sentimientos
        </h3>
        <p class="term-definition">
          Técnica del procesamiento del lenguaje natural (NLP) que permite identificar y clasificar emociones (positivas, negativas o neutras) expresadas en un texto.
        </p>
      </div>
      <div class="term-card fade-in" data-term="aprendizaje automático" style="--card-index: 2;">
        <div class="term-category">ML</div>
        <h3 class="term-name">
          <i class="fas fa-robot term-icon"></i>
          Aprendizaje automático (Machine Learning)
        </h3>
        <p class="term-definition">
          Subcampo de la inteligencia artificial que permite que las máquinas aprendan patrones a partir de datos, sin ser programadas explícitamente para cada tarea.
        </p>
      </div>
      <div class="term-card fade-in" data-term="aprendizaje profundo" style="--card-index: 3;">
        <div class="term-category">DL</div>
        <h3 class="term-name">
          <i class="fas fa-brain term-icon"></i>
          Aprendizaje profundo (Deep Learning)
        </h3>
        <p class="term-definition">
          Rama del aprendizaje automático que utiliza redes neuronales artificiales con múltiples capas para modelar representaciones complejas y resolver tareas como clasificación, traducción o reconocimiento de imágenes.
        </p>
      </div>
      <div class="term-card fade-in" data-term="bert" style="--card-index: 4;">
        <div class="term-category">Modelos</div>
        <h3 class="term-name">
          <i class="fas fa-language term-icon"></i>
          BERT (Bidirectional Encoder Representations from Transformers)
        </h3>
        <p class="term-definition">
          Modelo de lenguaje desarrollado por Google basado en la arquitectura Transformer. Está entrenado para comprender el contexto de una palabra teniendo en cuenta su posición dentro de toda la frase, en ambas direcciones (izquierda y derecha).
        </p>
      </div>
      <div class="term-card fade-in" data-term="clasificación" style="--card-index: 5;">
        <div class="term-category">ML</div>
        <h3 class="term-name">
          <i class="fas fa-tags term-icon"></i>
          Clasificación
        </h3>
        <p class="term-definition">
          Problema de aprendizaje supervisado donde el objetivo es asignar una etiqueta o clase a una entrada (por ejemplo, spam o no spam).
        </p>
      </div>
      <div class="term-card fade-in" data-term="dataset" style="--card-index: 6;">
        <div class="term-category">Datos</div>
        <h3 class="term-name">
          <i class="fas fa-database term-icon"></i>
          Dataset
        </h3>
        <p class="term-definition">
          Conjunto estructurado de datos que se utiliza para entrenar, validar o probar modelos de inteligencia artificial.
        </p>
      </div>
      <div class="term-card fade-in" data-term="embeddings" style="--card-index: 7;">
        <div class="term-category">NLP</div>
        <h3 class="term-name">
          <i class="fas fa-vector-square term-icon"></i>
          Embeddings
        </h3>
        <p class="term-definition">
          Representación numérica densa y de baja dimensión de palabras o frases que captura relaciones semánticas y sintácticas entre ellas. Son fundamentales en tareas de NLP.
        </p>
      </div>
      <div class="term-card fade-in" data-term="feature extraction" style="--card-index: 8;">
        <div class="term-category">Procesamiento</div>
        <h3 class="term-name">
          <i class="fas fa-filter term-icon"></i>
          Feature Extraction (Extracción de características)
        </h3>
        <p class="term-definition">
          Proceso mediante el cual se transforman datos brutos en información útil que puede ser procesada por algoritmos de aprendizaje.
        </p>
      </div>
      <div class="term-card fade-in" data-term="inteligencia artificial" style="--card-index: 9;">
        <div class="term-category">Fundamentos</div>
        <h3 class="term-name">
          <i class="fas fa-microchip term-icon"></i>
          Inteligencia Artificial (IA)
        </h3>
        <p class="term-definition">
          Disciplina de la informática que se ocupa de crear sistemas capaces de realizar tareas que normalmente requieren inteligencia humana, como el razonamiento, aprendizaje o comprensión del lenguaje.
        </p>
      </div>
      <div class="term-card fade-in" data-term="machine learning" style="--card-index: 10;">
        <div class="term-category">ML</div>
        <h3 class="term-name">
          <i class="fas fa-chart-line term-icon"></i>
          Machine Learning (ML)
        </h3>
        <p class="term-definition">
          (Ver Aprendizaje automático) Subcampo de la inteligencia artificial que permite que las máquinas aprendan patrones a partir de datos, sin ser programadas explícitamente para cada tarea.
        </p>
      </div>
      <div class="term-card fade-in" data-term="modelo supervisado" style="--card-index: 11;">
        <div class="term-category">ML</div>
        <h3 class="term-name">
          <i class="fas fa-graduation-cap term-icon"></i>
          Modelo supervisado
        </h3>
        <p class="term-definition">
          Modelo de aprendizaje que se entrena con datos etiquetados, es decir, cada ejemplo de entrenamiento incluye tanto la entrada como la salida esperada.
        </p>
      </div>
      <div class="term-card fade-in" data-term="naive bayes" style="--card-index: 12;">
        <div class="term-category">Algoritmos</div>
        <h3 class="term-name">
          <i class="fas fa-calculator term-icon"></i>
          Naive Bayes
        </h3>
        <p class="term-definition">
          Algoritmo de clasificación basado en el Teorema de Bayes, que asume independencia entre las características. Es especialmente útil para tareas de clasificación de texto.
        </p>
      </div>
      <div class="term-card fade-in" data-term="nlp" style="--card-index: 13;">
        <div class="term-category">NLP</div>
        <h3 class="term-name">
          <i class="fas fa-comments term-icon"></i>
          NLP (Natural Language Processing / Procesamiento del Lenguaje Natural)
        </h3>
        <p class="term-definition">
          Campo de la IA que se encarga de la interacción entre computadoras y lenguaje humano. Incluye tareas como análisis de sentimientos, traducción automática, generación de texto y reconocimiento de voz.
        </p>
      </div>
      <div class="term-card fade-in" data-term="red neuronal" style="--card-index: 14;">
        <div class="term-category">DL</div>
        <h3 class="term-name">
          <i class="fas fa-project-diagram term-icon"></i>
          Red neuronal
        </h3>
        <p class="term-definition">
          Modelo computacional inspirado en el cerebro humano, formado por unidades llamadas neuronas artificiales que procesan y transmiten información entre capas conectadas.
        </p>
      </div>
      <div class="term-card fade-in" data-term="scikit-learn" style="--card-index: 15;">
        <div class="term-category">Herramientas</div>
        <h3 class="term-name">
          <i class="fab fa-python term-icon"></i>
          Scikit-learn
        </h3>
        <p class="term-definition">
          Biblioteca de Python utilizada para realizar tareas de aprendizaje automático como clasificación, regresión, clustering y reducción de dimensionalidad. Muy usada en enseñanza e investigación.
        </p>
      </div>
      <div class="term-card fade-in" data-term="spam" style="--card-index: 16;">
        <div class="term-category">Aplicaciones</div>
        <h3 class="term-name">
          <i class="fas fa-envelope term-icon"></i>
          Spam
        </h3>
        <p class="term-definition">
          Correo electrónico no deseado, comúnmente utilizado como ejemplo en clasificación automática de textos.
        </p>
      </div>
      <div class="term-card fade-in" data-term="test de turing" style="--card-index: 17;">
        <div class="term-category">Historia</div>
        <h3 class="term-name">
          <i class="fas fa-user-check term-icon"></i>
          Test de Turing
        </h3>
        <p class="term-definition">
          Propuesta realizada por Alan Turing para determinar si una máquina puede exhibir un comportamiento inteligente indistinguible del de un ser humano.
        </p>
      </div>
      <div class="term-card fade-in" data-term="tokenización" style="--card-index: 18;">
        <div class="term-category">NLP</div>
        <h3 class="term-name">
          <i class="fas fa-cut term-icon"></i>
          Tokenización
        </h3>
        <p class="term-definition">
          Proceso de dividir un texto en unidades más pequeñas llamadas “tokens”, como palabras, signos de puntuación o frases. Es un paso previo esencial en NLP.
        </p>
      </div>
      <div class="term-card fade-in" data-term="vectorización" style="--card-index: 19;">
        <div class="term-category">Procesamiento</div>
        <h3 class="term-name">
          <i class="fas fa-ruler-combined term-icon"></i>
          Vectorización
        </h3>
        <p class="term-definition">
          Conversión de texto en representaciones numéricas para que puedan ser procesadas por algoritmos de aprendizaje automático. Por ejemplo, con técnicas como Bag of Words o TF-IDF.
        </p>
      </div>

      <div class="no-results hidden">
        <i class="fas fa-search-minus"></i>
        <p>No se encontraron términos que coincidan con su búsqueda o filtro.</p>
      </div>

    </div>
  </div>

  <script>
    // IIFE para encapsular el script y evitar contaminación global
    (function() {
        // Esta variable simulará el localStorage para el tema dentro de esta instancia de la celda.
        let currentSessionTheme = 'light'; // Por defecto 'light'

        function applyInitialTheme() {
            document.body.setAttribute('data-theme', currentSessionTheme);
        }

        function toggleTheme() {
            const currentThemeFromBody = document.body.getAttribute('data-theme');
            const newTheme = currentThemeFromBody === 'dark' ? 'light' : 'dark';
            document.body.setAttribute('data-theme', newTheme);
            currentSessionTheme = newTheme;
        }

        const themeToggleButton = document.getElementById('themeToggleButton');
        const searchInput = document.getElementById('searchInput');
        const glossaryGrid = document.getElementById('glossaryGrid');
        const termCards = Array.from(glossaryGrid.querySelectorAll('.term-card:not(.no-results)'));
        const termCountElement = document.getElementById('termCount');
        const alphabetFilter = document.querySelector('.alphabet-filter');
        const letterButtons = alphabetFilter.querySelectorAll('.letter-btn');
        const clearFilterButton = document.getElementById('clearFilterButton');
        const noResultsElement = glossaryGrid.querySelector('.no-results');

        const totalTermsElement = document.getElementById('totalTerms');
        const categoriesCountElement = document.getElementById('categoriesCount');
        const visibleTermsElement = document.getElementById('visibleTerms');

        let currentFilterLetter = null;

        function countUniqueCategories() {
            const categories = new Set();
            termCards.forEach(card => {
                const categoryElement = card.querySelector('.term-category');
                if (categoryElement) {
                    categories.add(categoryElement.textContent.trim());
                }
            });
            return categories.size;
        }

        function updateStats(visibleCount) {
            if (totalTermsElement) totalTermsElement.textContent = termCards.length;
            if (categoriesCountElement) categoriesCountElement.textContent = countUniqueCategories();
            if (visibleTermsElement) visibleTermsElement.textContent = visibleCount;
        }

        function filterTerms() {
            if (!searchInput || !glossaryGrid) return;
            const searchTerm = searchInput.value.toLowerCase().trim();
            let visibleCount = 0;

            termCards.forEach(card => {
                const termName = card.dataset.term.toLowerCase();
                const termDefinitionElement = card.querySelector('.term-definition');
                if (!termDefinitionElement) return;
                const termDefinition = termDefinitionElement.textContent.toLowerCase();

                const matchesSearch = termName.includes(searchTerm) || termDefinition.includes(searchTerm);
                const matchesLetter = !currentFilterLetter || termName.startsWith(currentFilterLetter.toLowerCase());

                if (matchesSearch && matchesLetter) {
                    card.classList.remove('hidden');
                    card.classList.add('visible');
                    visibleCount++;
                } else {
                    card.classList.add('hidden');
                    card.classList.remove('visible');
                }
            });

            if(termCountElement) termCountElement.innerHTML = `<i class="fas fa-list-ul"></i> ${visibleCount} término${visibleCount !== 1 ? 's' : ''} encontrado${visibleCount !== 1 ? 's' : ''}`;
            updateStats(visibleCount);

            if (noResultsElement) {
                if (visibleCount === 0 && (searchTerm || currentFilterLetter)) {
                    noResultsElement.classList.remove('hidden');
                } else {
                    noResultsElement.classList.add('hidden');
                }
            }
        }

        function filterByLetter(letter) {
            if (currentFilterLetter === letter) {
                currentFilterLetter = null;
                letterButtons.forEach(btn => btn.classList.remove('active'));
            } else {
                currentFilterLetter = letter;
                letterButtons.forEach(btn => {
                    btn.classList.toggle('active', btn.dataset.letter === letter);
                });
            }
            filterTerms();
        }

        function clearAllFilters() {
            if(searchInput) searchInput.value = '';
            currentFilterLetter = null;
            letterButtons.forEach(btn => btn.classList.remove('active'));
            filterTerms();
        }

        if (themeToggleButton) {
            themeToggleButton.addEventListener('click', toggleTheme);
        }
        if (searchInput) {
            searchInput.addEventListener('input', filterTerms);
        }
        letterButtons.forEach(button => {
            button.addEventListener('click', () => filterByLetter(button.dataset.letter));
        });
        if (clearFilterButton) {
            clearFilterButton.addEventListener('click', clearAllFilters);
        }

        applyInitialTheme();
        filterTerms();

        termCards.forEach((card, index) => {
            card.style.setProperty('--card-index', index);
        });
    })();
  </script>
</body>
</html>
"""

# Mostrar el HTML en la salida de la celda
display(HTML(html_content))

```

```python
# Importar las funciones necesarias de IPython para mostrar HTML
from IPython.core.display import display, HTML

# Definir el contenido HTML de Preguntas y Respuestas
html_content_qa = """
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Preguntas y Respuestas - Clase 1: Inteligencia Artificial</title>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">
  <style>
    :root {
      /* Tema claro */
      --bg-primary-qa: linear-gradient(135deg, #52a7c1 0%, #2e7d8a 100%); /* Tonos azules/verdes */
      --bg-secondary-qa: #ffffff;
      --bg-tertiary-qa: #f0f9ff; /* Azul muy claro */
      --text-primary-qa: #1e293b; /* Azul oscuro/gris */
      --text-secondary-qa: #475569; /* Gris azulado */
      --text-accent-qa: #0ea5e9; /* Azul cielo */
      --border-color-qa: #cbd5e1; /* Gris claro */
      --shadow-sm-qa: 0 1px 2px 0 rgb(0 0 0 / 0.05);
      --shadow-md-qa: 0 4px 6px -1px rgb(0 0 0 / 0.1), 0 2px 4px -2px rgb(0 0 0 / 0.1);
      --shadow-lg-qa: 0 10px 15px -3px rgb(0 0 0 / 0.1), 0 4px 6px -4px rgb(0 0 0 / 0.1);

      --color-blue-qa: #0ea5e9;
      --color-green-qa: #22c55e;
      --color-purple-qa: #8b5cf6;
    }

    [data-theme="dark"] {
      --bg-primary-qa: linear-gradient(135deg, #1e3a8a 0%, #1e293b 100%); /* Azul oscuro a más oscuro */
      --bg-secondary-qa: #1e293b; /* Azul muy oscuro */
      --bg-tertiary-qa: #334155; /* Gris azulado oscuro */
      --text-primary-qa: #e2e8f0; /* Gris muy claro */
      --text-secondary-qa: #94a3b8; /* Gris claro */
      --text-accent-qa: #38bdf8; /* Azul cielo más brillante */
      --border-color-qa: #475569; /* Gris azulado oscuro */
      /* Sombras para modo oscuro (pueden ser más sutiles o usar colores oscuros) */
      --shadow-sm-qa: 0 1px 2px 0 rgba(0,0,0,0.2);
      --shadow-md-qa: 0 4px 6px -1px rgba(0,0,0,0.25), 0 2px 4px -2px rgba(0,0,0,0.2);
      --shadow-lg-qa: 0 10px 15px -3px rgba(0,0,0,0.25), 0 4px 6px -4px rgba(0,0,0,0.2);
    }

    * {
      margin: 0;
      padding: 0;
      box-sizing: border-box;
    }

    html {
        scroll-behavior: smooth;
    }

    body.qa-body { /* Clase específica para evitar conflictos si se usa en la misma página que el glosario */
      font-family: 'Inter', -apple-system, BlinkMacSystemFont, sans-serif;
      line-height: 1.8;
      background: var(--bg-primary-qa);
      color: var(--text-primary-qa);
      min-height: 100vh;
      overflow-x: hidden;
      transition: background 0.3s ease, color 0.3s ease;
    }

    .qa-container {
      max-width: 900px;
      margin: 0 auto;
      padding: 2rem;
      position: relative;
    }

    .qa-header {
      text-align: center;
      margin-bottom: 2.5rem;
      position: relative;
      z-index: 10;
    }

    .qa-main-title {
      font-size: clamp(2.2rem, 4.5vw, 3.2rem);
      font-weight: 700;
      background: linear-gradient(135deg, var(--color-blue-qa), var(--color-green-qa));
      -webkit-background-clip: text;
      -webkit-text-fill-color: transparent;
      background-clip: text;
      margin-bottom: 0.5rem;
      position: relative;
    }

    .qa-subtitle {
      font-size: 1.1rem;
      color: var(--text-secondary-qa);
      font-weight: 400;
      max-width: 600px;
      margin: 0 auto 1rem auto;
    }

    .qa-theme-toggle {
      position: fixed;
      top: 2rem;
      right: 2rem;
      width: 50px;
      height: 50px;
      border: 2px solid var(--border-color-qa);
      border-radius: 50%;
      background: var(--bg-secondary-qa);
      box-shadow: var(--shadow-md-qa);
      cursor: pointer;
      display: flex;
      align-items: center;
      justify-content: center;
      font-size: 1.25rem;
      color: var(--text-primary-qa);
      transition: all 0.3s cubic-bezier(0.4, 0, 0.2, 1);
      z-index: 1000;
    }

    .qa-theme-toggle:hover {
      transform: translateY(-2px) scale(1.05);
      box-shadow: var(--shadow-lg-qa);
      background: var(--text-accent-qa);
      color: white;
    }

    .qa-theme-icon {
      transition: all 0.3s ease;
      position: absolute;
    }
    .qa-theme-toggle .fa-sun { opacity: 0; transform: rotate(-90deg) scale(0.5); }
    .qa-theme-toggle .fa-moon { opacity: 1; transform: rotate(0deg) scale(1); }
    [data-theme="dark"] .qa-theme-toggle .fa-moon { opacity: 0; transform: rotate(90deg) scale(0.5); }
    [data-theme="dark"] .qa-theme-toggle .fa-sun { opacity: 1; transform: rotate(0deg) scale(1); }


    .qa-item {
      background: var(--bg-secondary-qa);
      margin-bottom: 1.5rem;
      border-radius: 0.75rem;
      box-shadow: var(--shadow-md-qa);
      border: 1px solid var(--border-color-qa);
      overflow: hidden;
      transition: box-shadow 0.3s ease;
    }
    .qa-item:hover {
        box-shadow: var(--shadow-lg-qa);
    }

    .qa-question-btn {
      display: flex;
      justify-content: space-between;
      align-items: center;
      width: 100%;
      padding: 1.25rem 1.5rem;
      background: var(--bg-tertiary-qa);
      border: none;
      /* border-bottom: 1px solid var(--border-color-qa); Sacado para que no haya doble borde al expandir */
      text-align: left;
      cursor: pointer;
      transition: background-color 0.3s ease;
    }
    .qa-item.expanded .qa-question-btn {
        background-color: var(--bg-secondary-qa);
        border-bottom: 1px solid var(--border-color-qa); /* Añadir borde solo cuando está expandido y antes del panel */
    }


    .qa-question-btn:hover {
      /* Usar una variable CSS con fallback si no está definida */
      background-color: var(--bg-quaternary-qa, #eef2ff);
    }
    [data-theme="dark"] .qa-question-btn:hover {
        background-color: var(--bg-quaternary-qa, #3e4c5f);
    }


    .qa-question-text {
      font-size: 1.1rem;
      font-weight: 600;
      color: var(--text-primary-qa);
      margin-right: 1rem;
    }

    .qa-toggle-icon {
      font-size: 1rem;
      color: var(--text-accent-qa);
      transition: transform 0.3s ease;
      flex-shrink: 0; /* Evitar que el ícono se encoja */
    }

    .qa-item.expanded .qa-toggle-icon {
      transform: rotate(45deg);
    }

    .qa-answer-panel {
      padding: 1.5rem;
      background: var(--bg-secondary-qa);
      display: none;
      /* border-top: 1px solid var(--border-color-qa); El borde ahora está en el botón expandido */
    }

    .qa-answer-panel.visible {
        display: block;
        animation: fadeInAnswer 0.5s ease forwards;
    }

    @keyframes fadeInAnswer {
        from { opacity: 0; transform: translateY(-10px); }
        to { opacity: 1; transform: translateY(0); }
    }

    .qa-answer-panel p {
      margin-bottom: 1em;
      color: var(--text-primary-qa);
    }
    .qa-answer-panel p:last-child {
      margin-bottom: 0;
    }
    .qa-answer-panel ul {
        list-style-position: outside;
        padding-left: 1.5rem;
        margin-bottom: 1em;
    }
    .qa-answer-panel li {
        margin-bottom: 0.5em;
    }
    .qa-answer-panel strong, .qa-answer-panel em {
        color: var(--text-accent-qa);
        font-weight: 600; /* Asegurar que strong sea negrita */
    }
    [data-theme="dark"] .qa-answer-panel strong, [data-theme="dark"] .qa-answer-panel em {
        color: var(--color-blue-qa);
    }


    .fade-in {
      opacity: 0;
      transform: translateY(20px);
      animation: fadeInUp 0.6s ease forwards;
    }

    .qa-item.fade-in {
        animation: fadeInUp 0.6s ease forwards;
        animation-delay: calc(var(--item-index) * 0.1s);
    }

    @keyframes fadeInUp {
      to {
        opacity: 1;
        transform: translateY(0);
      }
    }
     @media (max-width: 768px) {
      .qa-container { padding: 1.5rem 1rem; }
      .qa-main-title { font-size: 1.8rem; }
      .qa-subtitle { font-size: 1rem; }
      .qa-theme-toggle { top: 1rem; right: 1rem; width: 45px; height: 45px; font-size: 1.1rem;}
      .qa-question-text { font-size: 1rem; }
    }

  </style>
</head>
<body class="qa-body" data-theme="light">

  <button class="qa-theme-toggle" id="qaThemeToggleButton" title="Cambiar tema" aria-label="Cambiar tema">
    <i class="fas fa-moon qa-theme-icon"></i>
    <i class="fas fa-sun qa-theme-icon"></i>
  </button>

  <div class="qa-container">
    <header class="qa-header">
      <h1 class="qa-main-title fade-in">Preguntas y Respuestas Fundamentales</h1>
      <p class="qa-subtitle fade-in">Clase 1: Inteligencia Artificial</p>
    </header>

    <div id="qaAccordion">
      <!-- Pregunta 1 -->
      <div class="qa-item fade-in" style="--item-index: 0;">
        <button class="qa-question-btn" aria-expanded="false" aria-controls="answer1">
          <span class="qa-question-text">1. ¿Qué es el Test de Turing y cuál fue su propósito original?</span>
          <i class="fas fa-plus qa-toggle-icon"></i>
        </button>
        <div id="answer1" class="qa-answer-panel">
          <p>El Test de Turing, propuesto por el pionero de la computación <strong>Alan Turing</strong> en su influyente artículo de 1950, <em>“Computing Machinery and Intelligence”</em>, es un experimento mental diseñado para abordar la compleja pregunta: <em>“¿Pueden las máquinas pensar?”</em>. En lugar de definir directamente el "pensamiento", Turing propuso una prueba operacional conocida como el "Juego de Imitación".</p>
          <p>El propósito original del test era establecer un criterio estándar y empírico para evaluar si una máquina podía exhibir un comportamiento inteligente indistinguible del de un ser humano. En su configuración clásica, un interrogador humano se comunica (generalmente a través de texto) con dos entidades ocultas: una máquina y un ser humano. Si el interrogador, después de un período de conversación, no puede determinar con fiabilidad cuál es la máquina y cuál es el humano, se dice que la máquina ha "pasado" el Test de Turing. Es crucial entender que el test <em>no mide conciencia, autoconciencia o comprensión genuina</em>, sino la capacidad de la máquina para <strong>simular una conversación humana</strong> de manera convincente y coherente.</p>
        </div>
      </div>

      <!-- Pregunta 2 -->
      <div class="qa-item fade-in" style="--item-index: 1;">
        <button class="qa-question-btn" aria-expanded="false" aria-controls="answer2">
          <span class="qa-question-text">2. ¿Cuáles son las principales críticas al Test de Turing?</span>
          <i class="fas fa-plus qa-toggle-icon"></i>
        </button>
        <div id="answer2" class="qa-answer-panel">
          <p>A pesar de su influencia histórica, el Test de Turing ha sido objeto de numerosas críticas. Las principales objeciones incluyen:</p>
          <ul>
            <li><strong>Simulación no implica comprensión (Argumento de la Habitación China de Searle):</strong> Filósofos como John Searle argumentan que una máquina podría pasar el test manipulando símbolos según un conjunto de reglas (como un programa) sin entender realmente el significado de la conversación. La máquina simularía la comprensión, pero no la poseería.</li>
            <li><strong>Limitación al lenguaje y antropocentrismo:</strong> El test se centra exclusivamente en la comunicación lingüística escrita, ignorando otras facetas de la inteligencia como la creatividad, la inteligencia emocional, la capacidad de aprendizaje visual, la resolución de problemas espaciales o la interacción física con el entorno. Además, define la inteligencia en términos muy humanos.</li>
            <li><strong>Posibilidad de falsos positivos (engaño):</strong> Un sistema podría ser programado específicamente para engañar al interrogador utilizando trucos, evasivas, o explotando sesgos humanos, sin poseer una inteligencia general subyacente. Programas como ELIZA en sus inicios demostraron esta capacidad de simulación superficial.</li>
            <li><strong>Relevancia práctica decreciente:</strong> A medida que la IA ha avanzado, el foco se ha desplazado hacia la creación de sistemas que resuelven problemas específicos de manera útil, en lugar de simplemente imitar la conversación humana. La capacidad de una IA para realizar tareas complejas (ej. diagnóstico médico, conducción autónoma) se considera a menudo un mejor indicador de su "inteligencia" o utilidad.</li>
            <li><strong>Dificultad de implementación y subjetividad:</strong> La duración de la prueba, la sofisticación de los interrogadores y los criterios exactos para "pasar" el test son difíciles de estandarizar, lo que lleva a resultados variables y subjetivos.</li>
          </ul>
        </div>
      </div>

      <!-- Pregunta 3 -->
      <div class="qa-item fade-in" style="--item-index: 2;">
        <button class="qa-question-btn" aria-expanded="false" aria-controls="answer3">
          <span class="qa-question-text">3. ¿Qué diferencia fundamental existe entre Machine Learning y Deep Learning?</span>
          <i class="fas fa-plus qa-toggle-icon"></i>
        </button>
        <div id="answer3" class="qa-answer-panel">
          <p>La diferencia fundamental entre <em>Machine Learning (ML)</em> y <em>Deep Learning (DL)</em>, que es un subcampo del ML, radica en cómo se procesan y aprenden las características (features) de los datos y en la complejidad de los modelos utilizados:</p>
          <ul>
            <li>
              <strong>Machine Learning (ML):</strong>
              <ul>
                <li><strong>Ingeniería de características manual:</strong> En el ML tradicional, los científicos de datos o expertos en el dominio deben realizar un proceso de ingeniería de características. Esto implica seleccionar, transformar y crear manualmente las variables (features) más relevantes a partir de los datos brutos que el algoritmo utilizará para aprender. La calidad de estas características es crucial para el rendimiento del modelo.</li>
                <li><strong>Modelos más simples o especializados:</strong> Utiliza una variedad de algoritmos como regresión lineal y logística, máquinas de vectores de soporte (SVM), árboles de decisión, Random Forest, k-Nearest Neighbors, Naive Bayes, entre otros. Estos modelos pueden ser muy efectivos pero a menudo funcionan mejor con datos estructurados y cuando las características importantes son bien conocidas.</li>
                <li><strong>Menor necesidad de datos (relativamente):</strong> Algunos algoritmos de ML pueden funcionar bien con cantidades de datos menores en comparación con el Deep Learning.</li>
              </ul>
            </li>
            <li>
              <strong>Deep Learning (DL):</strong>
              <ul>
                <li><strong>Aprendizaje automático de características (representaciones jerárquicas):</strong> La principal distinción del DL es su capacidad para aprender automáticamente las características relevantes directamente de los datos brutos (ej. píxeles de una imagen, texto crudo). Esto se logra a través de arquitecturas de redes neuronales artificiales con múltiples capas (de ahí el término "profundo"). Cada capa aprende representaciones de los datos a diferentes niveles de abstracción, desde características simples en las primeras capas hasta características más complejas y abstractas en las capas más profundas.</li>
                <li><strong>Redes Neuronales Profundas:</strong> Utiliza arquitecturas como Redes Neuronales Convolucionales (CNNs) para visión por computadora, Redes Neuronales Recurrentes (RNNs) y Transformers para datos secuenciales como texto o series temporales, y Redes Neuronales Densas (DNNs) para una variedad de tareas.</li>
                <li><strong>Gran cantidad de datos y cómputo:</strong> Los modelos de DL generalmente requieren grandes volúmenes de datos para entrenar eficazmente y suelen ser computacionalmente intensivos, beneficiándose del uso de GPUs (Unidades de Procesamiento Gráfico).</li>
              </ul>
            </li>
          </ul>
          <p>En resumen, mientras que el ML a menudo depende de la intervención humana para definir las entradas significativas para el modelo, el DL automatiza este proceso de extracción de características, lo que le permite abordar problemas más complejos con datos no estructurados de manera más efectiva, siempre que se disponga de suficientes datos y recursos computacionales.</p>
        </div>
      </div>

      <!-- Pregunta 4 -->
      <div class="qa-item fade-in" style="--item-index: 3;">
        <button class="qa-question-btn" aria-expanded="false" aria-controls="answer4">
          <span class="qa-question-text">4. ¿Qué tipo de tareas puede resolver el Deep Learning de forma efectiva?</span>
          <i class="fas fa-plus qa-toggle-icon"></i>
        </button>
        <div id="answer4" class="qa-answer-panel">
          <p>El Deep Learning ha revolucionado numerosas áreas y ha demostrado ser excepcionalmente efectivo en tareas que involucran el procesamiento de grandes volúmenes de datos, especialmente datos no estructurados, y donde los patrones subyacentes son complejos y difíciles de definir manualmente. Algunas de las tareas más destacadas incluyen:</p>
          <ul>
            <li><strong>Visión por Computadora:</strong>
                <ul>
                    <li><em>Clasificación de imágenes:</em> Identificar el objeto principal en una imagen (ej. gato, perro, coche).</li>
                    <li><em>Detección de objetos:</em> Localizar y clasificar múltiples objetos dentro de una imagen.</li>
                    <li><em>Segmentación de imágenes:</em> Dividir una imagen en regiones que corresponden a diferentes objetos o partes.</li>
                    <li><em>Reconocimiento facial:</em> Identificar o verificar la identidad de una persona a partir de una imagen facial.</li>
                    <li><em>Generación de imágenes:</em> Crear imágenes nuevas y realistas (ej. GANs, modelos de difusión).</li>
                </ul>
            </li>
            <li><strong>Procesamiento del Lenguaje Natural (NLP):</strong>
                <ul>
                    <li><em>Traducción automática:</em> Traducir texto de un idioma a otro (ej. Google Translate).</li>
                    <li><em>Análisis de sentimientos:</em> Determinar la polaridad emocional (positiva, negativa, neutra) de un texto.</li>
                    <li><em>Generación de texto:</em> Crear texto coherente y contextualmente relevante (ej. ChatGPT, modelos de lenguaje grandes).</li>
                    <li><em>Resumen automático:</em> Generar resúmenes concisos de documentos largos.</li>
                    <li><em>Reconocimiento de entidades nombradas (NER):</em> Identificar y clasificar entidades como nombres de personas, organizaciones, lugares en el texto.</li>
                    <li><em>Sistemas de pregunta-respuesta (Question Answering):</em> Responder preguntas basadas en un contexto dado.</li>
                </ul>
            </li>
            <li><strong>Reconocimiento de Voz:</strong>
                <ul>
                    <li><em>Conversión de voz a texto (Speech-to-Text):</em> Transcribir el habla humana en texto escrito (ej. asistentes virtuales como Siri, Alexa).</li>
                    <li><em>Identificación del hablante:</em> Reconocer quién está hablando.</li>
                </ul>
            </li>
            <li><strong>Juegos y Sistemas de Decisión Complejos:</strong>
                <ul>
                    <li>Superar a campeones humanos en juegos complejos como Go (AlphaGo), ajedrez y videojuegos (ej. Dota 2, StarCraft II con AlphaStar).</li>
                    <li>Optimización de estrategias en entornos complejos.</li>
                </ul>
            </li>
            <li><strong>Descubrimiento de Fármacos y Bioinformática:</strong>
                <ul>
                    <li>Predecir propiedades de moléculas, interacciones farmacológicas y ayudar en el diseño de nuevos medicamentos.</li>
                    <li>Análisis de secuencias genómicas.</li>
                </ul>
            </li>
             <li><strong>Vehículos Autónomos:</strong>
                <ul>
                    <li>Percepción del entorno, toma de decisiones y control del vehículo.</li>
                </ul>
            </li>
          </ul>
          <p>Estas tareas se benefician enormemente de la capacidad de las redes neuronales profundas para aprender representaciones jerárquicas de características directamente de los datos, lo que les permite capturar matices y relaciones que serían muy difíciles de codificar manualmente.</p>
        </div>
      </div>

      <!-- Pregunta 5 -->
      <div class="qa-item fade-in" style="--item-index: 4;">
        <button class="qa-question-btn" aria-expanded="false" aria-controls="answer5">
          <span class="qa-question-text">5. ¿Qué es el Procesamiento del Lenguaje Natural (NLP) y por qué es relevante?</span>
          <i class="fas fa-plus qa-toggle-icon"></i>
        </button>
        <div id="answer5" class="qa-answer-panel">
          <p>El <strong>Procesamiento del Lenguaje Natural (NLP)</strong> es un subcampo interdisciplinario de la lingüística, la informática y la inteligencia artificial. Su objetivo principal es dotar a las computadoras de la capacidad de <em>comprender, interpretar, generar y manipular el lenguaje humano</em> (tanto hablado como escrito) de una manera que sea útil y significativa.</p>
          <p>La relevancia del NLP radica en varios aspectos cruciales:</p>
          <ul>
            <li><strong>Mejora la interacción humano-computadora:</strong> Permite a las personas interactuar con la tecnología utilizando su lenguaje natural, en lugar de tener que aprender lenguajes de programación o comandos específicos. Esto hace que la tecnología sea más accesible e intuitiva para una audiencia más amplia (ej. asistentes virtuales, chatbots).</li>
            <li><strong>Automatización de tareas intensivas en lenguaje:</strong> Muchas tareas que tradicionalmente requerían un esfuerzo humano considerable, como la traducción de documentos, el resumen de textos largos, la moderación de contenido o el análisis de grandes volúmenes de opiniones de clientes, pueden ser automatizadas o asistidas significativamente por sistemas de NLP.</li>
            <li><strong>Extracción de conocimiento de datos no estructurados:</strong> La gran mayoría de la información digital del mundo se encuentra en formato de texto no estructurado (correos electrónicos, redes sociales, artículos de noticias, informes médicos, documentos legales). El NLP proporciona las herramientas para analizar esta vasta cantidad de datos, extraer información valiosa, identificar tendencias y obtener insights.</li>
            <li><strong>Avances en otras áreas de la IA:</strong> La comprensión del lenguaje es fundamental para muchas otras aplicaciones de la IA, como los sistemas de recomendación, la robótica con capacidad de diálogo o la inteligencia artificial general.</li>
            <li><strong>Impacto social y económico:</strong> El NLP tiene un impacto transformador en sectores como la atención al cliente (chatbots), la salud (análisis de historiales médicos), las finanzas (análisis de sentimiento del mercado), la educación (tutores virtuales), el periodismo (generación de noticias) y muchos más.</li>
          </ul>
          <p>En esencia, el NLP es clave para cerrar la brecha de comunicación entre los humanos y las máquinas, desbloqueando el potencial de la vasta cantidad de información textual y oral disponible en el mundo.</p>
        </div>
      </div>

      <!-- Pregunta 6 -->
      <div class="qa-item fade-in" style="--item-index: 5;">
        <button class="qa-question-btn" aria-expanded="false" aria-controls="answer6">
          <span class="qa-question-text">6. ¿Qué desafíos presenta el lenguaje humano para los sistemas de NLP?</span>
          <i class="fas fa-plus qa-toggle-icon"></i>
        </button>
        <div id="answer6" class="qa-answer-panel">
          <p>El lenguaje humano, a pesar de ser utilizado sin esfuerzo por las personas, es inherentemente complejo y presenta numerosos desafíos para los sistemas de Procesamiento del Lenguaje Natural (NLP). Algunos de los más significativos son:</p>
          <ul>
            <li><strong>Ambigüedad:</strong> Esta es quizás la dificultad más grande.
                <ul>
                    <li><em>Ambigüedad léxica:</em> Una misma palabra puede tener múltiples significados (ej. "banco" puede ser una entidad financiera o un asiento).</li>
                    <li><em>Ambigüedad sintáctica (o estructural):</em> Una frase puede tener varias interpretaciones gramaticales válidas (ej. "Vi al hombre en el parque con un telescopio" - ¿Quién tiene el telescopio?).</li>
                    <li><em>Ambigüedad semántica:</em> El significado de una frase puede ser incierto incluso si la gramática es clara (ej. "El pollo está listo para comer" - ¿El pollo va a comer o va a ser comido?).</li>
                    <li><em>Ambigüedad pragmática:</em> El significado depende del contexto de la conversación, la intención del hablante o el conocimiento compartido (ej. el sarcasmo o la ironía).</li>
                </ul>
            </li>
            <li><strong>Variabilidad y Creatividad:</strong> Las personas expresan la misma idea de innumerables maneras, utilizando diferente vocabulario, estructura de frases, y estilos. El lenguaje es dinámico y constantemente se crean nuevas palabras, expresiones y usos.</li>
            <li><strong>Dependencia del Contexto:</strong> El significado de las palabras y frases a menudo depende fuertemente del contexto circundante, tanto el contexto lingüístico (palabras anteriores y posteriores) como el contexto situacional o del mundo real.</li>
            <li><strong>Conocimiento del Mundo (Sentido Común):</strong> La comprensión profunda del lenguaje requiere una vasta cantidad de conocimiento del mundo y sentido común que los humanos adquieren a través de la experiencia, algo muy difícil de codificar en las máquinas.</li>
            <li><strong>Emociones, Tono y Sutilezas Culturales:</strong> El lenguaje transmite emociones, intenciones, ironía, sarcasmo, humor y matices culturales que son difíciles de interpretar para los algoritmos sin un modelado sofisticado de estos aspectos.</li>
            <li><strong>Errores, Ruido e Informalidad:</strong> El lenguaje real está lleno de errores gramaticales, tipográficos, jerga, abreviaturas, lenguaje informal y dialectos, lo que dificulta el procesamiento por parte de sistemas entrenados con datos más formales.</li>
            <li><strong>Escasez de Datos para Lenguajes Minoritarios:</strong> Muchos modelos de NLP de vanguardia requieren grandes cantidades de datos para su entrenamiento, los cuales no están disponibles para todos los idiomas, especialmente los menos hablados.</li>
            <li><strong>Evolución del Lenguaje:</strong> El lenguaje cambia con el tiempo, lo que significa que los modelos de NLP deben ser actualizados continuamente para mantenerse relevantes.</li>
          </ul>
          <p>Superar estos desafíos es un área activa de investigación en NLP, impulsada por avances en modelos de aprendizaje profundo, representaciones de lenguaje más ricas y conjuntos de datos más grandes y diversos.</p>
        </div>
      </div>

      <!-- Pregunta 7 -->
      <div class="qa-item fade-in" style="--item-index: 6;">
        <button class="qa-question-btn" aria-expanded="false" aria-controls="answer7">
          <span class="qa-question-text">7. ¿Cómo funciona un modelo de análisis de sentimientos?</span>
          <i class="fas fa-plus qa-toggle-icon"></i>
        </button>
        <div id="answer7" class="qa-answer-panel">
          <p>Un modelo de <strong>análisis de sentimientos</strong> (también conocido como minería de opiniones) es una aplicación del NLP que tiene como objetivo identificar y clasificar la polaridad emocional expresada en un fragmento de texto. Generalmente, esta polaridad se categoriza como <em>positiva, negativa o neutra</em>, aunque modelos más avanzados pueden intentar detectar un espectro más amplio de emociones (ej. alegría, tristeza, enojo, sorpresa).</p>
          <p>El flujo de trabajo típico para construir y utilizar un modelo de análisis de sentimientos supervisado implica varios pasos:</p>
          <ol>
            <li><strong>Recolección y Etiquetado de Datos (Dataset):</strong> Se necesita un conjunto de datos (dataset) donde cada texto (ej. una reseña de producto, un tweet, un comentario) esté etiquetado con el sentimiento correspondiente. Este etiquetado puede ser manual o, en algunos casos, inferido (ej. reseñas con 1-2 estrellas como negativas, 4-5 estrellas como positivas).</li>
            <li><strong>Preprocesamiento del Texto:</strong> Antes de que el texto pueda ser procesado por un algoritmo, necesita ser limpiado y normalizado. Esto puede incluir:
                <ul>
                    <li>Conversión a minúsculas.</li>
                    <li>Eliminación de signos de puntuación, caracteres especiales o URLs.</li>
                    <li><em>Tokenización:</em> División del texto en unidades más pequeñas (tokens), generalmente palabras.</li>
                    <li>Eliminación de <em>stopwords</em> (palabras comunes como "el", "la", "es" que aportan poco significado para esta tarea).</li>
                    <li><em>Lematización o Stemming:</em> Reducción de las palabras a su forma raíz o base (ej. "corriendo" -> "correr").</li>
                </ul>
            </li>
            <li><strong>Extracción de Características y Vectorización:</strong> Las máquinas no entienden texto directamente, por lo que debe convertirse en una representación numérica (vectores). Algunas técnicas comunes son:
                <ul>
                    <li><em>Bag of Words (BoW):</em> Representa el texto como una colección de sus palabras, ignorando la gramática y el orden, pero manteniendo la multiplicidad.
                        <ul>
                            <li><strong>CountVectorizer:</strong> Cuenta la frecuencia de cada palabra en el texto.</li>
                            <li><strong>TF-IDF (Term Frequency-Inverse Document Frequency):</strong> Pondera las palabras no solo por su frecuencia en un documento, sino también por cuán raras son en todo el corpus de documentos, dando más importancia a las palabras distintivas.</li>
                        </ul>
                    </li>
                    <li><em>Word Embeddings (ej. Word2Vec, GloVe, FastText):</em> Representaciones vectoriales densas que capturan relaciones semánticas entre palabras. Son más comunes en modelos de deep learning.</li>
                </ul>
            </li>
            <li><strong>División del Dataset:</strong> El dataset etiquetado se divide típicamente en un conjunto de entrenamiento (para entrenar el modelo) y un conjunto de prueba (para evaluar su rendimiento en datos no vistos).</li>
            <li><strong>Entrenamiento del Modelo Clasificador:</strong> Se elige un algoritmo de aprendizaje supervisado y se entrena con el conjunto de entrenamiento. Algoritmos comunes incluyen:
                <ul>
                    <li><em>Naive Bayes (especialmente MultinomialNB para texto).</em></li>
                    <li>Regresión Logística.</li>
                    <li>Máquinas de Vectores de Soporte (SVM).</li>
                    <li>Redes Neuronales (incluyendo RNNs, LSTMs, Transformers para enfoques de Deep Learning).</li>
                </ul>
                El modelo aprende a asociar los patrones en los vectores de características con las etiquetas de sentimiento.
            </li>
            <li><strong>Evaluación del Modelo:</strong> Se utiliza el conjunto de prueba para evaluar qué tan bien generaliza el modelo a datos nuevos. Las métricas comunes incluyen:
                <ul>
                    <li><em>Precisión (Accuracy):</em> Proporción de predicciones correctas.</li>
                    <li><em>Precision (Precisión por clase):</em> De todas las predicciones positivas, ¿cuántas fueron realmente positivas?</li>
                    <li><em>Recall (Sensibilidad):</em> De todos los ejemplos realmente positivos, ¿cuántos identificó correctamente el modelo?</li>
                    <li><em>F1-Score:</em> Media armónica de Precision y Recall.</li>
                    <li><em>Matriz de Confusión:</em> Muestra el número de verdaderos positivos, verdaderos negativos, falsos positivos y falsos negativos.</li>
                </ul>
            </li>
            <li><strong>Aplicación (Inferencia):</strong> Una vez que el modelo se considera satisfactorio, se puede utilizar para predecir el sentimiento de nuevos textos no etiquetados.</li>
          </ol>
        </div>
      </div>

      <!-- Pregunta 8 -->
      <div class="qa-item fade-in" style="--item-index: 7;">
        <button class="qa-question-btn" aria-expanded="false" aria-controls="answer8">
          <span class="qa-question-text">8. ¿Qué es Naive Bayes y por qué se utiliza en NLP?</span>
          <i class="fas fa-plus qa-toggle-icon"></i>
        </button>
        <div id="answer8" class="qa-answer-panel">
          <p><strong>Naive Bayes</strong> es una familia de algoritmos de clasificación probabilísticos simples basados en la aplicación del <strong>Teorema de Bayes</strong> con una fuerte (de ahí "naive" o ingenua) suposición de <em>independencia condicional entre las características</em>, dado el valor de la clase.</p>
          <p>En términos más sencillos, para un problema de clasificación, Naive Bayes calcula la probabilidad de que una instancia pertenezca a una clase particular, basándose en la presencia de ciertas características. La suposición "ingenua" es que la presencia (o ausencia) de una característica particular es independiente de la presencia (o ausencia) de cualquier otra característica, una vez que conocemos la clase.</p>
          <p>El Teorema de Bayes se expresa como:</p>
          <p><code>P(clase | características) = [P(características | clase) * P(clase)] / P(características)</code></p>
          <ul>
            <li><code>P(clase | características)</code> es la probabilidad posterior de la clase dada las características (lo que queremos predecir).</li>
            <li><code>P(características | clase)</code> es la verosimilitud (likelihood) de las características dada la clase.</li>
            <li><code>P(clase)</code> es la probabilidad previa (prior) de la clase.</li>
            <li><code>P(características)</code> es la probabilidad previa de las características (evidencia).</li>
          </ul>
          <p>Para la clasificación, se calcula esta probabilidad para cada clase y se elige la clase con la probabilidad posterior más alta.</p>
          <p><strong>¿Por qué se utiliza Naive Bayes en NLP?</strong></p>
          <p>A pesar de su suposición de independencia simplista (que rara vez es cierta en el lenguaje real, ya que las palabras suelen tener dependencias), Naive Bayes es sorprendentemente efectivo y popular en muchas tareas de NLP, especialmente en la clasificación de texto, por varias razones:</p>
          <ul>
            <li><strong>Simplicidad y Eficiencia Computacional:</strong> Es fácil de implementar y computacionalmente muy rápido tanto para el entrenamiento como para la predicción. Esto lo hace ideal para conjuntos de datos grandes o aplicaciones en tiempo real.</li>
            <li><strong>Buen Rendimiento como Baseline:</strong> A menudo proporciona un buen rendimiento base (baseline) con el que se pueden comparar modelos más complejos. En muchos casos, su rendimiento es competitivo con algoritmos más sofisticados, especialmente cuando el número de características es alto (como en el caso de vocabularios grandes en texto).</li>
            <li><strong>Manejo de Alta Dimensionalidad:</strong> Funciona bien con datos de alta dimensionalidad, como los que se encuentran en la representación de texto usando Bag of Words, donde cada palabra del vocabulario puede ser una característica.</li>
            <li><strong>Requerimientos de Datos Relativamente Bajos:</strong> Puede funcionar razonablemente bien incluso con cantidades de datos de entrenamiento más pequeñas en comparación con modelos más complejos como las redes neuronales profundas.</li>
            <li><strong>Robustez a Características Irrelevantes:</strong> Aunque no las ignora, la presencia de muchas características irrelevantes no suele degradar drásticamente su rendimiento.</li>
            <li><strong>Variantes Específicas para Texto:</strong>
                <ul>
                    <li><em>Multinomial Naive Bayes:</em> Adecuado para datos que representan conteos (como la frecuencia de palabras en un documento, típico de CountVectorizer).</li>
                    <li><em>Bernoulli Naive Bayes:</em> Adecuado para características binarias (presencia o ausencia de una palabra).</li>
                </ul>
            </li>
          </ul>
          <p>Se utiliza comúnmente para tareas como:</p>
          <ul>
            <li>Clasificación de spam en correos electrónicos.</li>
            <li>Análisis de sentimientos.</li>
            <li>Clasificación de documentos por tema.</li>
          </ul>
          <p>Aunque la suposición de independencia es una simplificación, en la práctica, Naive Bayes a menudo funciona bien porque la estimación de las probabilidades no necesita ser perfectamente precisa para tomar la decisión de clasificación correcta; solo necesita que la clase con la mayor probabilidad sea la correcta.</p>
        </div>
      </div>

      <!-- Pregunta 9 -->
      <div class="qa-item fade-in" style="--item-index: 8;">
        <button class="qa-question-btn" aria-expanded="false" aria-controls="answer9">
          <span class="qa-question-text">9. ¿Qué son los embeddings de palabras y qué papel cumplen?</span>
          <i class="fas fa-plus qa-toggle-icon"></i>
        </button>
        <div id="answer9" class="qa-answer-panel">
          <p>Los <strong>embeddings de palabras (word embeddings)</strong> son una forma de representación del vocabulario donde las palabras o frases se mapean a <em>vectores de números reales</em> en un espacio de baja dimensión (comparado con la alta dimensionalidad de técnicas como one-hot encoding o Bag of Words). Estos vectores son densos (la mayoría de sus componentes no son cero) y están diseñados para capturar el <strong>significado semántico y las relaciones sintácticas</strong> entre las palabras.</p>
          <p><strong>Características Clave de los Word Embeddings:</strong></p>
          <ul>
            <li><strong>Representación Densa:</strong> A diferencia de las representaciones dispersas (como one-hot encoding, donde cada palabra es un vector largo con un solo '1' y el resto ceros), los embeddings son vectores densos de una dimensionalidad mucho menor (ej. 50, 100, 300 dimensiones).</li>
            <li><strong>Captura de Semántica:</strong> Palabras con significados similares tienden a tener vectores de embedding cercanos en el espacio vectorial. Por ejemplo, el vector de "gato" estaría más cerca del vector de "perro" que del vector de "coche".</li>
            <li><strong>Relaciones Lineales:</strong> A menudo, se pueden capturar relaciones analógicas mediante operaciones aritméticas simples con los vectores. El ejemplo clásico es <em>vector("rey") - vector("hombre") + vector("mujer") ≈ vector("reina")</em>.</li>
            <li><strong>Aprendidos de los Datos:</strong> Los embeddings no se definen manualmente, sino que se aprenden a partir de grandes corpus de texto utilizando algoritmos de aprendizaje no supervisado. Estos algoritmos analizan los contextos en los que aparecen las palabras.</li>
          </ul>
          <p><strong>Papel que Cumplen en NLP:</strong></p>
          <p>Los embeddings de palabras han sido fundamentales para el avance de muchas tareas de NLP y cumplen varios roles importantes:</p>
          <ol>
            <li><strong>Mejora del Rendimiento de los Modelos:</strong> Al proporcionar a los modelos de machine learning y deep learning representaciones de palabras más ricas y significativas, los embeddings suelen conducir a un mejor rendimiento en tareas como clasificación de texto, análisis de sentimientos, traducción automática, reconocimiento de entidades nombradas, etc.</li>
            <li><strong>Generalización:</strong> Como las palabras similares tienen embeddings similares, los modelos pueden generalizar mejor a partir de los datos de entrenamiento. Si un modelo ha aprendido algo sobre la palabra "rápido", puede inferir propiedades similares para la palabra "veloz" si sus embeddings son cercanos.</li>
            <li><strong>Reducción de la Dimensionalidad:</strong> Convierten vocabularios muy grandes (alta dimensionalidad) en espacios vectoriales de menor dimensión, lo que hace que los modelos sean más eficientes y menos propensos al sobreajuste (curse of dimensionality).</li>
            <li><strong>Entrada para Redes Neuronales:</strong> Son la forma estándar de alimentar texto a las redes neuronales profundas. La primera capa de muchos modelos de NLP basados en deep learning es una capa de embedding que convierte los índices de palabras (o tokens) en sus correspondientes vectores de embedding.</li>
            <li><strong>Transferencia de Aprendizaje (Embeddings Preentrenados):</strong> Se pueden utilizar embeddings preentrenados en grandes corpus de texto (como Wikipedia o Google News) y luego aplicarlos a tareas específicas donde los datos etiquetados pueden ser escasos. Modelos como <em>Word2Vec, GloVe y FastText</em> proporcionan estos embeddings preentrenados. Modelos de lenguaje más grandes como BERT, GPT, etc., también generan embeddings contextuales muy potentes.</li>
            <li><strong>Comprensión Contextual (en modelos más avanzados):</strong> Mientras que los embeddings tradicionales asignan un vector fijo a cada palabra, los modelos más modernos (como ELMo, BERT, GPT) generan <em>embeddings contextuales</em>, donde la representación vectorial de una palabra depende del contexto específico en el que aparece en la oración.</li>
          </ol>
          <p>En resumen, los embeddings de palabras transforman el texto en un formato numérico que no solo es manejable por los algoritmos, sino que también encapsula el significado y las relaciones inherentes al lenguaje, lo que permite a los modelos de IA "entender" y procesar el lenguaje de manera más efectiva.</p>
        </div>
      </div>

      <!-- Pregunta 10 -->
      <div class="qa-item fade-in" style="--item-index: 9;">
        <button class="qa-question-btn" aria-expanded="false" aria-controls="answer10">
          <span class="qa-question-text">10. ¿Qué herramientas se usaron en el módulo para implementar el análisis de sentimientos?</span>
          <i class="fas fa-plus qa-toggle-icon"></i>
        </button>
        <div id="answer10" class="qa-answer-panel">
          <p>Para la implementación práctica del análisis de sentimientos en el módulo, se utilizó un conjunto de herramientas populares dentro del ecosistema de <strong>Python</strong> para ciencia de datos y machine learning. Estas herramientas facilitan cada etapa del proceso, desde la manipulación de datos hasta el entrenamiento y la evaluación del modelo:</p>
          <ul>
            <li>
              <strong>Python:</strong> Como lenguaje de programación principal, debido a su sintaxis clara, amplia adopción en la comunidad científica y la vasta cantidad de bibliotecas disponibles.
            </li>
            <li>
              <strong>Pandas:</strong>
              <ul>
                <li><strong>Propósito:</strong> Utilizada para la creación, manipulación y análisis de datos estructurados.</li>
                <li><strong>Uso en el módulo:</strong> Específicamente, se empleó para crear un <em>DataFrame</em> (una estructura de datos tabular similar a una hoja de cálculo) que contenía las opiniones (textos) y sus correspondientes etiquetas de sentimiento (ej. "positivo", "negativo"). Esto facilitó la organización y el acceso a los datos.</li>
              </ul>
            </li>
            <li>
              <strong>Scikit-learn (sklearn):</strong> Una biblioteca integral de machine learning en Python. Varias de sus funcionalidades fueron cruciales:
              <ul>
                <li>
                  <strong><code>CountVectorizer</code> (del módulo <code>sklearn.feature_extraction.text</code>):</strong>
                  <ul>
                    <li><strong>Propósito:</strong> Convertir una colección de documentos de texto en una matriz de conteos de tokens (palabras). Es una forma de implementar la técnica de Bag of Words.</li>
                    <li><strong>Uso en el módulo:</strong> Transformó las opiniones textuales en vectores numéricos que el modelo de machine learning puede procesar. Cada columna representa una palabra del vocabulario y cada fila un documento, con el valor siendo el conteo de esa palabra en ese documento.</li>
                  </ul>
                </li>
                <li>
                  <strong><code>MultinomialNB</code> (del módulo <code>sklearn.naive_bayes</code>):</strong>
                  <ul>
                    <li><strong>Propósito:</strong> Implementación del algoritmo clasificador Naive Bayes para datos distribuidos multinomialmente (adecuado para conteos de palabras).</li>
                    <li><strong>Uso en el módulo:</strong> Fue el algoritmo elegido para entrenar el modelo de clasificación de sentimientos. Aprendió a partir de los vectores de palabras y sus etiquetas de sentimiento asociadas.</li>
                  </ul>
                </li>
                <li>
                  <strong><code>train_test_split</code> (del módulo <code>sklearn.model_selection</code>):</strong>
                  <ul>
                    <li><strong>Propósito:</strong> Dividir arrays o matrices en subconjuntos aleatorios de entrenamiento y prueba.</li>
                    <li><strong>Uso en el módulo:</strong> Se utilizó para separar el dataset de opiniones en un conjunto de entrenamiento (con el que el modelo aprende) y un conjunto de prueba (para evaluar el rendimiento del modelo en datos no vistos, evitando el sobreajuste).</li>
                  </ul>
                </li>
                <li>
                  <strong><code>classification_report</code> (del módulo <code>sklearn.metrics</code>):</strong>
                  <ul>
                    <li><strong>Propósito:</strong> Construir un informe de texto que muestra las principales métricas de clasificación (precision, recall, F1-score, support) para cada clase.</li>
                    <li><strong>Uso en el módulo:</strong> Permitió evaluar de manera detallada el rendimiento del modelo Naive Bayes en la tarea de clasificación de sentimientos, mostrando qué tan bien predecía las etiquetas positivas y negativas.</li>
                  </ul>
                </li>
                 <li>
                  (Opcionalmente, aunque no mencionado explícitamente en la pregunta pero común en estos flujos) <strong><code>accuracy_score</code>, <code>confusion_matrix</code> (del módulo <code>sklearn.metrics</code>):</strong> Para obtener la precisión general y una matriz de confusión detallada.
                </li>
              </ul>
            </li>
            <li>
              <strong>Matplotlib (específicamente <code>matplotlib.pyplot</code>):</strong>
              <ul>
                <li><strong>Propósito:</strong> Una biblioteca de trazado 2D que produce figuras de calidad de publicación en una variedad de formatos impresos e interactivos.</li>
                <li><strong>Uso en el módulo (potencial):</strong> Aunque la pregunta original menciona "visualización de la distribución de predicciones", Matplotlib es la herramienta estándar para esto. Se podría haber usado para crear histogramas de las predicciones del modelo, gráficos de barras de la distribución de sentimientos, o visualizar la matriz de confusión.</li>
              </ul>
            </li>
          </ul>
          <p>La combinación de estas herramientas proporciona un flujo de trabajo robusto y eficiente para abordar problemas de NLP supervisado como el análisis de sentimientos, desde la preparación de los datos hasta la evaluación del modelo final.</p>
        </div>
      </div>

    </div> <!-- Fin de qaAccordion -->
  </div> <!-- Fin de qa-container -->

  <script>
    (function() {
        let qaCurrentSessionTheme = 'light';

        function qaApplyInitialTheme() {
            document.body.setAttribute('data-theme', qaCurrentSessionTheme);
        }

        function qaToggleTheme() {
            const currentThemeFromBody = document.body.getAttribute('data-theme');
            const newTheme = currentThemeFromBody === 'dark' ? 'light' : 'dark';
            document.body.setAttribute('data-theme', newTheme);
            qaCurrentSessionTheme = newTheme;
        }

        const qaThemeToggleButton = document.getElementById('qaThemeToggleButton');
        if (qaThemeToggleButton) {
            qaThemeToggleButton.addEventListener('click', qaToggleTheme);
        }

        const qaAccordion = document.getElementById('qaAccordion');
        if (qaAccordion) {
            const qaItems = qaAccordion.querySelectorAll('.qa-item');
            qaItems.forEach((item, index) => {
                item.style.setProperty('--item-index', index);
                const btn = item.querySelector('.qa-question-btn');
                const panel = item.querySelector('.qa-answer-panel');
                const icon = btn.querySelector('.qa-toggle-icon');

                if (btn && panel && icon) {
                    btn.addEventListener('click', () => {
                        const isExpanded = btn.getAttribute('aria-expanded') === 'true';

                        // Cierra todos los demás paneles antes de abrir el actual (opcional)
                        /*
                        if (!isExpanded) { // Solo si se va a expandir este
                            qaItems.forEach(otherItem => {
                                if (otherItem !== item) {
                                    otherItem.querySelector('.qa-question-btn').setAttribute('aria-expanded', 'false');
                                    otherItem.querySelector('.qa-answer-panel').classList.remove('visible');
                                    otherItem.classList.remove('expanded');
                                    const otherIcon = otherItem.querySelector('.qa-toggle-icon');
                                    otherIcon.classList.remove('fa-minus');
                                    otherIcon.classList.add('fa-plus');
                                }
                            });
                        }
                        */

                        btn.setAttribute('aria-expanded', !isExpanded);
                        panel.classList.toggle('visible');
                        item.classList.toggle('expanded');

                        if (!isExpanded) {
                            icon.classList.remove('fa-plus');
                            icon.classList.add('fa-minus');
                        } else {
                            icon.classList.remove('fa-minus');
                            icon.classList.add('fa-plus');
                        }
                    });
                }
            });
        }

        qaApplyInitialTheme();
    })();
  </script>
</body>
</html>
"""

# Mostrar el HTML en la salida de la celda
display(HTML(html_content_qa))

```
