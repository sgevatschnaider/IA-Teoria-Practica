```python
from IPython.display import display, HTML

# ==============================================================================
# 1. PLANTILLA HTML PARA CONJUNTOS (SETS) CON FOOTER AÑADIDO
# ==============================================================================
html_template_set = """
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <title>Guía Interactiva de Conjuntos (Sets) en Python</title>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">
  <style>
    :root {{
      --bg-primary: #1a1a2e; --bg-secondary: rgba(27, 40, 69, 0.85); --bg-tertiary: rgba(45, 55, 72, 0.7);
      --text-primary: #e0e0e0; --text-secondary: #a0aec0; --accent-primary: #4eacfa; --accent-secondary: #c471ed;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(255, 255, 255, 0.15); --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.2);
      --border-radius: 15px; --transition: all 0.4s cubic-bezier(0.25, 0.8, 0.25, 1);
    }}
    * {{ margin: 0; padding: 0; box-sizing: border-box; }}
    html {{ scroll-behavior: smooth; }}
    body {{ font-family: 'Inter', sans-serif; line-height: 1.8; background-color: var(--bg-primary); color: var(--text-primary); }}
    .container {{ max-width: 1000px; margin: 0 auto; padding: 2rem; }}
    .header {{ text-align: center; margin-bottom: 3rem; }}
    .main-title {{ font-size: 3.5rem; font-weight: 800; background: var(--accent-gradient); -webkit-background-clip: text; color: transparent; background-clip: text; margin-bottom: 1rem; }}
    .subtitle {{ font-size: 1.3rem; color: var(--text-secondary); font-weight: 400; max-width: 800px; margin: auto; }}
    .lesson-container {{ display: flex; flex-direction: column; gap: 1.5rem; }}
    .topic-card {{ background: var(--bg-secondary); backdrop-filter: blur(10px); border-radius: var(--border-radius); box-shadow: var(--shadow-card); border: 1px solid var(--border-color); overflow: hidden; transition: var(--transition); }}
    .topic-header {{ cursor: pointer; padding: 1.5rem 2rem; display: flex; justify-content: space-between; align-items: center; }}
    .topic-title {{ font-size: 1.3rem; font-weight: 600; color: var(--text-primary); }}
    .expand-icon {{ font-size: 1.2rem; color: var(--text-secondary); transition: var(--transition); }}
    .topic-card.open .expand-icon {{ transform: rotate(180deg); }}
    .topic-content {{ max-height: 0; overflow: hidden; transition: max-height 1s ease-in-out, padding 1s ease-in-out; background: var(--bg-tertiary); }}
    .topic-card.open .topic-content {{ max-height: 5000px; padding: 1.5rem 2rem; border-top: 1px solid var(--border-color); }}
    .topic-content h3 {{ font-size: 1.2rem; color: var(--accent-primary); margin: 1.5rem 0 1rem 0; padding-bottom: 0.5rem; border-bottom: 2px solid var(--accent-secondary); }}
    .topic-content h3:first-child {{ margin-top: 0; }}
    .topic-content p, .topic-content li {{ margin-bottom: 1rem; color: var(--text-secondary); line-height: 1.7; }}
    .topic-content ul {{ padding-left: 20px; }}
    .topic-content strong {{ color: #ffffff; font-weight: 600; }}
    pre, .diagram {{ background: #0f172a; padding: 1.2rem; border-radius: 10px; border: 1px solid var(--border-color); overflow-x: auto; margin: 1.5rem 0; font-family: 'JetBrains Mono', monospace; font-size: 0.9em; color: #cbd5e1; white-space: pre-wrap; word-wrap: break-word; }}
    .code-output {{ background: #1e293b; color: #e2e8f0; font-family: 'JetBrains Mono', monospace; padding: 1rem; border-radius: 0 0 10px 10px; margin-top: -1.5rem; border: 1px solid var(--border-color); border-top:none; white-space: pre; }}
    .tip-box {{ background: rgba(78, 172, 250, 0.1); border-left: 4px solid var(--accent-primary); border-radius: 8px; padding: 1rem; margin: 1.5rem 0; }}
    .tip-box::before {{ content: '💡 Pythonic Tip: '; font-weight: bold; color: var(--accent-primary);}}

    /* === ESTILOS PARA EL FOOTER AÑADIDO === */
    footer {{
      text-align: center;
      margin-top: 3rem;
      padding-top: 1.5rem;
      border-top: 1px solid var(--border-color);
    }}
    footer p {{
      color: var(--text-secondary);
      font-size: 0.9rem;
    }}
  </style>
</head>
<body>
  <div class="container">
    <header class="header">
      <h1 class="main-title">Guía Interactiva de Conjuntos (Sets) en Python</h1>
      <p class="subtitle">Descubre el poder de los conjuntos para gestionar colecciones de elementos únicos y realizar operaciones matemáticas de forma ultra-rápida.</p>
    </header>
    <div class="lesson-container">
        <!-- Parte 1 -->
        <div class="topic-card open">
            <div class="topic-header"><span class="topic-title">Parte 1: ¿Qué es un Conjunto y Cómo se Crea?</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <p>Un conjunto (set) es una colección con dos características clave que la diferencian de listas y tuplas:</p>
                <ul><li><strong>Sin Orden:</strong> Los elementos no mantienen un orden de inserción. No puedes acceder a ellos por un índice.</li><li><strong>Elementos Únicos:</strong> No puede haber elementos duplicados. Si intentas añadir un elemento que ya existe, simplemente se ignora.</li><li><strong>Mutable:</strong> Puedes añadir y eliminar elementos después de su creación.</li></ul>
                <h3>Formas de Crear un Conjunto</h3>
                <p>La forma más común es con el constructor <code>set()</code> a partir de un iterable (como una lista), lo cual es ideal para eliminar duplicados.</p>
                <div class="tip-box">¡Cuidado! Para crear un conjunto vacío, <strong>debes usar <code>set()</code></strong>. Usar <code>{{}}</code> crea un diccionario vacío, no un conjunto.</div>
                <pre># Creación a partir de una lista (elimina duplicados automáticamente)\nlista_con_duplicados = [1, 2, "a", 3, 2, "a"]\nmi_set = set(lista_con_duplicados)\n\n# Creación con literales (si no está vacío)\nset_con_literales = {{'rojo', 'verde', 'azul'}}\n\n# Creación de un conjunto vacío (la única forma correcta)\nset_vacio = set()</pre>
                <div class="code-output">{output_set_p1_creacion}</div>
            </div>
        </div>
        <!-- Parte 2 -->
        <div class="topic-card">
            <div class="topic-header"><span class="topic-title">Parte 2: Añadir y Eliminar Elementos</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <p>Al no tener orden, no se accede a los elementos, sino que se gestiona su pertenencia al conjunto.</p>
                <h3>Añadir Elementos</h3>
                <ul><li><code>.add(elemento)</code>: Añade un único elemento. Si ya existe, no hace nada.</li><li><code>.update(otro_iterable)</code>: Añade todos los elementos de otro iterable (lista, tupla, otro set).</li></ul>
                <pre>tags = {{"python", "dev"}}\n# Añadir un elemento\ntags.add("web")\n\n# Añadir múltiples elementos\ntags.update(["database", "api"])</pre>
                <div class="code-output">{output_set_p2_add}</div>
                <h3>Eliminar Elementos: <code>.remove()</code> vs <code>.discard()</code></h3>
                <p>Esta es una distinción importante:</p>
                <ul><li><code>.remove(elemento)</code>: Elimina el elemento. Si <strong>no existe</strong>, lanza un error <code>KeyError</code>.</li><li><code>.discard(elemento)</code>: Elimina el elemento. Si <strong>no existe</strong>, no hace nada y no lanza error.</li></ul>
                <pre># Partiendo de: {{"python", "dev", "web", "database", "api"}}\n# Eliminar un elemento existente\ntags.remove("api")\n\n# Descartar un elemento que NO existe (sin error)\ntags.discard("mobile")</pre>
                <div class="code-output">{output_set_p2_remove}</div>
            </div>
        </div>
        <!-- Parte 3 -->
        <div class="topic-card">
            <div class="topic-header"><span class="topic-title">Parte 3: El Poder de los Sets: Operaciones Matemáticas</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <p>La verdadera magia de los conjuntos reside en su capacidad para realizar operaciones de la teoría de conjuntos de forma muy eficiente.</p>
                <pre>set_a = {{1, 2, 3, 4}}\nset_b = {{3, 4, 5, 6}}</pre>
                <h3>Unión ( | ): Todos los elementos de ambos conjuntos.</h3>
                <pre>union = set_a | set_b</pre>
                <div class="code-output">{output_set_p3_union}</div>
                <h3>Intersección ( & ): Elementos que están en AMBOS conjuntos.</h3>
                <pre>interseccion = set_a & set_b</pre>
                <div class="code-output">{output_set_p3_intersection}</div>
                <h3>Diferencia ( - ): Elementos en A pero NO en B.</h3>
                <pre>diferencia = set_a - set_b</pre>
                <div class="code-output">{output_set_p3_difference}</div>
                <h3>Diferencia Simétrica ( ^ ): Elementos en A o en B, pero NO en ambos.</h3>
                <pre>dif_simetrica = set_a ^ set_b</pre>
                <div class="code-output">{output_set_p3_symmetric_difference}</div>
            </div>
        </div>
        <!-- Parte 4 -->
        <div class="topic-card">
            <div class="topic-header"><span class="topic-title">Parte 4: Iteración y Pertenencia</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <h3>Iterando sobre un Conjunto</h3>
                <p>Puedes usar un bucle <code>for</code> para recorrer los elementos de un conjunto. Recuerda que el orden no está garantizado.</p>
                <pre>permisos = {{"lectura", "escritura", "ejecucion"}}\nfor p in permisos:\n    print(f"Permiso: {{p}}")</pre>
                <div class="code-output">{output_set_p4_iteracion}</div>
                <h3>Comprobación de Pertenencia (<code>in</code>)</h3>
                <div class="tip-box">Una de las mayores ventajas de los conjuntos es que comprobar si un elemento está dentro (<code>elemento in mi_set</code>) es extremadamente rápido (operación de tiempo constante, O(1)), mucho más que en una lista (O(n)).</div>
                <pre># Partiendo de: {{"lectura", "escritura", "ejecucion"}}\n'lectura' in permisos\n'admin' in permisos</pre>
                <div class="code-output">{output_set_p4_membership}</div>
            </div>
        </div>
        <!-- Parte 5 -->
        <div class="topic-card">
            <div class="topic-header"><span class="topic-title">Parte 5: Técnicas Avanzadas y Set Comprehensions</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <h3>Set Comprehensions</h3>
                <p>Al igual que listas y diccionarios, los conjuntos tienen su propia sintaxis de "comprensión" para crearlos de forma concisa. La estructura es <code>{{expresion for item in iterable}}</code>.</p>
                <pre># Crear un conjunto con los cuadrados de los números pares del 0 al 9\ncuadrados_pares = {{x**2 for x in range(10) if x % 2 == 0}}</pre>
                <div class="code-output">{output_set_p5_comprehension}</div>
                <h3>Comparación de Conjuntos</h3>
                <p>Puedes comprobar relaciones entre conjuntos, como si uno es subconjunto de otro.</p>
                <ul><li><code>A.issubset(B)</code> o <code>A <= B</code>: ¿Son todos los elementos de A también de B?</li><li><code>A.issuperset(B)</code> o <code>A >= B</code>: ¿Contiene A todos los elementos de B?</li><li><code>A.isdisjoint(B)</code>: ¿No tienen elementos en común A y B?</li></ul>
                <pre>a = {{1, 2}}\nb = {{1, 2, 3}}\nc = {{4, 5}}</pre>
                <div class="code-output">{output_set_p5_comparison}</div>
            </div>
        </div>
    </div>
    <!-- === FOOTER AÑADIDO === -->
    <footer>
      <p>Material elaborado por prof. Sergio Gevatschnaider</p>
    </footer>
  </div>
  <script>
    (function() {{
        document.querySelectorAll('.topic-header').forEach(header => {{
            header.addEventListener('click', () => {{
                document.querySelectorAll('.topic-card.open').forEach(openCard => {{
                    if (openCard !== header.parentElement) {{
                        openCard.classList.remove('open');
                    }}
                }});
                header.parentElement.classList.toggle('open');
            }});
        }});
    }})();
  </script>
</body>
</html>
"""

# ==============================================================================
# 2. LÓGICA PYTHON PARA GENERAR LOS RESULTADOS DE CONJUNTOS (SETS)
# ==============================================================================

# Parte 1: Creación
lista_con_duplicados = [1, 2, "a", 3, 2, "a"]
mi_set = set(lista_con_duplicados)
set_con_literales = {'rojo', 'verde', 'azul'}
set_vacio = set()
output_set_p1_creacion = (f"Lista original: {lista_con_duplicados}\n"
                          f"Set creado desde lista (duplicados eliminados): {mi_set}\n"
                          f"Set con literales: {set_con_literales}\n"
                          f"Set vacío con set(): {set_vacio} (tipo: {type(set_vacio)})")

# Parte 2: Añadir y Eliminar
tags = {"python", "dev"}
tags_antes = tags.copy()
tags.add("web")
tags.update(["database", "api"])
output_set_p2_add = (f"Set original: {tags_antes}\n"
                     f"Tras .add('web') y .update(['database', 'api']): {tags}")

tags_remove = tags.copy()
tags_remove.remove("api")
tags_remove.discard("mobile")
output_set_p2_remove = (f"Set original: {tags}\n"
                        f"Tras .remove('api'): {tags_remove}\n"
                        f".discard('mobile') no hace nada ni da error.")

# Parte 3: Operaciones Matemáticas
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
output_set_p3_union = f"A | B = {set_a | set_b}"
output_set_p3_intersection = f"A & B = {set_a & set_b}"
output_set_p3_difference = f"A - B = {set_a - set_b}"
output_set_p3_symmetric_difference = f"A ^ B = {set_a ^ set_b}"

# Parte 4: Iteración y Pertenencia
permisos = {"lectura", "escritura", "ejecucion"}
iteracion_res = [f"Permiso: {p}" for p in permisos]
output_set_p4_iteracion = "\n".join(iteracion_res)

output_set_p4_membership = (f"¿'lectura' in permisos? -> {'lectura' in permisos}\n"
                            f"¿'admin' in permisos? -> {'admin' in permisos}")

# Parte 5: Técnicas Avanzadas
cuadrados_pares = {x**2 for x in range(10) if x % 2 == 0}
output_set_p5_comprehension = f"Conjunto resultante: {cuadrados_pares}"

a = {1, 2}
b = {1, 2, 3}
c = {4, 5}
output_set_p5_comparison = (f"a.issubset(b)? -> {a.issubset(b)} (o a <= b)\n"
                            f"b.issuperset(a)? -> {b.issuperset(a)} (o b >= a)\n"
                            f"a.isdisjoint(c)? -> {a.isdisjoint(c)}")

# ==============================================================================
# 3. INYECTAR LOS RESULTADOS EN LA PLANTILLA DE CONJUNTOS
# ==============================================================================
final_html_set = html_template_set.format(
    output_set_p1_creacion=output_set_p1_creacion,
    output_set_p2_add=output_set_p2_add,
    output_set_p2_remove=output_set_p2_remove,
    output_set_p3_union=output_set_p3_union,
    output_set_p3_intersection=output_set_p3_intersection,
    output_set_p3_difference=output_set_p3_difference,
    output_set_p3_symmetric_difference=output_set_p3_symmetric_difference,
    output_set_p4_iteracion=output_set_p4_iteracion,
    output_set_p4_membership=output_set_p4_membership,
    output_set_p5_comprehension=output_set_p5_comprehension,
    output_set_p5_comparison=output_set_p5_comparison,
)

# Renderizar el resultado en el notebook.
display(HTML(final_html_set))
```

```python
# Celda 1
print("--- Creación de Conjuntos y Unicidad ---")

# 1. La forma más común: desde una lista para eliminar duplicados.
numeros_con_duplicados = [1, 2, 3, 2, 4, 1, 5, 5, 'a', 'b', 'a']
conjunto_unicos = set(numeros_con_duplicados)

print(f"Lista original: {numeros_con_duplicados}")
print(f"Conjunto creado (duplicados eliminados): {conjunto_unicos}")

# 2. Usando llaves {} (si el conjunto no está vacío)
conjunto_colores = {'rojo', 'verde', 'azul'}
print(f"Conjunto creado con llaves: {conjunto_colores}")

# 3. ¡Importante! Cómo crear un conjunto VACÍO.
#    {} crea un diccionario vacío, no un conjunto.
diccionario_vacio = {}
conjunto_vacio = set()

print("-" * 20)
print(f"Usar {{}} crea un: {type(diccionario_vacio)}")
print(f"Para un conjunto vacío, debes usar set(): {type(conjunto_vacio)}")
```

```python
# Celda 2
print("--- Añadir y Eliminar Elementos ---")

tags = {'python', 'data'}
print(f"Conjunto inicial: {tags}")

# 1. Añadir un solo elemento con .add()
tags.add('science')
print(f"Tras .add('science'): {tags}")

# 2. Añadir múltiples elementos desde otro iterable con .update()
tags.update(['machine learning', 'ia', 'python']) # 'python' ya existía, así que se ignora
print(f"Tras .update(...): {tags}")

# 3. Eliminar elementos: .remove() vs .discard()
#    .remove() da error si el elemento no existe.
tags.remove('ia')
print(f"Tras .remove('ia'): {tags}")

#    .discard() NO da error si el elemento no existe. Es más seguro.
tags.discard('data')       # Elimina 'data' que sí existe
tags.discard('java')       # Intenta eliminar 'java' que no existe, pero no pasa nada
print(f"Tras .discard('data') y .discard('java'): {tags}")

# Demostración del error de .remove()
try:
    tags.remove('java')
except KeyError as e:
    print(f"\nIntentar .remove('java') causa un error: {type(e).__name__}: {e}")
```

```python
# Celda 3
print("--- Operaciones de Conjuntos ---")

desarrolladores_backend = {'Python', 'Java', 'SQL', 'Docker'}
desarrolladores_frontend = {'JavaScript', 'CSS', 'React', 'Docker'}

print(f"Backend:   {desarrolladores_backend}")
print(f"Frontend:  {desarrolladores_frontend}")
print("-" * 20)

# 1. Unión (|): Todos los desarrolladores de ambos equipos.
todos_los_desarrolladores = desarrolladores_backend | desarrolladores_frontend
print(f"Unión (todos): {todos_los_desarrolladores}")

# 2. Intersección (&): Tecnologías que ambos equipos conocen (Full-Stack).
tecnologias_comunes = desarrolladores_backend & desarrolladores_frontend
print(f"Intersección (comunes): {tecnologias_comunes}")

# 3. Diferencia (-): Tecnologías que solo conoce el equipo de backend.
solo_backend = desarrolladores_backend - desarrolladores_frontend
print(f"Diferencia (solo backend): {solo_backend}")

# 4. Diferencia Simétrica (^): Tecnologías que son exclusivas de un equipo (no compartidas).
tecnologias_exclusivas = desarrolladores_backend ^ desarrolladores_frontend
print(f"Diferencia Simétrica (exclusivas): {tecnologias_exclusivas}")
```

```python
# Celda 4
print("--- Pertenencia (in) e Iteración ---")

ingredientes_disponibles = {'tomate', 'lechuga', 'queso', 'pan', 'cebolla'}

# 1. Comprobación de Pertenencia (extremadamente rápido)
#    Es el uso principal de los conjuntos: ¿Tengo este elemento?
tiene_queso = 'queso' in ingredientes_disponibles
tiene_jamon = 'jamon' in ingredientes_disponibles

print(f"Ingredientes: {ingredientes_disponibles}")
print(f"¿Tenemos 'queso'? -> {tiene_queso}")
print(f"¿Tenemos 'jamon'? -> {tiene_jamon}")
print("\nLa operación 'in' es mucho más rápida en un conjunto que en una lista.")

# 2. Iteración
#    Puedes recorrer un conjunto, pero recuerda que no hay garantía de orden.
print("\nListando ingredientes disponibles:")
for ingrediente in ingredientes_disponibles:
    print(f" - {ingrediente}")
```

```python
# Celda 5
print("--- Set Comprehensions y Comparación ---")

# 1. Set Comprehensions
#    Crear conjuntos de forma concisa, similar a las listas.
#    Ej: Crear un conjunto con las iniciales de una lista de palabras.
palabras = ['hola', 'mundo', 'python', 'es', 'potente']
iniciales = {palabra[0] for palabra in palabras}
print(f"Conjunto de iniciales (creado con comprehension): {iniciales}")

# 2. Comparación de Conjuntos
numeros_pares = {2, 4, 6}
numeros_bajos = {1, 2, 3, 4, 5, 6}
numeros_impares = {1, 3, 5}

print("\n--- Comparando conjuntos ---")
print(f"Pares: {numeros_pares}")
print(f"Bajos: {numeros_bajos}")
print(f"Impares: {numeros_impares}")

# .issubset(): ¿Es un conjunto subconjunto de otro?
print(f"¿Pares es subconjunto de Bajos? -> {numeros_pares.issubset(numeros_bajos)}")

# .issuperset(): ¿Es un conjunto superconjunto de otro?
print(f"¿Bajos es superconjunto de Pares? -> {numeros_bajos.issuperset(numeros_pares)}")

# .isdisjoint(): ¿No tienen elementos en común?
print(f"¿Pares e Impares son disjuntos? -> {numeros_pares.isdisjoint(numeros_impares)}")
```
