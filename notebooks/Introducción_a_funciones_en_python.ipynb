```python
    """
    Recibe una lista de números y devuelve una nueva lista
    conteniendo solo los valores no negativos.
    """
    valores_validos = [] # Lista para almacenar los resultados
    for medicion in mediciones:
        if medicion < 0:
            continue # Ignora este valor y salta a la siguiente iteración
        
        valores_validos.append(medicion)
        
    return valores_validos

# -- Uso de la función --
mediciones_sensor = [10.5, -2.1, 15.0, 8.2, -1.0, 0.0, 5.5]
mediciones_limpias = filtrar_positivos(mediciones_sensor)

print("\nMediciones originales:", mediciones_sensor)
print("Mediciones filtradas (solo válidas):", mediciones_limpias)
```

### 9. Función para Generar Tablas de Multiplicar (Basado en Bucles Anidados)

**Escenario:** Crear una función que pueda generar y mostrar la tabla de multiplicar para cualquier número, hasta un límite definido.

**Explicación de Experto:**
La función `mostrar_tabla_de_multiplicar` encapsula la complejidad de los bucles anidados. El usuario solo necesita especificar qué tabla quiere (`numero_tabla`) y hasta dónde (`limite`). Esto hace que el código sea mucho más semántico y fácil de usar. Por ejemplo, `mostrar_tabla_de_multiplicar(7, 12)` es auto-explicativo.

**Código Python:**
```python
def mostrar_tabla_de_multiplicar(numero_tabla, limite=10):
    """Muestra la tabla de multiplicar para un número dado hasta un límite."""
    print(f"\n--- Tabla de Multiplicar del {numero_tabla} ---")
    for i in range(1, limite + 1):
        resultado = numero_tabla * i
        print(f"  {numero_tabla} x {i:2} = {resultado}")

# -- Uso de la función --
mostrar_tabla_de_multiplicar(7)       # Muestra la tabla del 7 hasta 10
mostrar_tabla_de_multiplicar(3, 5)    # Muestra la tabla del 3 hasta 5
```

### 10. Función para Verificar Condición en Lista (Basado en `for-else`)

**Escenario:** Crear una función booleana que verifique si **todos** los elementos de una lista cumplen una condición (en este caso, ser aprobados).

Esta es una función de validación perfecta. `verificar_todos_aprobados` devuelve `True` o `False`, un resultado booleano claro y conciso. La lógica `for-else` es ideal para este patrón: asumimos que todos cumplen (el `else` se prepara para devolver `True`), pero si encontramos un solo contraejemplo, salimos inmediatamente con `break` y devolvemos `False`. Es eficiente y elegante.

**Código Python:**
```python
def verificar_todos_aprobados(lista_notas, nota_minima=5):
    """
    Verifica si todas las notas en una lista son iguales o superiores a la nota mínima.
    Devuelve True si todos aprueban, False en caso contrario.
    """
    for nota in lista_notas:
        if nota < nota_minima:
            print(f"  -> Verificación fallida. Nota encontrada: {nota}")
            return False # Se encontró un suspenso, no hay necesidad de seguir.
    
    # Este bloque solo se alcanza si el bucle termina sin encontrar un suspenso.
    print("  -> Verificación exitosa. ¡Todos aprobados!")
    return True

# -- Uso de la función --
notas_grupo_a = [8, 9, 7, 10, 9]
notas_grupo_b = [8, 4, 7, 10, 9]

print("\nVerificando Grupo A:")
if verificar_todos_aprobados(notas_grupo_a):
    print("Resultado: El Grupo A puede pasar de curso.")
else:
    print("Resultado: El Grupo A necesita revisión.")

print("\nVerificando Grupo B:")
if verificar_todos_aprobados(notas_grupo_b):
    print("Resultado: El Grupo B puede pasar de curso.")
else:
    print("Resultado: El Grupo B necesita revisión.")
```
