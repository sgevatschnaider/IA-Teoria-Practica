```python
# Celda 1
print("--- Creaci√≥n de Listas ---")

# 1. Usando corchetes [] (la forma m√°s com√∫n)
lista_vacia = []
lista_con_elementos = [1, "hola", 3.14, True] # Pueden contener tipos de datos mezclados

# 2. Usando el constructor list() para convertir otro objeto (como un string)
lista_desde_string = list("python")

print(f"Una lista vac√≠a: {lista_vacia}")
print(f"El tipo de dato es: {type(lista_vacia)}")
print(f"Una lista con elementos variados: {lista_con_elementos}")
print(f"Una lista creada desde el string 'python': {lista_desde_string}")
```

```python
# Celda 2
print("--- Acceso a Elementos y Slicing ---")

elementos = ['P', 'y', 't', 'h', 'o', 'n', '!']
print(f"Nuestra lista de ejemplo: {elementos}\n")

# --- Indexing (acceder a un solo elemento) ---
print("-> Indexing:")
print(f"El primer elemento (√≠ndice 0): {elementos[0]}")
print(f"El tercer elemento (√≠ndice 2): {elementos[2]}")
print(f"El √∫ltimo elemento (√≠ndice -1): {elementos[-1]}")
print(f"El pen√∫ltimo elemento (√≠ndice -2): {elementos[-2]}\n")


# --- Slicing (acceder a una sub-lista) ---
# La sintaxis es: lista[inicio : fin : paso]
# El elemento en el √≠ndice 'fin' NO se incluye.
print("-> Slicing:")
sub_lista = elementos[1:4] # Del √≠ndice 1 hasta el 3 (el 4 no se incluye)
print(f"Slice de [1:4]: {sub_lista}")

desde_el_principio = elementos[:3] # Desde el inicio hasta el √≠ndice 2
print(f"Slice de [:3]: {desde_el_principio}")

hasta_el_final = elementos[3:] # Desde el √≠ndice 3 hasta el final
print(f"Slice de [3:]: {hasta_el_final}")

# Un truco muy com√∫n para copiar una lista
copia_de_lista = elementos[:]
print(f"Una copia completa con [:]: {copia_de_lista}")

# Y el truco para invertir una lista
lista_invertida = elementos[::-1]
print(f"Lista invertida con [::-1]: {lista_invertida}")
```

```python
# Celda 3
print("--- .append() vs .extend() ---")

# Usando .append()
lista_append = [1, 2, 3]
lista_append.append([4, 5]) # A√±ade la lista [4, 5] como UN SOLO elemento
print(f"Resultado con .append([4, 5]): {lista_append}")
print(f"La longitud de la lista ahora es: {len(lista_append)}\n")


# Usando .extend()
lista_extend = [1, 2, 3]
lista_extend.extend([4, 5]) # A√±ade los elementos 4 y 5 de forma individual
print(f"Resultado con .extend([4, 5]): {lista_extend}")
print(f"La longitud de la lista ahora es: {len(lista_extend)}")
```

```python
# Celda 4
print("--- Insertar, eliminar y extraer ---")
mi_lista = ['manzana', 'cereza', 'uva']
print(f"Lista inicial: {mi_lista}\n")

# 1. Insertar 'banana' en el √≠ndice 1
mi_lista.insert(1, 'banana')
print(f"Despu√©s de .insert(1, 'banana'): {mi_lista}\n")

# 2. Eliminar la primera ocurrencia de 'cereza'
mi_lista.remove('cereza')
print(f"Despu√©s de .remove('cereza'): {mi_lista}\n")

# 3. Extraer el elemento en el √≠ndice 2 ('uva')
# .pop() elimina el elemento PERO tambi√©n te lo devuelve
elemento_extraido = mi_lista.pop(2)
print(f"Despu√©s de .pop(2): {mi_lista}")
print(f"El elemento que se extrajo fue: '{elemento_extraido}'")
```

```python
# Celda 5
print("--- .sort() vs sorted() ---")

# 1. Usando .sort() (modifica la lista original)
lista_para_sort = [3, 1, 4, 1, 5, 9, 2]
print(f"Lista original para .sort(): {lista_para_sort}")
lista_para_sort.sort() # No devuelve nada, solo modifica
print(f"Lista despu√©s de .sort(): {lista_para_sort}\n")

# 2. Usando sorted() (devuelve una nueva lista ordenada)
lista_para_sorted = [3, 1, 4, 1, 5, 9, 2]
print(f"Lista original para sorted(): {lista_para_sorted}")
nueva_lista_ordenada = sorted(lista_para_sorted)
print(f"La nueva lista creada por sorted(): {nueva_lista_ordenada}")
print(f"La lista original NO cambi√≥: {lista_para_sorted}")
```

```python
# Celda 6
print("--- Operadores + y * ---")
lista_a = [1, 2]
lista_b = ['A', 'B']

# Concatenaci√≥n con '+'
concatenada = lista_a + lista_b
print(f"Unir {lista_a} + {lista_b} = {concatenada}\n")

# Repetici√≥n con '*'
repetida = lista_a * 3
print(f"Repetir {lista_a} * 3 = {repetida}")
```

```python
# Celda 7
print("--- Iterando con un bucle for ---")
frutas = ['manzana', 'banana', 'cereza']
for indice, fruta in enumerate(frutas):
  print(f"Elemento en el √≠ndice {indice} es: {fruta}")
```

```python
# Celda 8
print("--- List Comprehensions ---")

# Ejemplo 1: Crear una lista con los cuadrados de los n√∫meros del 0 al 9
cuadrados = [numero**2 for numero in range(10)]
print(f"Cuadrados del 0 al 9: {cuadrados}\n")

# Ejemplo 2: Crear una lista solo con los n√∫meros pares del 0 al 19
numeros_pares = [n for n in range(20) if n % 2 == 0]
print(f"N√∫meros pares del 0 al 19: {numeros_pares}\n")

# Ejemplo 3: Convertir una lista de palabras a may√∫sculas
palabras = ['hola', 'mundo', 'python']
mayusculas = [p.upper() for p in palabras]
print(f"Palabras en may√∫sculas: {mayusculas}")
```
