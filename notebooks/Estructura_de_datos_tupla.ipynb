```python
# Celda 1
print("--- Creaci√≥n de Tuplas ---")

# 1. Usando par√©ntesis () (la forma m√°s com√∫n)
tupla_vacia = ()
tupla_con_elementos = (1, "hola", 3.14, True) # Tambi√©n pueden contener tipos de datos mezclados

# 2. Usando el constructor tuple() para convertir otro iterable (como una lista o un string)
tupla_desde_lista = tuple([1, 2, 3])
tupla_desde_string = tuple("python")

# 3. ¬°Caso especial y muy importante! Tupla de un solo elemento
#    Se necesita una coma al final para que Python no lo confunda con un n√∫mero entre par√©ntesis.
tupla_un_elemento = (42,)
no_es_una_tupla = (42) # Esto es simplemente el n√∫mero 42

print(f"Una tupla vac√≠a: {tupla_vacia}")
print(f"El tipo de dato es: {type(tupla_vacia)}")
print(f"Una tupla con elementos variados: {tupla_con_elementos}")
print(f"Una tupla creada desde una lista: {tupla_desde_lista}")
print(f"Una tupla creada desde el string 'python': {tupla_desde_string}")
print("-" * 20)
print(f"Tupla de un solo elemento (nota la coma): {tupla_un_elemento}, tipo: {type(tupla_un_elemento)}")
print(f"Esto NO es una tupla: {no_es_una_tupla}, tipo: {type(no_es_una_tupla)}")
```

```python
# Celda 2
print("--- Acceso a Elementos (Funciona igual que en las listas) ---")

elementos = ('P', 'y', 't', 'h', 'o', 'n')

# 1. Acceso por √≠ndice (empieza en 0)
primer_elemento = elementos[0]

# 2. Acceso con √≠ndice negativo (el -1 es el √∫ltimo)
ultimo_elemento = elementos[-1]

# 3. Slicing o rebanado para obtener una sub-tupla
sub_tupla = elementos[1:4] # Elementos desde el √≠ndice 1 hasta el 3

# 4. Slicing para invertir la tupla
tupla_invertida = elementos[::-1]

print(f"Tupla original: {elementos}")
print(f"Primer elemento [0]: {primer_elemento}")
print(f"√öltimo elemento [-1]: {ultimo_elemento}")
print(f"Sub-tupla con slice [1:4]: {sub_tupla}")
print(f"Tupla invertida con slice [::-1]: {tupla_invertida}")
```

```python
# Celda 3
print("--- La Inmutabilidad en Acci√≥n (Intentos de Modificaci√≥n) ---")

coordenadas = (1920, 1080)
print(f"Tupla original: {coordenadas}")

# 1. Intentar cambiar un elemento por su √≠ndice (¬°esto fallar√°!)
try:
    coordenadas[0] = 1280
except TypeError as e:
    print(f"\nError al intentar cambiar un elemento (coordenadas[0] = 1280):")
    print(f"  -> {e}")

# 2. Intentar a√±adir un elemento (las tuplas no tienen el m√©todo .append())
try:
    coordenadas.append(720)
except AttributeError as e:
    print(f"\nError al intentar a√±adir un elemento (coordenadas.append(720)):")
    print(f"  -> {e}")

print("\nLa tupla original permanece intacta:", coordenadas)
```

```python
# Celda 4
print("--- 'Modificar' Creando una Nueva Tupla ---")

punto_2d = (10, 20)
print(f"Tupla original (punto 2D): {punto_2d}")
print(f"ID de memoria de la tupla original: {id(punto_2d)}")

# Para "a√±adir" un elemento, creamos una nueva tupla concatenando la original con otra
punto_3d = punto_2d + (30,) # Nota la coma para crear una tupla de un elemento

print(f"\nNueva tupla (punto 3D): {punto_3d}")
print(f"ID de memoria de la nueva tupla: {id(punto_3d)}")
print("\n¬°Observa que el ID de memoria es diferente! Hemos creado un objeto completamente nuevo.")
```

```python
# Celda 5
print("--- M√©todos de Tupla y Desempaquetado (Unpacking) ---")

notas = (10, 9, 7, 8, 9, 9, 10)

# 1. M√©todos .count() y .index()
conteo_de_9 = notas.count(9)
posicion_del_8 = notas.index(8)

print(f"Tupla de notas: {notas}")
print(f"El n√∫mero 9 aparece {conteo_de_9} veces (.count(9))")
print(f"El n√∫mero 8 est√° en la posici√≥n {posicion_del_8} (.index(8))")

# 2. Desempaquetado de Tuplas (Tuple Unpacking)
#    Se asignan los elementos de una tupla a variables individuales.
#    ¬°Esto es extremadamente √∫til y muy Pythonic!
persona = ("Ana", "Garc√≠a", 30)
nombre, apellido, edad = persona

print("\n--- Desempaquetado ---")
print(f"Tupla original: {persona}")
print(f"Variable 'nombre': {nombre}")
print(f"Variable 'apellido': {apellido}")
print(f"Variable 'edad': {edad}")
```

```python
# Celda 6
print("--- Casos de Uso Pr√°cticos ---")

# 1. Devolver m√∫ltiples valores de una funci√≥n
def obtener_min_max(datos):
    """Devuelve el valor m√≠nimo y m√°ximo de una secuencia."""
    return (min(datos), max(datos)) # Python empaqueta los valores en una tupla

numeros = [4, 1, 88, -10, 42]
minimo, maximo = obtener_min_max(numeros) # Se desempaqueta el resultado directamente

print("1. Devolviendo m√∫ltiples valores de una funci√≥n:")
print(f"  Los n√∫meros son: {numeros}")
print(f"  El valor m√≠nimo es {minimo} y el m√°ximo es {maximo}")


# 2. Usar como claves en un diccionario
#    Las claves de un diccionario DEBEN ser inmutables. Una tupla lo es, una lista no.
capitales = {
    ("Espa√±a", "Europa"): "Madrid",
    ("Colombia", "Am√©rica"): "Bogot√°",
    ("Jap√≥n", "Asia"): "Tokio"
}
# Esto fallar√≠a: capitales_error = {["Espa√±a", "Europa"]: "Madrid"} -> TypeError

clave_consulta = ("Espa√±a", "Europa")
print("\n2. Usando tuplas como claves de diccionario:")
print(f"  La capital de {clave_consulta} es: {capitales[clave_consulta]}")
```
