```python
import random
import collections
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

def ejecutar_simulacion():
    # =============================================================================
    # SIMULACIÓN DE DADOS: ESTRUCTURAS DE DATOS Y ESTADÍSTICA
    # =============================================================================

    print("="*80)
    print("ANÁLISIS COMPLETO: LANZAMIENTO DE DOS DADOS")
    print("="*80)

    # Parámetros de la simulación - INPUT DEL USUARIO
    while True:
        try:
            N_LANZAMIENTOS = int(input("Ingrese el número de lanzamientos para la simulación (ej: 1000, 100000): "))
            if N_LANZAMIENTOS > 0:
                break
            else:
                print("Por favor ingrese un número positivo.")
        except ValueError:
            print("Por favor ingrese un número entero válido.")

    # =============================================================================
    # 1. ESTRUCTURAS DE DATOS BÁSICAS
    # =============================================================================
    print("\n1. ESTRUCTURAS DE DATOS UTILIZADAS:")
    print("-" * 40)

    # Lista para almacenar resultados (estructura de datos lineal)
    resultados = []
    lanzamientos_individuales = []  # Lista de tuplas para cada lanzamiento

    print(f"• LISTA (resultados): Almacena {N_LANZAMIENTOS:,} sumas.")
    print(f"• LISTA DE TUPLAS: Almacena cada lanzamiento como (dado1, dado2, suma).")

    # Simulación con registro detallado
    print("\nGenerando datos...")
    for i in range(N_LANZAMIENTOS):
        dado1 = random.randint(1, 6)
        dado2 = random.randint(1, 6)
        suma = dado1 + dado2

        # Almacenar en diferentes estructuras
        resultados.append(suma)  # Lista simple
        lanzamientos_individuales.append((dado1, dado2, suma))  # Lista de tuplas # [(1,2,5), (1,2,6)]

        # Mostrar los primeros 10 lanzamientos
        if i < 10:
            print(f"  Lanzamiento {i+1:2d}: Dado1={dado1}, Dado2={dado2}, Suma={suma}")

    if N_LANZAMIENTOS > 10:
        print(f"... (continuando hasta {N_LANZAMIENTOS:,} lanzamientos)")

    # =============================================================================
    # 2. COUNTER - ESTRUCTURA DE DATOS ESPECIALIZADA
    # =============================================================================
    print(f"\n2. COUNTER (collections.Counter):")
    print("-" * 40)

    # Counter es un diccionario especializado para contar [6,5, 2, 6]  2 1  5 1 6 2
    conteo = collections.Counter(resultados)
    print(f"• Counter es una subclase de dict optimizada para el conteo de frecuencias.")
    print(f"• Tipo de dato: {type(conteo)}")

    print("\nFrecuencias empíricas encontradas:")
    for suma in sorted(conteo.keys()):
        print(f"  Suma {suma:2d}: {conteo[suma]:,} veces")

    # Métodos útiles del Counter
    print(f"\n• Suma más común: {conteo.most_common(1)[0]}")
    print(f"• Total de elementos contados: {sum(conteo.values()):,}")

    # =============================================================================
    # 3. ESTADÍSTICAS DESCRIPTIVAS
    # =============================================================================
    print(f"\n3. ESTADÍSTICAS DESCRIPTIVAS:")
    print("-" * 40)

    # Convertir a array de NumPy para cálculos estadísticos eficientes
    datos_np = np.array(resultados)# [6,2,6] 14/3

    # Medidas de tendencia central
    media = np.mean(datos_np)
    mediana = np.median(datos_np)
    moda_resultado = stats.mode(datos_np)
    moda_valor = moda_resultado.mode
    moda_conteo = moda_resultado.count

    print(f"• Media (promedio empírico): {media:.4f} (Teórica: 7.0)")
    print(f"• Mediana (valor central): {mediana:.4f}")
    print(f"• Moda (valor más frecuente): {moda_valor} (apareció {moda_conteo:,} veces)")

    # Medidas de dispersión
    varianza = np.var(datos_np)
    desviacion_std = np.std(datos_np)
    rango = np.max(datos_np) - np.min(datos_np)

    print(f"• Varianza: {varianza:.4f} (Teórica: ~5.833)")
    print(f"• Desviación estándar: {desviacion_std:.4f} (Teórica: ~2.415)")
    print(f"• Rango: {rango} (de {np.min(datos_np)} a {np.max(datos_np)})")

    # =============================================================================
    # 4. PROBABILIDADES TEÓRICAS VS EMPÍRICAS
    # =============================================================================
    print(f"\n4. ANÁLISIS DE PROBABILIDADES:")
    print("-" * 40)

    # Diccionario con probabilidades teóricas (estructura clave-valor)
    prob_teorica = {s: len([(d1, d2) for d1 in range(1, 7) for d2 in range(1, 7) if d1 + d2 == s]) / 36 for s in range(2, 13)}

    print("SUMA │ FRECUENCIA │ PROB.EMPÍRICA │ PROB.TEÓRICA │ DIFERENCIA ABS.")
    print("─────┼────────────┼───────────────┼──────────────┼────────────────")

    diferencias_totales = 0
    for suma in range(2, 13):
        freq = conteo.get(suma, 0) # Usar .get para evitar errores si una suma nunca aparece
        prob_empirica = freq / N_LANZAMIENTOS
        prob_teorica_val = prob_teorica[suma]
        diferencia = abs(prob_empirica - prob_teorica_val)
        diferencias_totales += diferencia

        print(f"{suma:4d} │ {freq:10,} │ {prob_empirica:13.5f} │ {prob_teorica_val:12.5f} │ {diferencia:14.5f}")

    print(f"\n• Diferencia promedio entre probabilidades: {diferencias_totales/11:.6f}")
    print("  (Este valor tiende a cero a medida que aumenta el número de lanzamientos)")

    # =============================================================================
    # 5. VISUALIZACIÓN AVANZADA
    # =============================================================================
    print(f"\n5. GENERANDO VISUALIZACIONES...")

    x = list(range(2, 13))
    y_empirica = [conteo.get(s, 0) for s in x]

    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle(f'Análisis Completo: {N_LANZAMIENTOS:,} Lanzamientos de Dos Dados', fontsize=16, fontweight='bold')

    # Gráfico 1: Distribución de frecuencias vs Frecuencia Teórica Esperada
    y_teorica_esperada = [prob_teorica[s] * N_LANZAMIENTOS for s in x]
    ax1.bar(x, y_empirica, color='royalblue', alpha=0.7, label='Frecuencia Empírica')
    ax1.plot(x, y_teorica_esperada, 'ro-', linewidth=2, markersize=6, label='Frecuencia Teórica Esperada')
    ax1.set_title('Distribución de Frecuencias (Empírica vs. Teórica)')
    ax1.set_xlabel('Suma de los dados')
    ax1.set_ylabel('Frecuencia')
    ax1.legend()
    ax1.grid(axis='y', linestyle='--', alpha=0.6)

    # Gráfico 2: Comparación de Probabilidades
    prob_emp = [f/N_LANZAMIENTOS for f in y_empirica]
    prob_teo = [prob_teorica[s] for s in x]
    ax2.bar([i-0.2 for i in x], prob_emp, width=0.4, color='lightcoral', label='Prob. Empírica')
    ax2.bar([i+0.2 for i in x], prob_teo, width=0.4, color='lightgreen', label='Prob. Teórica')
    ax2.set_title('Comparación de Probabilidades')
    ax2.set_xlabel('Suma de los dados')
    ax2.set_ylabel('Probabilidad')
    ax2.legend()
    ax2.grid(axis='y', linestyle='--', alpha=0.6)

    # Gráfico 3: Error Absoluto
    diferencias = [abs(p_emp - p_teo) for p_emp, p_teo in zip(prob_emp, prob_teo)]
    ax3.bar(x, diferencias, color='orange', alpha=0.8)
    ax3.set_title('Error Absoluto |P(Empírica) - P(Teórica)|')
    ax3.set_xlabel('Suma de los dados')
    ax3.set_ylabel('Diferencia Absoluta')
    ax3.grid(axis='y', linestyle='--', alpha=0.6)

    # Gráfico 4: Distribución Acumulativa
    cdf_empirica = np.cumsum(prob_emp)
    ax4.plot(x, cdf_empirica, 'b-', marker='o', label='CDF Empírica')
    ax4.set_title('Función de Distribución Acumulativa (CDF)')
    ax4.set_xlabel('Suma de los dados')
    ax4.set_ylabel('Probabilidad Acumulada')
    ax4.legend()
    ax4.grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

    # =============================================================================
    # 6. ANÁLISIS DE CONVERGENCIA (LEY DE LOS GRANDES NÚMEROS)
    # =============================================================================
    print(f"\n6. ANÁLISIS DE CONVERGENCIA (Ley de los Grandes Números):")
    print("-" * 60)

    media_teorica = 7.0
    prob_7_teorica = 1/6

    # Generar puntos de muestra de forma logarítmica para ver mejor la convergencia
    if N_LANZAMIENTOS < 100:
        tamaños_muestra = np.unique(np.linspace(10, N_LANZAMIENTOS, 10, dtype=int))
    else:
        tamaños_muestra = np.unique(np.logspace(1, np.log10(N_LANZAMIENTOS), num=50, dtype=int))
        if tamaños_muestra[-1] != N_LANZAMIENTOS:
             tamaños_muestra = np.append(tamaños_muestra, N_LANZAMIENTOS)

    # Asegurarse de que no haya tamaños de muestra cero
    tamaños_muestra = tamaños_muestra[tamaños_muestra > 0]

    print(f"Analizando la convergencia en {len(tamaños_muestra)} puntos de muestra (de 1 a {N_LANZAMIENTOS:,})...")

    medias_empiricas = [np.mean(datos_np[:n]) for n in tamaños_muestra]
    probs_7_empiricas = [(np.sum(datos_np[:n] == 7) / n) for n in tamaños_muestra]

    errores_media = [abs(m - media_teorica) for m in medias_empiricas]
    errores_prob7 = [abs(p - prob_7_teorica) for p in probs_7_empiricas]

    # Visualización de la convergencia
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
    fig.suptitle('Convergencia hacia Valores Teóricos', fontsize=16, fontweight='bold')

    ax1.plot(tamaños_muestra, medias_empiricas, 'bo-', markersize=4, label='Media Empírica')
    ax1.axhline(y=media_teorica, color='r', linestyle='--', label=f'Media Teórica ({media_teorica})')
    ax1.set_title('Convergencia de la Media')
    ax1.set_xlabel('Número de Lanzamientos')
    ax1.set_ylabel('Media de las Sumas')
    ax1.set_xscale('log')
    ax1.legend()
    ax1.grid(True, which="both", ls="--", alpha=0.6)

    ax2.plot(tamaños_muestra, probs_7_empiricas, 'go-', markersize=4, label='P(Suma=7) Empírica')
    ax2.axhline(y=prob_7_teorica, color='r', linestyle='--', label=f'P(Suma=7) Teórica ({prob_7_teorica:.4f})')
    ax2.set_title('Convergencia de la Probabilidad para la Suma = 7')
    ax2.set_xlabel('Número de Lanzamientos')
    ax2.set_ylabel('Probabilidad')
    ax2.set_xscale('log')
    ax2.legend()
    ax2.grid(True, which="both", ls="--", alpha=0.6)

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

    print(f"\n📊 ANÁLISIS FINAL DE CONVERGENCIA:")
    print("-" * 60)
    print(f"Tras {N_LANZAMIENTOS:,} lanzamientos:")
    print(f"• Media final: {medias_empiricas[-1]:.6f} (Error: {errores_media[-1]:.6f})")
    print(f"• P(Suma=7) final: {probs_7_empiricas[-1]:.6f} (Error: {errores_prob7[-1]:.6f})")

    if len(tamaños_muestra) > 10:
        # Calcular la tasa de convergencia usando los últimos puntos
        ultimos_errores = np.log(errores_media[-5:])
        ultimos_n = np.log(tamaños_muestra[-5:])
        # La teoría (Teorema del Límite Central) sugiere que el error es proporcional a 1/sqrt(N)
        # Por lo tanto, log(error) ≈ C - 0.5 * log(N). La pendiente debería ser cercana a -0.5.
        pendiente, _, _, _, _ = stats.linregress(ultimos_n, ultimos_errores)
        print(f"• Tasa de convergencia del error (pendiente log-log): {pendiente:.4f} (Teórica: -0.5)")

    # =============================================================================
    # 7. RESUMEN DE ESTRUCTURAS DE DATOS UTILIZADAS
    # =============================================================================
    print(f"\n7. RESUMEN DE ESTRUCTURAS DE DATOS:")
    print("="*50)
    print(f"• LISTA (list): Almacenamiento secuencial de {N_LANZAMIENTOS:,} resultados.")
    print(f"• TUPLA (tuple): Estructura inmutable para registrar cada lanzamiento (dado1, dado2, suma).")
    print(f"• DICCIONARIO (dict): Pares clave-valor para las probabilidades teóricas. Acceso O(1).")
    print(f"• COUNTER (collections.Counter): Subclase de dict para conteo de frecuencias optimizado.")
    print(f"• ARRAY DE NUMPY (np.array): Estructura optimizada para cálculo numérico y estadístico masivo.")

    # =============================================================================
    # 8. CONCEPTOS ESTADÍSTICOS DEMOSTRADOS
    # =============================================================================
    print(f"\n8. CONCEPTOS ESTADÍSTICOS DEMOSTRADOS:")
    print("="*50)
    print("• Distribución de Frecuencias: Cómo se distribuyen los resultados observados.")
    print("• Medidas de Tendencia Central: Media, mediana y moda de los datos.")
    print("• Medidas de Dispersión: Varianza y desviación estándar que miden la volatilidad.")
    print("• Probabilidad Teórica vs. Empírica: Comparación entre lo que la matemática predice y lo que la simulación muestra.")
    print("• Ley de los Grandes Números: La probabilidad empírica converge a la teórica a medida que N aumenta.")
    print("• Teorema del Límite Central: La distribución de las sumas se aproxima a una distribución normal.")
    print("• Análisis de Convergencia: Visualización de cómo el error disminuye con más datos.")

# =============================================================================
# BLOQUE PRINCIPAL DE EJECUCIÓN
# =============================================================================
if __name__ == "__main__":
    while True:
        ejecutar_simulacion()

        # Preguntar al usuario si desea realizar otra simulación
        respuesta = input("\n¿Desea realizar otra simulación? (s/n): ").lower()
        if respuesta not in ['s', 'si', 'yes', 'y']:
            print("\n¡Gracias por usar la simulación del casino! ¡Juegue con responsabilidad!")
            break

        print("\n" + "="*80)
        print("REINICIANDO SIMULACIÓN...")
```
