```python
from IPython.display import display, HTML

# ==============================================================================
# 1. PLANTILLA HTML PARA DICCIONARIOS (CON FOOTER A√ëADIDO)
# ==============================================================================
html_template_dict = """
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <title>Gu√≠a Interactiva de Diccionarios en Python</title>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">
  <style>
    :root {{
      --bg-primary: #1a1a2e; --bg-secondary: rgba(27, 40, 69, 0.85); --bg-tertiary: rgba(45, 55, 72, 0.7);
      --text-primary: #e0e0e0; --text-secondary: #a0aec0; --accent-primary: #4eacfa; --accent-secondary: #c471ed;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(255, 255, 255, 0.15); --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.2);
      --border-radius: 15px; --transition: all 0.4s cubic-bezier(0.25, 0.8, 0.25, 1);
    }}
    * {{ margin: 0; padding: 0; box-sizing: border-box; }}
    html {{ scroll-behavior: smooth; }}
    body {{ font-family: 'Inter', sans-serif; line-height: 1.8; background-color: var(--bg-primary); color: var(--text-primary); }}
    .container {{ max-width: 1000px; margin: 0 auto; padding: 2rem; }}
    .header {{ text-align: center; margin-bottom: 3rem; }}
    .main-title {{ font-size: 3.5rem; font-weight: 800; background: var(--accent-gradient); -webkit-background-clip: text; color: transparent; background-clip: text; margin-bottom: 1rem; }}
    .subtitle {{ font-size: 1.3rem; color: var(--text-secondary); font-weight: 400; max-width: 800px; margin: auto; }}
    .lesson-container {{ display: flex; flex-direction: column; gap: 1.5rem; }}
    .topic-card {{ background: var(--bg-secondary); backdrop-filter: blur(10px); border-radius: var(--border-radius); box-shadow: var(--shadow-card); border: 1px solid var(--border-color); overflow: hidden; transition: var(--transition); }}
    .topic-header {{ cursor: pointer; padding: 1.5rem 2rem; display: flex; justify-content: space-between; align-items: center; }}
    .topic-title {{ font-size: 1.3rem; font-weight: 600; color: var(--text-primary); }}
    .expand-icon {{ font-size: 1.2rem; color: var(--text-secondary); transition: var(--transition); }}
    .topic-card.open .expand-icon {{ transform: rotate(180deg); }}
    .topic-content {{ max-height: 0; overflow: hidden; transition: max-height 1s ease-in-out, padding 1s ease-in-out; background: var(--bg-tertiary); }}
    .topic-card.open .topic-content {{ max-height: 5000px; padding: 1.5rem 2rem; border-top: 1px solid var(--border-color); }}
    .topic-content h3 {{ font-size: 1.2rem; color: var(--accent-primary); margin: 1.5rem 0 1rem 0; padding-bottom: 0.5rem; border-bottom: 2px solid var(--accent-secondary); }}
    .topic-content h3:first-child {{ margin-top: 0; }}
    .topic-content p, .topic-content li {{ margin-bottom: 1rem; color: var(--text-secondary); line-height: 1.7; }}
    .topic-content ul {{ padding-left: 20px; }}
    .topic-content strong {{ color: #ffffff; font-weight: 600; }}
    pre, .diagram {{ background: #0f172a; padding: 1.2rem; border-radius: 10px; border: 1px solid var(--border-color); overflow-x: auto; margin: 1.5rem 0; font-family: 'JetBrains Mono', monospace; font-size: 0.9em; color: #cbd5e1; white-space: pre-wrap; word-wrap: break-word; }}
    .code-output {{ background: #1e293b; color: #e2e8f0; font-family: 'JetBrains Mono', monospace; padding: 1rem; border-radius: 0 0 10px 10px; margin-top: -1.5rem; border: 1px solid var(--border-color); border-top:none; white-space: pre; }}
    .tip-box {{ background: rgba(78, 172, 250, 0.1); border-left: 4px solid var(--accent-primary); border-radius: 8px; padding: 1rem; margin: 1.5rem 0; }}
    .tip-box::before {{ content: 'üí° Pythonic Tip: '; font-weight: bold; color: var(--accent-primary);}}

    /* === ESTILOS PARA EL FOOTER A√ëADIDO === */
    footer {{
      text-align: center;
      margin-top: 3rem;
      padding-top: 1.5rem;
      border-top: 1px solid var(--border-color);
    }}
    footer p {{
      color: var(--text-secondary);
      font-size: 0.9rem;
    }}
  </style>
</head>
<body>
  <div class="container">
    <header class="header">
      <h1 class="main-title">Gu√≠a Interactiva de Diccionarios en Python</h1>
      <p class="subtitle">El "mapa" de Python para datos. Aprende a dominar los pares clave-valor, la estructura de datos m√°s flexible y fundamental para cualquier programador.</p>
    </header>
    <div class="lesson-container">
        <!-- Parte 1 -->
        <div class="topic-card open">
            <div class="topic-header"><span class="topic-title">Parte 1: ¬øQu√© es un Diccionario y C√≥mo se Crea?</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <p>Un diccionario es una colecci√≥n de <strong>pares clave-valor</strong>. En lugar de acceder a los elementos por su posici√≥n (como en las listas), los accedes mediante una "clave" √∫nica.</p>
                <ul><li><strong>Mutable:</strong> Puedes a√±adir, modificar y eliminar pares clave-valor despu√©s de su creaci√≥n.</li><li><strong>Desordenado (hist√≥ricamente):</strong> Antes de Python 3.7, no se garantizaba el orden. <strong>Desde Python 3.7, los diccionarios recuerdan el orden de inserci√≥n</strong>.</li><li><strong>Claves √önicas e Inmutables:</strong> Cada clave en un diccionario debe ser √∫nica. Adem√°s, las claves deben ser de un tipo de dato inmutable (strings, n√∫meros, tuplas).</li></ul>
                <h3>Formas de Crear un Diccionario</h3>
                <p>Puedes crear un diccionario usando llaves <code>{{}}</code> o con el constructor <code>dict()</code>.</p>
                <pre># Creaci√≥n con literales {{}}\ndict_vacio = {{}}\nusuario = {{"nombre": "Ada", "edad": 42, "profesion": "Programadora"}}\n\n# Creaci√≥n con el constructor dict()\ndict_con_constructor = dict(marca="Ford", modelo="Mustang", a√±o=1969)</pre>
                <div class="code-output">{output_dict_p1_creacion}</div>
            </div>
        </div>
        <!-- Parte 2 -->
        <div class="topic-card">
            <div class="topic-header"><span class="topic-title">Parte 2: Acceso y Modificaci√≥n de Datos</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <h3>Acceso a Valores por Clave</h3>
                <p>Usas la sintaxis de corchetes <code>diccionario[clave]</code> para obtener el valor asociado. Si la clave no existe, se produce un error <code>KeyError</code>.</p>
                <div class="tip-box">Usa el m√©todo <code>.get(clave, valor_por_defecto)</code> para acceder de forma segura. Si la clave no existe, devuelve <code>None</code> o el valor por defecto que especifiques, sin causar un error.</div>
                <pre>usuario = {{"nombre": "Ada", "id": 101}}\n# Acceso directo\nnombre_usuario = usuario["nombre"]\n\n# Acceso seguro con .get()\nprofesion = usuario.get("profesion", "No especificada")\n\n# Intento de acceso a una clave que no existe (causar√° error)\n# email = usuario["email"] # KeyError!</pre>
                <div class="code-output">{output_dict_p2_acceso}</div>
                <h3>A√±adir y Modificar Pares</h3>
                <p>La misma sintaxis <code>diccionario[clave] = valor</code> se usa tanto para modificar el valor de una clave existente como para a√±adir un nuevo par clave-valor.</p>
                <pre># Modificar una clave existente\nusuario["id"] = 102 # Cambiamos el id de 101 a 102\n\n# A√±adir una nueva clave\nusuario["ciudad"] = "Londres"</pre>
                <div class="code-output">{output_dict_p2_modificacion}</div>
            </div>
        </div>
        <!-- Parte 3 -->
        <div class="topic-card">
            <div class="topic-header"><span class="topic-title">Parte 3: Eliminar Elementos</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <p>Hay dos formas principales de eliminar pares clave-valor de un diccionario.</p>
                <h3>Usando la declaraci√≥n <code>del</code></h3>
                <p><code>del diccionario[clave]</code> elimina el par clave-valor. Si la clave no existe, produce un <code>KeyError</code>.</p>
                <pre>config = {{"host": "localhost", "port": 8080, "debug": True}}\ndel config["debug"]</pre>
                <div class="code-output">{output_dict_p3_del}</div>
                <h3>Usando el m√©todo <code>.pop()</code></h3>
                <p><code>diccionario.pop(clave)</code> elimina el par y <strong>devuelve el valor eliminado</strong>. Es √∫til cuando necesitas usar el valor que est√°s quitando. Tambi√©n produce un <code>KeyError</code> si la clave no existe.</p>
                <pre># Partiendo del diccionario anterior: {{"host": "localhost", "port": 8080}}\npuerto_eliminado = config.pop("port")</pre>
                <div class="code-output">{output_dict_p3_pop}</div>
            </div>
        </div>
        <!-- Parte 4 -->
        <div class="topic-card">
            <div class="topic-header"><span class="topic-title">Parte 4: Iteraci√≥n: Recorriendo el Diccionario</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <p>Iterar sobre diccionarios es una tarea com√∫n. Python ofrece varias formas de hacerlo eficientemente.</p>
                <div class="tip-box">La forma m√°s com√∫n y "Pythonic" es usar el m√©todo <code>.items()</code>, que te da acceso tanto a la clave como al valor en cada paso del bucle.</div>
                <ul>
                    <li><code>for clave in mi_dict:</code> - Itera sobre las claves.</li>
                    <li><code>for valor in mi_dict.values():</code> - Itera sobre los valores.</li>
                    <li><code>for clave, valor in mi_dict.items():</code> - Itera sobre los pares (clave, valor).</li>
                </ul>
                <pre>inventario = {{"manzanas": 50, "bananas": 25, "naranjas": 100}}\n\nfor fruta, cantidad in inventario.items():\n    print(f"Hay {{cantidad}} de {{fruta}}s.")</pre>
                <div class="code-output">{output_dict_p4_iteracion}</div>
            </div>
        </div>
        <!-- Parte 5 -->
        <div class="topic-card">
            <div class="topic-header"><span class="topic-title">Parte 5: T√©cnicas Avanzadas y Dict Comprehensions</span><i class="fas fa-chevron-down expand-icon"></i></div>
            <div class="topic-content">
                <h3>Dict Comprehensions</h3>
                <p>Al igual que las listas, los diccionarios tienen su propia sintaxis de "comprensi√≥n" para crear diccionarios de forma concisa. La estructura es <code>{{clave_expr: valor_expr for item in iterable}}</code>.</p>
                <pre># Crear un diccionario con n√∫meros y sus cuadrados\ncuadrados = {{x: x**2 for x in range(5)}}\n\n# Crear un diccionario a partir de una lista\nfrutas = ["manzana", "banana", "cereza"]\nstock = {{fruta: 10 for fruta in frutas}}</pre>
                <div class="code-output">{output_dict_p5_comprehension}</div>
                <h3>Fusionar Diccionarios</h3>
                <p>Desde Python 3.9, puedes usar el operador de uni√≥n <code>|</code> para fusionar dos diccionarios de forma muy elegante. Si hay claves repetidas, prevalece el valor del diccionario de la derecha.</p>
                <pre>default_config = {{"theme": "dark", "lang": "es"}}\nuser_config = {{"lang": "en", "user": "root"}}\n\n# Fusionar usando el operador |\nfinal_config = default_config | user_config</pre>
                <div class="code-output">{output_dict_p5_merge}</div>
            </div>
        </div>
    </div>
    <!-- === FOOTER A√ëADIDO === -->
    <footer>
      <p>Material elaborado por prof. Sergio Gevatschnaider</p>
    </footer>
  </div>
  <script>
    (function() {{
        document.querySelectorAll('.topic-header').forEach(header => {{
            header.addEventListener('click', () => {{
                document.querySelectorAll('.topic-card.open').forEach(openCard => {{
                    if (openCard !== header.parentElement) {{
                        openCard.classList.remove('open');
                    }}
                }});
                header.parentElement.classList.toggle('open');
            }});
        }});
    }})();
  </script>
</body>
</html>
"""

# ==============================================================================
# 2. L√ìGICA PYTHON PARA GENERAR LOS RESULTADOS DE DICCIONARIOS
# ==============================================================================

# Parte 1: Creaci√≥n
dict_vacio = {}
usuario_dict = {"nombre": "Ada", "edad": 42, "profesion": "Programadora"}
dict_con_constructor = dict(marca="Ford", modelo="Mustang", a√±o=1969)
output_dict_p1_creacion = (f"Diccionario vac√≠o: {dict_vacio} (tipo: {type(dict_vacio)})\n"
                           f"Diccionario de usuario: {usuario_dict}\n"
                           f"Creado con dict(): {dict_con_constructor}")

# Parte 2: Acceso y Modificaci√≥n
usuario_acceso = {"nombre": "Ada", "id": 101}
acceso_msg = ""
try:
    _ = usuario_acceso["email"]
except KeyError as e:
    acceso_msg = f"Error al acceder a usuario['email']: KeyError: {e}"

output_dict_p2_acceso = (f"Diccionario: {usuario_acceso}\n"
                         f"Acceso con ['nombre']: '{usuario_acceso['nombre']}'\n"
                         f"Acceso seguro con .get('profesion', 'No especificada'): '{usuario_acceso.get('profesion', 'No especificada')}'\n"
                         f"{acceso_msg}")

usuario_mod = {"nombre": "Ada", "id": 101}
dict_antes = usuario_mod.copy()
usuario_mod["id"] = 102
usuario_mod["ciudad"] = "Londres"
output_dict_p2_modificacion = (f"Diccionario original: {dict_antes}\n"
                               f"Tras modificar 'id' y a√±adir 'ciudad': {usuario_mod}")

# Parte 3: Eliminaci√≥n
config_del = {"host": "localhost", "port": 8080, "debug": True}
config_antes_del = config_del.copy()
del config_del["debug"]
output_dict_p3_del = (f"Diccionario original: {config_antes_del}\n"
                      f"Tras 'del config[\"debug\"]': {config_del}")

config_pop = config_del.copy() # Usamos el resultado del 'del' anterior
puerto_eliminado = config_pop.pop("port")
output_dict_p3_pop = (f"Diccionario original: {config_del}\n"
                      f"Tras '.pop(\"port\")': {config_pop}\n"
                      f"Valor devuelto por .pop(): {puerto_eliminado}")

# Parte 4: Iteraci√≥n
inventario = {"manzanas": 50, "bananas": 25, "naranjas": 100}
iteracion_res = [f"Hay {cantidad} de {fruta}s." for fruta, cantidad in inventario.items()]
output_dict_p4_iteracion = "\n".join(iteracion_res)

# Parte 5: T√©cnicas Avanzadas
cuadrados_comp = {x: x**2 for x in range(5)}
frutas_list = ["manzana", "banana", "cereza"]
stock_comp = {fruta: 10 for fruta in frutas_list}
output_dict_p5_comprehension = (f"Cuadrados de 0 a 4: {cuadrados_comp}\n"
                                f"Stock creado desde una lista: {stock_comp}")

default_config = {"theme": "dark", "lang": "es"}
user_config = {"lang": "en", "user": "root"}
final_config = default_config | user_config
output_dict_p5_merge = (f"Configuraci√≥n por defecto: {default_config}\n"
                        f"Configuraci√≥n de usuario: {user_config}\n"
                        f"Resultado de la fusi√≥n (default | user): {final_config}")

# ==============================================================================
# 3. INYECTAR LOS RESULTADOS EN LA PLANTILLA DE DICCIONARIOS
# ==============================================================================
final_html_dict = html_template_dict.format(
    output_dict_p1_creacion=output_dict_p1_creacion,
    output_dict_p2_acceso=output_dict_p2_acceso,
    output_dict_p2_modificacion=output_dict_p2_modificacion,
    output_dict_p3_del=output_dict_p3_del,
    output_dict_p3_pop=output_dict_p3_pop,
    output_dict_p4_iteracion=output_dict_p4_iteracion,
    output_dict_p5_comprehension=output_dict_p5_comprehension,
    output_dict_p5_merge=output_dict_p5_merge,
)

# Renderizar el resultado en el notebook.
display(HTML(final_html_dict))
```

```python
# Celda 1
print("--- Creaci√≥n de Diccionarios ---")

# 1. Usando llaves {} (la forma m√°s com√∫n)
dict_vacio = {}
usuario = {"nombre": "Alan", "profesion": "Cient√≠fico de Datos", "edad": 41}

# 2. Usando el constructor dict() (√∫til con argumentos de palabra clave)
configuracion = dict(host="localhost", puerto=5432, debug=False)

# Las claves deben ser de un tipo inmutable (string, n√∫mero, tupla)
# El valor puede ser de cualquier tipo de dato.
dict_mixto = {
    "texto": "Hola",
    123: "es un n√∫mero",
    (1, 2): "es una tupla",
    "lista_valores": [10, 20, 30]
}

print(f"Un diccionario vac√≠o: {dict_vacio}")
print(f"El tipo de dato es: {type(dict_vacio)}")
print(f"Un diccionario de usuario: {usuario}")
print(f"Un diccionario creado con dict(): {configuracion}")
print(f"Un diccionario con claves mixtas: {dict_mixto}")
```

```python
# Celda 2
print("--- Acceso a Datos en un Diccionario ---")

usuario = {"nombre": "Grace", "profesion": "Almirante", "id": 1}

# 1. Acceso usando la clave entre corchetes []
nombre_usuario = usuario["nombre"]
print(f"Acceso directo a la clave 'nombre': {nombre_usuario}")

# 2. El m√©todo seguro .get()
#    Si la clave no existe, devuelve None en lugar de un error.
profesion_usuario = usuario.get("profesion")
print(f"Acceso seguro con .get('profesion'): {profesion_usuario}")

#    .get() tambi√©n puede recibir un valor por defecto si la clave no se encuentra.
pais_usuario = usuario.get("pais", "Desconocido")
print(f"Acceso a una clave inexistente ('pais') con valor por defecto: {pais_usuario}")

# 3. ¬øQu√© pasa si la clave no existe con el acceso directo? ¬°KeyError!
try:
    email = usuario["email"]
except KeyError as e:
    print(f"\nIntentar acceder a 'email' con [] causa un error: {type(e).__name__}: {e}")
```

```python
# Celda 3
print("--- Modificar y A√±adir Datos ---")

coche = {"marca": "Tesla", "modelo": "Model 3", "a√±o": 2020}
print(f"Diccionario original: {coche}")

# 1. Modificar el valor de una clave existente
coche["a√±o"] = 2021
print(f"Tras modificar el 'a√±o': {coche}")

# 2. A√±adir un nuevo par clave-valor
#    Si la clave no existe, se crea autom√°ticamente.
coche["color"] = "Rojo"
print(f"Tras a√±adir el 'color': {coche}")

# 3. Usando el m√©todo .update() para a√±adir/modificar varios pares a la vez
coche.update({"autonomia": 500, "color": "Rojo Perla"})
print(f"Tras usar .update(): {coche}")
```

```python
# Celda 4
print("--- Eliminar Datos ---")

config = {"host": "localhost", "port": 8080, "debug": True, "user": "admin"}
print(f"Diccionario original: {config}")

# 1. Usando la declaraci√≥n `del`
#    Elimina el par clave-valor. No devuelve nada.
del config["debug"]
print(f"Tras 'del config[\"debug\"]': {config}")

# 2. Usando el m√©todo .pop()
#    Elimina el par y DEVUELVE el valor asociado.
usuario_eliminado = config.pop("user")
print(f"Tras 'config.pop(\"user\")': {config}")
print(f"El valor devuelto por .pop() fue: '{usuario_eliminado}'")
```

```python
# Celda 5
print("--- Iteraci√≥n: Recorriendo un Diccionario ---")

inventario = {"manzanas": 50, "bananas": 25, "naranjas": 100}

# 1. Iterar sobre las claves (comportamiento por defecto)
print("\n1. Iterando sobre las claves:")
for clave in inventario:
    print(f"  Clave: {clave} -> Valor: {inventario[clave]}")

# 2. Iterar sobre los valores usando .values()
print("\n2. Iterando sobre los valores:")
for valor in inventario.values():
    print(f"  - {valor}")

# 3. La forma m√°s Pythonic: iterar sobre claves y valores con .items()
print("\n3. Iterando sobre pares (clave, valor) con .items():")
for fruta, cantidad in inventario.items():
    print(f"  Hay {cantidad} unidades de {fruta}.")
```

```python
# Celda 6
print("--- T√©cnicas √ötiles y Avanzadas ---")

# 1. Comprobar si una clave existe con el operador `in`
permisos = {"admin": True, "editor": True, "lector": False}
print(f"Diccionario de permisos: {permisos}")
if "admin" in permisos:
    print("La clave 'admin' S√ç existe en el diccionario.")
if "invitado" not in permisos:
    print("La clave 'invitado' NO existe en el diccionario.")


# 2. Dict Comprehensions: crear diccionarios de forma concisa
cuadrados = {x: x**2 for x in range(1, 6)}
print(f"\nDiccionario de cuadrados (creado con comprehension): {cuadrados}")


# 3. Fusionar diccionarios (Python 3.9+)
config_base = {"host": "localhost", "theme": "dark"}
config_usuario = {"theme": "light", "user": "ana"}
# El operador | fusiona. Si hay claves duplicadas, la de la derecha gana.
config_final = config_base | config_usuario
print("\nFusionando diccionarios:")
print(f"  Base:   {config_base}")
print(f"  Usuario:{config_usuario}")
print(f"  Final:  {config_final}")
```
